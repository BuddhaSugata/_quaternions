## 6. Статическая гравитация: решение Шварцшильда в униметрии

В униметрии гравитация статического сферически-симметричного поля моделируется как **фоновый вклад** во внутренний угол $\zeta$, т.е. в меру перепараметризации между фазой и собственным временем. Это позволяет получить стандартные формулы Шварцшильда в чисто «фазовой» геометрии, сохраняя связность и когерентность со СТО.

### 6.1. Калибр времени и фоновый угол
Общий Якобиан перепараметризации (§3.5) разлагаем на внутреннюю и внешнюю части:
$$
\frac{d\tau}{d\chi}=\cos\zeta(x)\,\cos\theta(x),\qquad
\zeta(x)=\zeta_{\rm int}\oplus\zeta_{\rm bg}(x),
\tag{6.1}
$$
где $\zeta_{\rm int}$ — постоянный для данного объекта (его «собственный образ»), а $\zeta_{\rm bg}(x)$ — вклад среды/фона.

Для статического сферического поля используем метрику Шварцшильда
$$
ds^2=f(r)\,\mathtt{c}^2 dt^2-f(r)^{-1}dr^2-r^2 d\Omega^2,\qquad
f(r)=1-\frac{r_s}{r}.
\tag{6.2}
$$
Естественная идентификация фонового угла:
$$
\boxed{\ \cos\zeta_{\rm bg}(r)=\sqrt{f(r)}=\sqrt{1-\frac{r_s}{r}}\ }\!,
\tag{6.3}
$$
так что «гравитационный множитель» есть $\gamma_{\rm g}(r):=1/\cos\zeta_{\rm bg}(r)=1/\sqrt{f(r)}$.

### 6.2. Замедление времени и метрические коэффициенты
Для статического наблюдателя ($\theta=0$) из \eqref{6.1} и \eqref{6.3}:
$$
\frac{d\tau}{dt}=\cos\zeta_{\rm int}\,\sqrt{f(r)}.
\tag{6.4}
$$
Выбирая калибр стандартных часов $\cos\zeta_{\rm int}=1$, получаем учебное $d\tau/dt=\sqrt{g_{tt}}$.

Если локально есть движение со скоростью $v$ относительно статического наблюдателя, то $\cos\theta=1/\gamma_{\rm kin}$ и
$$
\frac{d\tau}{dt}=\frac{\sqrt{f(r)}}{\gamma_{\rm kin}},
\qquad
\gamma_{\rm kin}=\frac{1}{\sqrt{1-\beta^2}},\ \beta=\frac{v}{\mathtt{c}}.
\tag{6.5}
$$
Это ровно стандартная композиция гравитационного и кинематического факторов.

### 6.3. Радиальная шкала и светоподобные траектории
Из \eqref{6.2} следует $g_{rr}=f(r)^{-1}$. В униметрии это реализуется выбором фонового пространственного масштабирования
$$
L'_{\rm bg}(r)=\frac{1}{\cos\zeta_{\rm bg}(r)}=\frac{1}{\sqrt{f(r)}},
\tag{6.6}
$$
что обеспечивает согласование с $g_{rr}$ при $dl=L'_{\rm bg}\,dr$.

Для луча (светоподобной траектории) условие $dx_0=dx_l$ с учётом \eqref{6.4}–\eqref{6.6} даёт
$$
\mathtt{c}\,\sqrt{f(r)}\,dt=\frac{dr}{\sqrt{f(r)}}\quad\Rightarrow\quad
\boxed{\ \frac{dr}{dt}=\mathtt{c}\,f(r)=\mathtt{c}\!\left(1-\frac{r_s}{r}\right)\ }\!,
\tag{6.7}
$$
т.е. координатная скорость света в координатах Шварцшильда.

### 6.4. Гравитационный красный сдвиг и композиция с кинематикой
Две статические радиальные позиции $r_{\rm e}\to r_{\rm o}$ дают
$$
\boxed{\ \frac{\nu_{\rm obs}}{\nu_{\rm src}}
=\frac{\cos\zeta_{\rm bg}(r_{\rm e})}{\cos\zeta_{\rm bg}(r_{\rm o})}
=\sqrt{\frac{f(r_{\rm e})}{f(r_{\rm o})}}\ }\!,
\tag{6.8}
$$
а при наличии локальных скоростей множится на СТО-фактор (например, продольный $e^{\pm\eta_{\rm kin}}$ или общий $\gamma_{\rm kin}(1-\beta\cos\varphi)$ в системе наблюдателя). В лог-форме композиция прозрачна:
$$
\ln\frac{\nu_{\rm obs}}{\nu_{\rm src}}
=\underbrace{\tfrac12\ln f(r_{\rm e})-\tfrac12\ln f(r_{\rm o})}_{\text{гравитация}}
\ \pm\ \eta_{\rm kin}\ \ (\text{или эквивалентный SR-фактор}).
\tag{6.9}
$$

### 6.5. Слабопольный предел и «фазовый потенциал»
Введём «фазовый потенциал» $\Phi(r)$:
$$
\cos\zeta_{\rm bg}(r)=e^{\Phi(r)},\qquad
\Phi(r)=\tfrac12\ln f(r)=\tfrac12\ln\!\left(1-\frac{r_s}{r}\right).
\tag{6.10}
$$
Тогда гравитационный сдвиг — просто $e^{\Phi(r_{\rm e})-\Phi(r_{\rm o})}$. В слабом поле $r\gg r_s$:
$$
\Phi(r)\simeq -\frac{GM}{r\,\mathtt{c}^2}
\quad\Rightarrow\quad
\frac{\Delta\nu}{\nu}\simeq\frac{\Delta\Phi_N}{\mathtt{c}^2},
\tag{6.11}
$$
что воспроизводит ньютоновскую аппроксимацию.

### 6.6. Замечание о калибрах
Идентификация $\cos\zeta_{\rm bg}=\sqrt{g_{tt}}$ относится к **временнo-ортогональным** статическим координатам (Шварцшильд). В других калибрах (Painlevé–Gullstrand, Эддингтон–Финкельштейн) та же физика перераспределяется между временной и смешанной частями метрики; в униметрии это эквивалентно иному разбиению фонового фактора между $\zeta_{\rm bg}$ и «эффективной» кинематикой. Наблюдаемые (частоты, задержки, скорость луча по локальным линейкам) остаются теми же.

### 6.7. Резюме
Статическая гравитация в униметрии — это фоновый внутренний поворот $\zeta_{\rm bg}(r)$ с $\cos\zeta_{\rm bg}=\sqrt{f(r)}$. Тогда:
- $d\tau/dt=\sqrt{f(r)}/\gamma_{\rm kin}$ — замедление времени;
- $dl=L'_{\rm bg}\,dr$ с $L'_{\rm bg}=1/\sqrt{f(r)}$ — радиальная шкала;
- $dr/dt=\mathtt{c} f(r)$ — координатная скорость света;
- $\nu_{\rm obs}/\nu_{\rm src}=\sqrt{f(r_{\rm e})/f(r_{\rm o})}$ — гравитационный сдвиг.

Все формулы возникают из одной схемы перепараметризации \eqref{6.1} и выбора калибра \eqref{6.3}, т.е. как чистая геометрия фазы, без введения новых законов.


## 7. От униметрии к уравнениям ОТО

Цель этого раздела — показать, как «фазовые» поля униметрии порождают произвольную (классическую) метрику и как стандартные уравнения Эйнштейна следуют из вариационного принципа, не вводя новой физики. Мы зададим словарь «фаза → метрика» через тетрады, а затем используем привычные геометрические конструкции (связность, кривизна, действие Эйнштейна–Гильберта).

### 7.1. Словарь фаз → тетрады → метрика

В униметрии локальное перепараметризование задаётся
$$
\frac{d\tau}{d\chi}=\cos\zeta(x)\,\cos\theta(x),\qquad
\zeta(x)=\zeta_{\rm int}\oplus\zeta_{\rm bg}(x),
\tag{7.1}
$$
где $\zeta$ — «внутренний» поворот (с фоновым вкладом), а $\theta$ — «внешний» (кинетический) поворот; см. §3.5. В кривом случае эти углы становятся полями на многообразии.

**Идея.** Построим ортонормированный кофрейм $e^A$ ($A=0,1,2,3$) из «фазовых» калибров:
- *временной масштаб (lapse)*: $N(x):=\cos\zeta(x)$;
- *локальный буст* по направлению единичного вектора $n^i(x)$ с «скоростью» $\beta(x)=\sin\theta(x)$, т.е. с быстротой $\eta=\operatorname{artanh}\beta$;
- *пространственная ориентация/масштаб* через триаду $E^a{}_i(x)$ ($a=1,2,3$), $\;h_{ij}=\delta_{ab}E^a{}_iE^b{}_j$.

Базовый (нелокально-вращённый) кофрейм:
$$
\bar e^{\,0}:=\mathtt{c}\,N\,d\tau,\qquad
\bar e^{\,a}:=E^a{}_i\,dx^i.
\tag{7.2}
$$
Локальный Лоренц-поворот $\Lambda^A{}_B(x)$, составленный из буста вдоль $n^i$ (параметр $\eta$) и трёхмерной ротации $R^a{}_b(x)$, даёт физический кофрейм
$$
e^A(x):=\Lambda^A{}_B(x)\,\bar e^{\,B}(x),\qquad
g=\eta_{AB}\,e^A\!\otimes e^B,\quad \eta_{AB}=\mathrm{diag}(+1,-1,-1,-1).
\tag{7.3}
$$
Тем самым «фазовые» поля $(N,\eta,n^i,R,E)$ однозначно задают метрику $g_{\mu\nu}$.

> В статике $N=\cos\zeta_{\rm bg}=\sqrt{g_{tt}}$ и $E$ фиксирует $h_{ij}$; см. §6 (Шварцшильд).

### 7.2. Эквивалентные формы: ADM и тетрады

Из (7.3) следует стандартная ADM-форма
$$
ds^2=\bigl(N^2-h_{ij}N^iN^j\bigr)\mathtt{c}^2 d\tau^2
-2h_{ij}N^i dx^j\,\mathtt{c} d\tau-h_{ij}dx^i dx^j,
\tag{7.4}
$$
где «сдвиг» $N^i$ выражается через локальный буст ($\eta,n^i$) и выбор срезов $\tau=\text{const}$ (это чисто кинематическое соответствие: фазовый буст кодируется как ADM-сдвиг в данном калибре времени). Величины $N,h_{ij},N^i$ — это просто иные координаты на многообразии метрик, эквивалентные $e^A$.

### 7.3. Геометрия: связность, кривизна, действие

Выбираем нулевую кручение (Леви–Чивита), что фиксирует спин-связность $\omega^A{}_B=\omega^A{}_{B\mu}dx^\mu$ по первому уравнению структуры
$$
de^A+\omega^A{}_B\wedge e^B=0\quad(\text{torsion}=0).
\tag{7.5}
$$
Кривизна — 2-форма
$$
R^A{}_B:=d\omega^A{}_B+\omega^A{}_C\wedge\omega^C{}_B.
\tag{7.6}
$$
Действие Эйнштейна–Гильберта в тетрадной форме (Палатино при нулевой кручении):
$$
S[e]=\frac{1}{2\kappa}\int \epsilon_{ABCD}\,e^A\wedge e^B\wedge R^{CD}
\;+\;S_{\rm matter}[e,\Psi],\qquad \kappa=\frac{8\pi G}{\mathtt{c}^4}.
\tag{7.7}
$$
Вариация по $e^A$ (и по $\omega$ при условии нулевой кручения) даёт уравнения Эйнштейна
$$
G_{\mu\nu}=\kappa\,T_{\mu\nu},\qquad
G_{\mu\nu}:=R_{\mu\nu}-\tfrac12 R\,g_{\mu\nu}.
\tag{7.8}
$$

**Вывод.** Как только «фазовые» поля $(N,\eta,n^i,R,E)$ определили тетрадный кофрейм $e^A$, дальше всё — стандартная ОТО. Динамика этих полей — ничто иное, как уравнения Эйнштейна для индуцированной метрики.

### 7.4. Проверка на частных случаях

- **Шварцшильд (статическое поле).** Выбор $N(r)=\sqrt{1-r_s/r}$, $E^a{}_i=\mathrm{diag}\bigl((1-r_s/r)^{-1/2},\,r,\,r\sin\theta\bigr)$, $N^i=0$ и отсутствие материи ($T_{\mu\nu}=0$) решает (7.8); см. §6 для интерпретации $N$ как $\cos\zeta_{\rm bg}$.
- **Чистый SR.** Константные поля $N\equiv1$, $E^a{}_i\equiv\delta^a{}_i$, произвольные глобальные бусты/ротации $\Lambda\in SO^+(1,3)$ дают плоскую метрику.

### 7.5. «Фазовый» взгляд на кривизну

Кривизна исходит из пространственно-временной вариации локальной Лоренц-матрицы $\Lambda(x)$ и масштабов $(N,E)$:
- Градиенты $\nabla N=-\nabla\ln\cos\zeta$ отражают **временную** часть гравитационного потенциала (см. §6.5).
- Производные $E^a{}_i$ и ротации $R^a{}_b(x)$ формируют **пространственную** кривизну $\,{}^{(3)}\!R[h]$.
- Поле буста $(\eta,n)$ даёт «смешанные» компоненты (экструзия срезов), т.е. экстремальную кривизну (экстратензор $K_{ij}$ в ADM).

В координатно-независимой форме всё это кодируется в $e^A$ и $\omega^A{}_B$ и приводит к стандартному тензору Эйнштейна $G_{\mu\nu}$.

### 7.6. Альтернатива: ADM-динамика в «фазовом» словаре

Эквивалентно можно переписать действие (7.7) в ADM-форме:
$$
S=\frac{1}{2\kappa}\int d\tau\,d^3x\,N\sqrt{h}\,\bigl({}^{(3)}\!R+K^2-K_{ij}K^{ij}\bigr)+S_{\rm matter},
\tag{7.9}
$$
где $K_{ij}=\tfrac1{2N}\bigl(\dot h_{ij}-D_iN_j-D_jN_i\bigr)$. Тогда уравнения Эйнштейна эквивалентны системе ограничений и эволюции:
\[
\begin{aligned}
&\text{(гамильтоново)}\quad {}^{(3)}\!R+K^2-K_{ij}K^{ij}=2\kappa\,\rho,\\
&\text{(импульсное)}\quad D_j(K^{ij}-h^{ij}K)=\kappa\,j^i,\\
&\text{(эволюция)}\quad \dot h_{ij}=-2N K_{ij}+D_iN_j+D_jN_i,\ \dots
\end{aligned}
\tag{7.10}
\]
а «фазовый» словарь — это просто
$$
N=\cos\zeta,\qquad
(N^i,\ h_{ij})\ \text{из}\ (\eta,n^i,E^a{}_i).
\tag{7.11}
$$
Т.е. униметрия даёт *параметризацию* переменных ОТО, а не модификацию динамики.

### 7.7. Концептуальное резюме

1) Униметрия предлагает **единый язык**: локальная геометрия — это поле локальных Лоренц-поворотов $\Lambda(x)$ и масштабов времени/пространства $(N,E)$, производных из фазовых углов $(\zeta,\theta)$ и ориентаций.  
2) Из этих полей строится тетрадный кофрейм $e^A$ и метрика $g_{\mu\nu}$ по (7.2)–(7.3).  
3) Вся динамика — это обычные уравнения Эйнштейна для $g_{\mu\nu}$, получаемые из (7.7) или (7.9).  
4) Частные решения (Шварцшильд, FRW и др.) возникают как специальные профили $N,E,\Lambda$; их «фазовая» интерпретация согласуется со знакомыми физическими факторами (замедление времени, «лоренцоподобные» множители и т.п.).

> **Смысловая выгода.** Мы не заменяем ОТО, а предоставляем связную параметризацию, в которой гравитационные и кинематические эффекты — разные аспекты одной конструкции перепараметризации ($\chi\!\to\!\tau$) и локальных Лоренц-поворотов.


## 8. Локальная цикличность времени и «массовый» параметр

В этом разделе мы **разделяем** внутренний угол на фоновый и локальный вклады и вводим гипотезу о **циклической природе локального времени**.

### 8.1. Разделение внутреннего угла
Пусть внутренний угол складывается из двух независимых факторов — фонового гравитационного и локального:
$$
\cos\zeta(x)=\cos\zeta_{\rm bg}(x)\,\cos\zeta_{\rm loc},
\qquad\Rightarrow\qquad
\frac{d\tau}{d\chi}=\cos\zeta_{\rm bg}(x)\,\cos\zeta_{\rm loc}\,\cos\theta(x).
\tag{8.1}
$$
Здесь $\zeta_{\rm bg}(x)$ кодирует «среду» (напр., $\cos\zeta_{\rm bg}=\sqrt{g_{tt}}$ в статике Шварцшильда; см. §6), а $\zeta_{\rm loc}$ — **внутренний локальный вклад объекта** (не «собственный» во избежание двусмысленности с собственным временем).

### 8.2. Гипотеза: локальное время как циклическое действие
Предположим, что локальное течение времени объекта возникает как периодическое действие: ось $x_0$ геометрически замкнута и её локальная модель — окружности $S^1_\chi$ и $S^1_{x_0}$ с радиусами $R_1$ и $R_2$ в соответствующих осях. За один цикл:
$$
\Delta\chi=2\pi R_1,\qquad
\Delta\tau=\frac{2\pi R_2}{\dot{H}},
\tag{8.2}
$$
где $\dot{H}:=dx_0/d\tau$ — темпоральная фазовая скорость в наблюдаемом времени (в наших калибрах §3.3 — это просто $\mathtt{c}$).

Частота в высокочастотном пределе (когда $R_1\to0$ как размер фазового шага) равна
$$
\nu:=\frac{\Delta\chi}{\Delta\tau}
=\frac{2\pi R_1}{2\pi R_2/\dot{H}}
=\underbrace{\frac{R_1}{R_2}}_{=:k}\,\dot{H},
\qquad
\Rightarrow\qquad
\boxed{\ \dot{\chi}=\frac{d\chi}{d\tau}=k\,\dot{H}\ }.
\tag{8.3}
$$
Таким образом, параметр $k:=R_1/R_2$ — **безразмерное геометрическое отношение** двух окружностей, а $\dot{\chi}$ — «скорость» фазы в собственном времени.

### 8.3. Наблюдаемая частота в поле и при движении
С учётом (8.1) наблюдаемая частота у удалённого наблюдателя получается из «голой» $\nu_{\rm loc}:=k\,\dot{H}$ умножением на Якобиан:
$$
\nu_{\rm obs}
=\frac{\nu_{\rm loc}}{\cos\zeta_{\rm bg}(x)\,\cos\theta(x)}
=\frac{k\,\dot{H}}{\cos\zeta_{\rm bg}\,\cos\theta}.
\tag{8.4}
$$
В статике и без движения ($\theta=0$) имеем $\nu_{\rm obs}=k\,\dot{H}/\cos\zeta_{\rm bg}$; в асимптотически плоской области $\cos\zeta_{\rm bg}\to1$ и при калибровке $\dot{H}\equiv\mathtt{c}$ — $\nu_{\rm obs}=k\,\mathtt{c}$.

### 8.4. От частоты к энергии и «появление» $\mathtt{c}^2$
Используя соотношение Планка $E=\hbar\nu$, для объекта в покое вдали от гравитационных источников:
$$
E_0=\hbar\,\nu_{\rm obs}=\hbar\,k\,\dot{H}
\ \stackrel{\dot{H}=\mathtt{c}}{=}\ 
\hbar\,k\,\mathtt{c}.
\tag{8.5}
$$
Если идентифицировать покоящуюся энергию с $E_0=m\,\mathtt{c}^2$, получаем
$$
\boxed{\ m=\frac{\hbar}{\mathtt{c}}\;k\ }.
\tag{8.6}
$$
То есть **масса пропорциональна** геометрическому отношению $k=R_1/R_2$ (с точностью до универсального масштаба $\hbar/\mathtt{c}$). С этой точки зрения «квадрат скорости света» в $E_0=m\mathtt{c}^2$ трактуется как композиция:
- один множитель $\mathtt{c}$ приходит из темпоральной калибровки $\dot{H}\equiv\mathtt{c}$;
- второй — из стандартного соотношения энергии покоя $E_0=m\mathtt{c}^2$.
Комбинация делает массу линейной функцией геометрического $k$.

> Замечание. Мы не утверждаем, что (8.6) — единственная возможная идентификация. Это **операциональная** привязка к наблюдаемым $E_0$ и $\nu$ в пределе слабого поля. В других калибровках роль $k$ может сыграть переопределённая функция от $\zeta_{\rm loc}$, см. ниже.

### 8.5. Связь с §5.3: масса и внутренний локальный угол
В §5.3 масса связывалась с неустранимой вещественной проекцией, т.е. с $\cos\zeta\neq0$. Разделяя фон и локаль по (8.1) и исходя из безмассового предела ($\zeta\to\pm\pi/2$), естественно требовать
$$
k=k(\zeta_{\rm loc}),\qquad
k(\pm\pi/2)=0,\qquad k(0)>0.
\tag{8.7}
$$
Минимальный (линейный по косинусу) анзац:
$$
k(\zeta_{\rm loc})=\kappa_0\,\cos\zeta_{\rm loc},
\qquad\Rightarrow\qquad
m=\frac{\hbar}{\mathtt{c}}\;\kappa_0\,\cos\zeta_{\rm loc}.
\tag{8.8}
$$
Тогда безмассовые состояния соответствуют $\cos\zeta_{\rm loc}=0$ (фотонное подпространство), а массивные — $|\zeta_{\rm loc}|<\pi/2$, что согласуется с §5.3. Параметр $\kappa_0$ задаёт **масштаб** (можно интерпретировать как эффективную связь с механизмом массогенеза).

### 8.6. Совместимость с гравитацией (фон) и доплером (кинетика)
Комбинируя (8.4) и (8.8), наблюдаемая частота при наличии поля и движения:
$$
\nu_{\rm obs}
=\frac{\kappa_0\,\cos\zeta_{\rm loc}}{\cos\zeta_{\rm bg}(x)\,\cos\theta(x)}\,\dot{H},
\qquad
\Rightarrow\qquad
\frac{\nu_{\rm obs}}{\nu_{\rm src}}
=\frac{\cos\zeta_{\rm loc,src}}{\cos\zeta_{\rm loc,obs}}\,
\frac{\cos\zeta_{\rm bg,src}}{\cos\zeta_{\rm bg,obs}}\,
\frac{\cos\theta_{\rm src}}{\cos\theta_{\rm obs}}.
\tag{8.9}
$$
В стандартных постановках (один и тот же объект, сравнение двух уровней поля и/или движения) остаются ровно известные факторы: гравитационный (через $\cos\zeta_{\rm bg}$, см. §6) и кинематический (через $\cos\theta$, см. §3.8).

### 8.7. Резюме главы
1) Внутренний угол **разделён** на фоновый и локальный вклады: $\cos\zeta=\cos\zeta_{\rm bg}\cos\zeta_{\rm loc}$.  
2) Локальное время **моделируется как циклическое действие**, где отношение радиусов $k=R_1/R_2$ задаёт «голую» частоту $\nu_{\rm loc}=k\,\dot{H}$.  
3) **Массовый параметр** можно ввести как $m=(\hbar/\mathtt{c})\,k$, а при выборе $k(\zeta_{\rm loc})\propto\cos\zeta_{\rm loc}$ он согласуется с критерием §5.3 (масса $\leftrightarrow$ ненулевая вещественная проекция).  
4) Гравитационные и кинематические эффекты остаются **мультипликативными** коррекциями (через $\zeta_{\rm bg}$ и $\theta$), сохраняя когерентность с §3 и §6.

> Эта глава задаёт связную интерпретацию: $E_0=\hbar\nu$ и $E_0=m\mathtt{c}^2$ согласуются, если $m\propto k$ и $\dot{H}\equiv\mathtt{c}$. Тогда «$\mathtt{c}^2$» в энергии покоя мыслится как композиция темпоральной калибровки и стандартной связи масса–энергия, а сама масса — как геометрический параметр локальной цикличности.


# Статическая гравитация и локальная цикличность времени в фазовой трактовке (униметрия)

## Аннотация
Мы даём самодостаточное изложение двух идей униметрии: (i) как статическая гравитация описывается через «фазовый» поворот внутренней оси времени, воспроизводя решения типа Шварцшильда; (ii) как локальная цикличность собственных «часов» объекта (в духе де Бройля) естественно встраивается в тот же язык и даёт простой геометрический параметр массы. Оба сюжета построены на одном тождестве перепараметризации: наблюдаемое время получается из скрытого фазового параметра путём локального масштабирования.

---

## 1. Фазовая кинематика: минимальный словарь

Пусть скрытый параметр — **фаза** $\chi$. Наблюдаемое собственное время $\tau$ возникает перепараметризацией
$$
\frac{d\tau}{d\chi}=\cos\zeta(x)\,\cos\theta(x),
\qquad \cos\zeta:=\cos\zeta_{\rm bg}(x)\,\cos\zeta_{\rm loc}.
\tag{1.1}
$$
Здесь:
- $\zeta_{\rm bg}(x)$ — **фоновый** внутренний угол (среда/гравитация),
- $\zeta_{\rm loc}$ — **локальный** внутренний угол объекта (его «образ»),
- $\theta(x)$ — **внешний** угол, связанный с относительным движением ($\beta=\sin\theta$, $\gamma=\sec\theta$).

Темпоральная фазовая скорость в собственном времени:
$$
\dot{H}:=\frac{dx_0}{d\tau}.
\tag{1.2}
$$
Мы используем калибровку
$$
\boxed{\ \dot{H}\equiv \mathtt{c}\ } \quad\text{(темпоральная шкала)}.
\tag{1.3}
$$
В статических временно-ортогональных координатах удобно фиксировать
$$
\boxed{\ \cos\zeta_{\rm bg}(x)=\sqrt{g_{tt}(x)}\ },
\tag{1.4}
$$
что просто означает: «гравитация» входит как фоновый вклад в меру времени.

---

## 2. Шварцшильд в фазовых величинах

Для статического сферического поля
$$
ds^2=f(r)\,\mathtt{c}^2dt^2-f(r)^{-1}dr^2-r^2d\Omega^2,
\qquad f(r)=1-\frac{r_s}{r}.
\tag{2.1}
$$
Идентификация фонового угла:
$$
\boxed{\ \cos\zeta_{\rm bg}(r)=\sqrt{f(r)}=\sqrt{1-\frac{r_s}{r}}\ }.
\tag{2.2}
$$

### 2.1. Замедление времени и композиция с движением
Из$(1.1)$, при локальной скорости $v$ относительно статического наблюдателя ($\cos\theta=1/\gamma_{\rm kin}$):
$$
\frac{d\tau}{dt}=\cos\zeta_{\rm bg}(r)\,\cos\theta
=\frac{\sqrt{f(r)}}{\gamma_{\rm kin}},
\qquad \gamma_{\rm kin}=\frac{1}{\sqrt{1-\beta^2}},\ \beta=\frac{v}{\mathtt{c}}.
\tag{2.3}
$$
В покое ($v=0$) получаем учебное $d\tau/dt=\sqrt{g_{tt}}$.

### 2.2. Радиальная шкала и светоподобные траектории
Согласование с$g_{rr}=f^{-1}$ достигается выбором фонового пространственного масштаба
$$
L'_{\rm bg}(r)=\frac{1}{\cos\zeta_{\rm bg}(r)}=\frac{1}{\sqrt{f(r)}},
\quad dl=L'_{\rm bg}\,dr.
\tag{2.4}
$$
Для луча условие $dx_0=dx_l$ даёт
$$
\mathtt{c}\,\sqrt{f(r)}\,dt=\frac{dr}{\sqrt{f(r)}}
\ \Rightarrow\
\boxed{\ \frac{dr}{dt}=\mathtt{c}\,f(r)\ }.
\tag{2.5}
$$

### 2.3. Гравитационный красный сдвиг и его композиция с СТО
Для двух статических уровней $r_{\rm e}\to r_{\rm o}$:
$$
\boxed{\ \frac{\nu_{\rm obs}}{\nu_{\rm src}}=\sqrt{\frac{f(r_{\rm e})}{f(r_{\rm o})}}\ }.
\tag{2.6}
$$
Если есть относительное движение (угол$\varphi$ в системе наблюдателя), доплеровский множитель СТО просто домножается:
$$
\frac{\nu_{\rm obs}}{\nu_{\rm src}}
=\sqrt{\frac{f(r_{\rm e})}{f(r_{\rm o})}}\;\gamma_{\rm kin}\,(1-\beta\cos\varphi).
\tag{2.7}
$$
В логарифмах композиция становится суммой:
$$
\ln\frac{\nu_{\rm obs}}{\nu_{\rm src}}
=\tfrac12\ln f(r_{\rm e})-\tfrac12\ln f(r_{\rm o})
+\ln\bigl[\gamma_{\rm kin}(1-\beta\cos\varphi)\bigr].
\tag{2.8}
$$

> **Смысл.** Шварцшильдовский фактор$\sqrt{f}$ — это просто фоновый поворот внутренней оси времени: $\cos\zeta_{\rm bg}=\sqrt{g_{tt}}$. Кинематика СТО и гравитация «склеиваются» как независимые множители одной перепараметризации.

---

## 3. Локальная цикличность времени (де Бройль в униметрии)

Теперь вводим **локальную цикличность**: предположим, что собственное время объекта возникает как результат периодического действия — локально ось $x_0$ топологически замкнута. Модель двух окружностей: фазовой$S^1_\chi$ радиуса $R_1$ и темпоральной$S^1_{x_0}$ радиуса $R_2$. За цикл:
$$
\Delta\chi=2\pi R_1,\qquad \Delta\tau=\frac{2\pi R_2}{\dot{H}}.
\tag{3.1}
$$
Определим **угловую частоту** локального цикла:
$$
\omega:=\frac{d\chi}{d\tau}
=\lim_{\text{цикл}}\frac{\Delta\chi}{\Delta\tau}
=\underbrace{\frac{R_1}{R_2}}_{=:k}\,\dot{H},
\qquad \nu=\frac{\omega}{2\pi}.
\tag{3.2}
$$
В калибре$(1.3)$:
$$
\boxed{\ \omega=k\,\mathtt{c}\ }.
\tag{3.3}
$$

### 3.1. Энергия покоя и геометрический параметр массы
С $E=\hbar\omega$ и $E_0=m\mathtt{c}^2$ получаем
$$
\boxed{\ m=\frac{\hbar}{\mathtt{c}}\,k\ }.
\tag{3.4}
$$
То есть масса — **безразмерный геометрический коэффициент** локальной цикличности, умноженный на универсальный масштаб $\hbar/\mathtt{c}$. Концептуально это комтон-частота в геометрическом обличье: $\omega_c=E_0/\hbar=m\mathtt{c}^2/\hbar=k\,\mathtt{c}$.

### 3.2. Связь с внутренним локальным углом
Чтобы безмассовый предел соответствовал «чисто мнимой» ориентации внутренней оси ($\cos\zeta_{\rm loc}=0$), удобно положить
$$
k=k(\zeta_{\rm loc})=\kappa_0\,\cos\zeta_{\rm loc},
\quad k(\pm\tfrac{\pi}{2})=0,\ k(0)=\kappa_0>0,
\tag{3.5}
$$
и тогда
$$
m=\frac{\hbar}{\mathtt{c}}\,\kappa_0\,\cos\zeta_{\rm loc}.
\tag{3.6}
$$
Параметр$\kappa_0$ фиксирует «семейство» объекта (масштаб), а сама масса пропорциональна вещественной проекции внутренней фазы.

### 3.3. Что видит удалённый наблюдатель
Наблюдаемая угловая частота при наличии поля и движения получается из «голой» $\omega_{\rm loc}=k\,\dot{H}$ делением на Якобиан$(1.1)$:
$$
\omega_{\rm obs}
=\frac{\omega_{\rm loc}}{\cos\zeta_{\rm bg}(x)\,\cos\theta(x)}
=\frac{k\,\dot{H}}{\cos\zeta_{\rm bg}\,\cos\theta}.
\tag{3.7}
$$
Если сравниваются **две высоты**$r_{\rm e}\to r_{\rm o}$ одного и того же объекта (значит,$\zeta_{\rm loc}$ одинаков), то локальный вклад сокращается и остаётся ровно композиция §2: гравитация$\,\times\,$доплер.

> **Итог секции.** Идея де Бройля о «внутренних часах» естественна в униметрии: локальная цикличность — геометрия оси $x_0$; её частота линейно задаётся калибром$\dot{H}$ и отношением$k$, а масса — это $m=(\hbar/\mathtt{c})\,k$.

---

## 4. Обсуждение и обобщения

1) **Единый механизм перепараметризации.** И в гравитации, и в кинематике, и в «внутренних часах» работает одна формула$(1.1)$: фоновая часть$\cos\zeta_{\rm bg}$, локальная$\cos\zeta_{\rm loc}$, внешняя$\cos\theta$.

2) **Калибр координат.** Идентификация$\cos\zeta_{\rm bg}=\sqrt{g_{tt}}$ относится к статическим временно-ортогональным координатам (Шварцшильд). В иных калибрах та же физика распределяется между временной и смешанной частями метрики; наблюдаемые коэффициенты не меняются.

3) **Выход за пределы статической сферы.** Для произвольной статической метрики с $g_{tt}>0$ можно положить$\cos\zeta_{\rm bg}=\sqrt{g_{tt}}$ и повторить рассуждения; в нестационарных задачах естественно работать в тетрадах, трактуя вариации$\zeta_{\rm bg},\theta$ как поля (см. общий словарь униметрии).

4) **Ограничения и интерпретации.** Связь $m=(\hbar/\mathtt{c})\,k$ — не «новая физика», а удобная геометризация комтон-частоты: $k$ инвариантен вдоль свободного движения и может зависеть от класса объекта через$\kappa_0$. Безмассовые состояния соответствуют$\cos\zeta_{\rm loc}=0$.

---

## 5. Выводы

- Статическая гравитация в фазовом языке — это **фоновый внутренний поворот** оси времени: $\cos\zeta_{\rm bg}=\sqrt{g_{tt}}$. Отсюда сразу следуют: $d\tau/dt=\sqrt{f}/\gamma_{\rm kin}$, $dr/dt=\mathtt{c}f$, $\nu_{\rm obs}/\nu_{\rm src}=\sqrt{f_{\rm e}/f_{\rm o}}$ и их композиция с доплером.

- Локальная цикличность в духе де Бройля встраивается в тот же каркас: $\omega=k\,\dot{H}$, при $\dot{H}\equiv\mathtt{c}$ получаем $E=\hbar\omega=\hbar k\,\mathtt{c}$ и $m=(\hbar/\mathtt{c})\,k$. Масса — геометрический параметр локальной фазы, согласованный с безмассовым пределом через$\zeta_{\rm loc}$.

- Одна перепараметризация объясняет «гравитацию, доплер и внутренние часы» как разные проекции и калибры одного фазового потока. Это даёт связанную концептуальную картину без добавления новых законов.
