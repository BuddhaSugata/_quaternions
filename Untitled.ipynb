Положим, существует наблюдатель в системе отсчёта S. Его обстоятельства классически (СТО) выражаются интервалом:
$$ s^2 = \mathtt{c}^2 t^2 - l^2 $$
Что позволяет определить величину собственного интервала покоя - самого наблюдателя:
$$ s^2 = \mathtt{c}^2 t^2 $$
Теперь положим существует второй наблюдатель:
$$ s'^2 = \mathtt{c}^2 t'^2 $$
Наличие относительного движения двух систем в терминах пространства первого наблюдателя будет:
$$ \mathtt{c}^2 t'^2 = \mathtt{c}^2 t^2 - l^2 $$
и напротив:
$$ \mathtt{c}^2 t^2 = \mathtt{c}^2 t'^2 - l'^2 $$
Откуда подстановкой следует, что:
$$ l'^2 = - l^2 $$
то есть в одной из систем отсчёта расстояние получилось мнимым относительно другой. Как это противоречие обходится в СТО?

Наличие относительного движения двух систем в терминах пространства первого наблюдателя будет:
$$ \mathtt{c}^2 t'^2 = \mathtt{c}^2 t^2 - l^2 $$

и напротив:
$$ \mathtt{c}^2 t^2 = \mathtt{c}^2 t'^2 - l'^2 $$

Откуда подстановкой следует, что:
$$ \mathtt{c}^2 t'^2 = \mathtt{c}^2 t'^2 - l'^2 - l^2 $$
