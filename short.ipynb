# Униметрия: Прото-функция и фазовое пространство

Мы воспринимаем реальность как данность, как «нулевой уровень дериватива».  
Но что если эта реальность вторична, а фундаментальной величиной является **фаза** \(\chi\)?  

Униметрия рассматривает **время и пространство как производные** от единого протопараметра.  
Метрика пространства-времени представляется как следствие изменения этого параметра, а не как самостоятельная сущность.  
В этом формализме эффекты СТО и релятивистский Доплер естественно вытекают из геометрии фазового вектора.

---

## 1. Время и пространство через прото-фазу

Предположим, что существует переменная \(\chi\), изменение которой порождает наблюдаемые эффекты времени и пространства.  
Введем проекции потока на временную (\(\tilde{S}\)) и пространственную (\(\tilde{L}\)) компоненты:

\[
dx_\mathcal{h} = \frac{d H}{d\chi} d\chi = \tilde{H} d\chi, \quad
dx_\mathcal{l} = \frac{d L}{d\chi} d\chi = \tilde{L} d\chi
\]

Интервал пространства-времени выражается через норму потока:

\[
\tilde{S}^2 = \tilde{H}^2 - \tilde{L}^2 \quad \Rightarrow \quad 
\tilde{\mathcal{H}}^2 = \tilde{S}^2 + \tilde{L}^2
\]

- \(\tilde{H}\) — полная прото-скорость  
- \(\tilde{S}\) — временная проекция (локальное время)  
- \(\tilde{L}\) — пространственная проекция  

Эта структура позволяет рассматривать релятивистские эффекты как **евклидовые проекции в фазовом пространстве**.

---

## 2. Сложные и простые объекты

- **Сложный объект**: собственная временная проекция максимальна (\(\tilde{S} = \tilde{H}\)), внутренний поток преобладает.  
- **Простой объект (фотон)**: движется по пространственной проекции (\(\tilde{L} = \tilde{H}\)), собственного времени нет.  

Комплексное представление потока:

\[
\tilde{V} = \tilde{S} + i \tilde{L}, \quad |\tilde{V}| = \tilde{H}
\]

Таким образом, замедление времени наблюдателя — это уменьшение проекции временной компоненты на фазовый вектор.

---

## 3. Лоренц-фактор как отношение фаз

В униметрии Лоренц-фактор возникает не как постулат, а **как отношение фаз двух систем**:

\[
\gamma = \frac{\chi_{\text{набл.}}}{\chi_{\text{собств.}}} = \frac{1}{\sqrt{1-\beta^2}}
\]

- \(\chi_{\text{набл.}}\) — фаза наблюдателя  
- \(\chi_{\text{собств.}}\) — фаза объекта  
- \(\beta = V / c\) — относительная скорость  

---

## 4. Фазовый релятивистский Доплер

Частота связана с фазой через производную по собственному времени:

\[
\omega_s = \frac{d\phi_s}{d\tau_s}, \quad \omega_0 = \frac{d\phi_0}{d\tau_0}
\]

Доплеровский сдвиг:

\[
r = \frac{\omega_0}{\omega_s} = \frac{d\phi_0 / d\tau_0}{d\phi_s / d\tau_s} = \frac{d\phi_0}{d\phi_s} \cdot \frac{d\tau_s}{d\tau_0}
\]

С учётом релятивистского замедления времени:

\[
\frac{d\phi_0}{d\phi_s} = \frac{r}{\gamma} = 1 - \beta
\]

Накопление фазового сдвига по углу поворота прото-скорости \(\theta\) даёт **экспоненциальное изменение частоты**:

\[
\frac{dr}{r} = \frac{d\phi}{\cos \theta}
\]

---

## 5. Геометрическая интерпретация

- Прото-скорость — вектор в **евклидовом пространстве Хора**.  
- Вращение прото-скорости соответствует видимому изменению времени и пространства.  
- Замедление времени возникает из уменьшения временной проекции фазового вектора.  

---

## 6. Отличие от СТО

1. **Исключительная система** — существует объект с максимальной собственной фазой.  
2. **Симметрия замедления** отсутствует: взаимное замедление зависит от фонового потока.  
3. **Световая скорость константна** для всех наблюдателей, так как любой измеритель масштабирует своё восприятие внутри протопотока.  

---

## 7. Парадокс близнецов

Разрешается через наличие **фона гравитационного потока**: менее подвижный объект «стареет» быстрее, так как его собственная фаза тратится медленнее.  
В отсутствие фона взаимные отношения замедления становятся симметричными, как в СТО.

Униметрия позволяет:

- Представить **время и пространство как производные от фазы**  
- Пояснить Лоренц-фактор и релятивистский Доплер через **геометрию фазового вектора**  
- Сохранить **константу скорости света** через внутреннюю структуру объектов  
- Рассматривать релятивистские эффекты как **евклидовые проекции**, без гиперболических преобразований

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Функция для визуализации фазового вектора
def plot_phase(theta_deg=0, H=1.0):
    theta = np.deg2rad(theta_deg)
    
    # Прото-скорость
    S = H * np.cos(theta)  # временная проекция
    L = H * np.sin(theta)  # пространственная проекция
    
    # Создаем фигуру
    plt.figure(figsize=(6,6))
    plt.axhline(0, color='gray', lw=1)
    plt.axvline(0, color='gray', lw=1)
    
    # Вектор фазового потока
    plt.quiver(0, 0, L, S, angles='xy', scale_units='xy', scale=1, color='blue', label='Phase vector')
    
    # Проекции
    plt.plot([0, L], [0,0], 'r--', label='Spatial projection L')
    plt.plot([0,0], [0, S], 'g--', label='Temporal projection S')
    
    plt.xlim(-H*1.1, H*1.1)
    plt.ylim(-H*1.1, H*1.1)
    plt.xlabel('L (space)')
    plt.ylabel('S (time)')
    plt.title(f'Phase vector rotation: theta={theta_deg}°')
    plt.legend()
    plt.grid(True)
    plt.show()
    
interact(plot_phase, theta_deg=FloatSlider(min=0, max=90, step=1, value=0));
```
