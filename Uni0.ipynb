# Униметрия. Прото-функция

Мы ощущаем эту реальность как данность, как нулевой уровень дериватива, но что если она вторична?
Что если реальность, какой мы привыкли её воспринимать - лишь [пещера Платона](https://ru.wikipedia.org/wiki/%D0%9C%D0%B8%D1%84_%D0%BE_%D0%BF%D0%B5%D1%89%D0%B5%D1%80%D0%B5)?

## 1. Время и пространство как производные
Предположим, что существует переменная величина, назовём её обезличенно $ \chi $, изменение которой порождает наблюдаемые эффекты времени и пространства. Алгебраически это означает просто, что время-пространственные базовые векторы (единицы, меры) являются скоростью - первой производной какого-то прото-движения:
$$ dx_i = \tilde{X}_i \ d\chi, \ i = 0..3 \in \mathbb{Z} \qquad (1.1) $$
Проще говоря, мы динамизируем тензор пространства-времени через единую общую переменную, введя скорость интервала СТО $ ds = \tilde{S} \ d\chi $ аналогично ранее представленым прото-скоростям::
$$ \tilde{S}^2 = \frac{ds^2}{d\chi^2} = \frac{g_{ij} dx^i dx^j}{d\chi^2} = g(\chi) = \begin{pmatrix} \tilde{X}_0^2 & 0 & 0 & 0 \\ 0 & -\tilde{X}_1^2 & 0 & 0 \\ 0 & 0 & -\tilde{X}_2^2 & 0 \\ 0 & 0 & 0 & -\tilde{X}_3^2 \end{pmatrix} = \tilde{X}_0^2 - \tilde{X}_1^2 - \tilde{X}_2^2 - \tilde{X}_3^2  \quad (1.2) $$
Что после объединения пространственной части $ \tilde{L}^2 = \tilde{X}_1^2 + \tilde{X}_2^2 + \tilde{X}_3^2 $ и перестановки равносильно:
$$ \tilde{X}_0^2 =  \tilde{S}^2 + \tilde{L}^2  \qquad (1.4) $$
Введём $ \tilde{\mathcal{H}} = \tilde{X}_0 $:
$$ \tilde{\mathcal{H}}^2 =  \tilde{S}^2 + \tilde{L}^2  \qquad (1.5) $$
что выглядит как формула сложения скоростей в евклидовом векторном пространстве. Предположим, что так и есть.  
Что у нас есть? Видно, что если представить время-пространство индуцированым изменением единого параметра, то величина скорости данного параметра вдоль темпоральной координатной оси может быть представлена как сумма скоростей вдоль двух отдельных компонент - пространственной и вне-пространственной (наблюдаемой и ненаблюдаемой, внешней и внутренней).
С математической точки зрения можно сказать, что (1.5) просто утверждает существование параметра, по которому темпоральная компонента интервала может быть разложена как сумма двух проекций на перпендикулярные оси в евклидовом пространстве. 

### 2. Концепт потока
Предположим, что слева в (1.5) - величина, являющаяся источником всей динамики времени-пространства __некоторого объекта__. Будем называть эту величину __потоком__, а векторное пространство (в математическом смысле, не в физическом), в котором поток существует - [__Хора__ (Платон)](https://en.wikipedia.org/wiki/Kh%C3%B4ra?utm_source=chatgpt.com).  
Примем также, что правая часть (1.5) состоит из "внутренней" $ \tilde{S} $ и "внешней" $ \tilde{L} $ динамики.  

### 3. Скорость потока
Мы стали обозначать скорость тильдой вместо точки, чтобы избежать возможной путаницы в будущем, потому что величина с точкой - это скорость по локальному времени объекта, а тильда - по абсолютному:
$$ [ \tilde{\mathcal{H}} ], [ \tilde{S} ], [ \tilde{L} ]   = \frac{ \mathtt{прото-пространство} }{ \mathtt{прото-параметр} } = \tilde{\chi} $$

### 4. Объект
Любое ненулевое значение потока ($ \tilde{\mathcal{H}} \ne 0 $) будет называться __объектом__. Если объект производит собственную систему отсчёта, в которой он находится в покое:
$$ \tilde{S} = \tilde{\mathcal{H}} \qquad (4.1) $$
то можно сказать, что весь поток объекта является ненаблюдаемым (для самого объекта) - внутренним. Будем говорить о таком объекте как о __сложном__ (__массивном__), в противопоставление объекту, который не может находится в покое:
$$ \tilde{L} = \tilde{\mathcal{H}} \qquad (4.2) $$
и весь поток которого внешний. Будем называть его __простым__ объектом.  
Таким образом, сложным объектом на языке современной физики будет являться любое __событие__ в смысле СТО или череда таковых - мировая линия, а простым - фрагмент подпространства нулевого интервала, содержащий информацию о __событии__.  
В общем случае, объект это вектор в пространстве $ (\tilde{S},\tilde{L}), L \in \mathbb{R}^3 $, которое удобно будет представить комплексным: $ \tilde{V} = e^{z}, z \in \mathbb{C}: \Re(\tilde{V}) = \tilde{S}; \ \Im(\tilde{V}) = \tilde{L}; \ |\tilde{V}| = \tilde{\mathcal{H}} $.  
Из (5.1) и (5.2) совместно следует, что поток сложного объекта всегда движется вдоль реальной оси с той же скоростью $ \tilde{\mathcal{H}} $, с которой простые объекты движутся перпендикулярно ему - вдоль мнимой. Таким образом, скорость потока является не свойством отдельно взятого изолированного объекта, а может быть скорее представлена гладко изменяющимся скалярным полем над прото-пространством. Кажется, это всё, что пока можно постулировать о ней на данном этапе.
Эффекты СТО же достигаются трансляцией точки в пространстве $ (\tilde{S}, \tilde{L}) $ вдоль мнимой оси (с сохранением реальной части - интервала). Очевидно, такое преобразование не является ортогональным - норма вектора $ |\tilde{V}| = \tilde{\mathcal{H}} $ увеличится.  
Противоречие возникает от того, что если объект А движется относительно объекта Б, то норма скорости потока А будет больше, но вместе с тем, объекта Б движется в ИСО объекта А, где норма потока А должна быть меньше, чем норма потока Б. Полагаю именно эта неурядица с нормами привела к использованию гиперболического инструментария. Что делать, если мы хотим оставить евклидову метрику?  
Принять, что норма относительна в части движения. Такое в физике уже было, это не большая проблема. Большая проблема возникает с сохранением постоянства скорости света. Но что если мы построим модель, в которой скорость света не фундаментальна, но сохраняется как воспринимаемая величина?

### 5. Время
Для начала необходимо разобраться с масштабированием времени и пространства скоростью потока, ведь скорость потока вдоль темпоральной оси прежде всего прочего нормирует воприятие обстоятельств:
$$ dx_0 = \tilde{X}_0 \ d\chi = \frac{dx_0}{d\chi} \textcolor{blue}{ \frac{d\chi}{d\tau} } \ d\tau = \tilde{X}_0 \textcolor{blue}{ \dot{\chi} } \ d\tau \qquad (5.1) $$
Скорость течения потока, выраженная в локальном времени, будет произвдением двух скоростей - скорости потока по единому прото-параметру и скорость этого параметра по локальному времени:
$$ \textcolor{blue}{ \dot{\chi} } = \frac{d\chi}{d\tau} = \frac{dx_0}{dx_0} \frac{d\chi}{d\tau} =\frac{ \dot{X}_0 }{ \tilde{X}_0 } \qquad (5.2) $$
Производная прото-параметра по пространственной компоненте:
$$ \chi' = \frac{ d\chi }{ dl } = \frac{ d\chi }{ \sqrt{dx_1^2 + dx_2^2 + dx_3^2 } } = \tilde{L}^{-1} \qquad (5.3) $$

Из наличия в окружающем мире большого числа частотных явлений неясного генеза, предположим, что локальное время возникает как результат совершения объектами некоторого циклического действия, как если бы геометрия оси $ x_0 $ была замкнутой. Упрощённо представим её окружностю варьируемого радиуса $ R $, тогда в высокочастотном пределе имеем:  
$$ \nu = \frac{\Delta \chi}{\Delta \tau} \overset{R \rightarrow 0}{\implies} \nu = \frac{d\chi}{d\tau} = \textcolor{blue}{ \dot{\chi} } \quad (6.3) $$
где $ \Delta \chi = 2 \pi R_1, \ \Delta \tau = 2 \pi R_2 / \tilde{X}_0 $, следовательно, частота $ \nu = k \tilde{X}_0, \ k = R_1 / R_2 $, а значит и скорость прото-параметра: $$ \boxed{ \textcolor{blue}{ \dot{\chi} } = k \tilde{X}_0 }  \quad (6.4)$$
<div class="alert alert-block alert-info"> <b>Примечание.</b> При заданных упрощённо в виде окружности циклических формах, частота всегда будет линейной функцией скорости потока и без предельного условия, однако важно понимать, что это лишь упрощение, и реальные формы конфигурации потока микрообъектов могут быть сложнее, а значит, введённый коэффициент <b>k</b> совсем не обязан быть линейным для объектов такого масштаба, но для сложносоставных объектов будет уже их количественной характеристикой. </div>
Это ключевая точка всей теории. Мало того, что метрика генерируется изменением какого-то прото-параметра, но она ещё должна быть описана в динамике воспринимаемой реальности - относительно времени как мы его регистрируем. Для этого сначала получим множитель приведения из прото-пространственных скоростей к локально значимым для всех компонентов (2.4). Он будет $ k \tilde{X}_0 $:
$$ \dot{\mathcal{H}} = \frac{d\mathcal{H}}{d\chi} \frac{d\chi}{d\tau} = \tilde{\mathcal{H}} \ k \tilde{X}_0 \quad \Rightarrow \quad \tilde{\mathcal{H}} = \frac{\dot{\mathcal{H}}}{k \tilde{X}_0} = \tilde{X}_0 ; \quad \dot{X}_0 = k \tilde{X}_0^2; \quad \tilde{S} = \frac{\dot{S}}{k \tilde{X}_0} ; \quad \tilde{L} = \frac{\dot{L}}{k \tilde{X}_0} \qquad (6.5) $$
Полная величина потока в локальном времени полуается:
$$ \boxed{ \dot{\mathcal{H}} = k \tilde{X}_0^2 }  \qquad (6.6) $$
И чтобы она сохранялась при изменении темпоральной проекции прото-скорости, необходимо, чтобы $ k = \tilde{X}_0^{-2} $
Теперь проверим форму интервала (из 6.1 и 6.4):
$$ dS^2 = ds^2 = \frac{g_{ij} dx^i dx^j}{d\chi^2} \frac{d\chi^2}{d\tau^2} d\tau^2 = g(\chi) k^2 \tilde{X}_0^2 d\tau^2 = \dot{X}_0^2 \left( 1 -\frac{\tilde{X}_1^2}{\tilde{X}_0^2} - \frac{\tilde{X}_2^2}{\tilde{X}_0^2} - \frac{\tilde{X}_3^2}{\tilde{X}_0^2} \right) \ d\tau^2 = $$
$$ \dot{X}_0^2 \left( d\tau^2 -\frac{dx_1^2}{dx_0^2} d\tau^2 - \frac{dx_2^2} {dx_0^2}d\tau^2 - \frac{dx_3^2}{dx_0^2} d\tau^2 \right) = \left| dx_0 = \dot{X}_0 \ d\tau \right| = \dot{X}_0^2 d\tau^2 - dx_1^2 - dx_2^2 - dx_3^2 \quad (6.7) $$
Значит, можно утверждать, что сохранение гиперболической формы интервала представимо поворотом вектора прото-скорости в евклидовом пространстве с учётом соответствующей геометризации времени.  
Можно показать это несколько иначе, заметив, что единый прото-параметр задаёт величину, которая будет инвариантна для преобразований затрагивающих и время, и пространство (из 6.4):
$$ d\chi = k \tilde{X}_0 \ d\tau \qquad (6.8) $$
В том смысле, что если слева величина единая для течения времени и пространства, которая будет сохраняться при её интегрировании в любых обстоятельствах, то выражение справа - тоже инвариант. И как видно из сравнения (6.4) и (6.6), частотная сущность воспринимаемого времени ведёт к инвариантности относительно себя именно "вывернутой" формы - интервала, а не полной энергии.  
Также из (6.8) следует, что для объекта фиксированной сложности ($ k = const $) величина $ \tilde{X}_0 $ должна компенсировать изменение течения времени. Нагляднее будет его переписать так:
$$ d\chi =  \frac{ \dot{X}_0 \ d\tau }{\tilde{X}_0}  \qquad (6.9) $$
Скорость света в единицах локального времени постоянна, пока постоянна скорость потока относительно прото-параметра.  


Из наблюдений мы можем заключить, что сама метрика при рассмотрении пространственной скорости сохраняется, то есть полная скорость потока $ \tilde{\mathcal{H}} = const $ перераспределяется внутри континуума. Но этого самого по себе недостаточно, чтобы сохранить $ \dot{X}_0 = const = k \tilde{X}_0^2 $

## Частота
Возьмём релятивистскую формулу доплеровского изменения частоты как включающую в себя все затрагиваемые эффекты (доплеровский эффект и релятивистское движение, не затрагивает вращение и гравитацию):
$$ \omega_s = \omega_0 \sqrt{ \frac{ 1 - \frac{ v }{ \mathtt{c} } }{ 1 + \frac{ v }{ \mathtt{c} } } } $$
и выразим в обратную сторону как отношение скорости к скорости света через отношение частот:
$$ \frac{ v }{ \mathtt{c} } = \frac{ 1 - \frac{ \omega_s^2 }{\omega_0^2} }{ 1 + \frac{ \omega_s^2 }{\omega_0^2} } $$


### 7. Пространство
#### 7.1 Самый простой случай
Пусть два сложных объекта имеют нулевую пространственную составляющую потока друг с другом, расстояние $ L $ в прото-пространстве между собой, и эмитируют в направлении друг друга простые объекты с временными промежутками $ T_1 $ и $ T_2 $, соответственно:
$$ \tilde{V}_1 = \tilde{V}_2 = \tilde{X}_0 \qquad (7.1) $$
Из равенства скоростей потоков следует равенство течения локальных времён и скоростей простых объектов, так что время необходимое для регистрации события эмиссии объекта 1 объектом 2 и наоборот будет одинаковым:
$$ \tau_{12} = \tau_{21} = \frac{ L }{ \dot{X}_0 } \qquad (7.2) $$
Скорости течения времени у объектов одинаковые, поэтому и периодические промежутки переносятся от одного объекта к другому без искажения:  

#### 7.2 Относительное движение
Теперь положим, что между объектами есть смещение - потоки взаимно видны под углом $ \theta $. Тогда утверждения (7.1) по прежнему в силе - каждый из объектов видит себя как прежде, но:
$$ \tilde{V}_{21} = -\tilde{V}_{12} = \tilde{X}_0 \ e^{\imath \theta} \qquad (7.3) $$
Что можно представить как два непараллельных вектора равной длины на комплексной плоскости.




Что после приведения к локальному времени даст:
$$ \dot{V} = k \ \tilde{X}_0^2 \  \ e^{ \imath \theta } = k \ \tilde{X}_0^2 \ \cos \theta \left( 1 + \imath \tan \theta \right) $$

(из (6.5)):
$$ \dot{L}^2 = \begin{pmatrix} 0 & 0 & 0 & 0 \\ 0 & \tilde{X}_1^2 & 0 & 0 \\ 0 & 0 & \tilde{X}_2 & 0 \\ 0 & 0 & 0 & \tilde{X}_3^2 \end{pmatrix} \begin{pmatrix} 0 \\ b_1^2 \\ b_2^2 \\ b_3^2 \end{pmatrix} k^2 \tilde{X}_0^2, \ b^2 = b_1^2 + b_2^2 + b_3^2 \quad (7.2) $$
Тогда зависимость между полной прото-скоростью и скоростью для наблюдаемой и ненаблюдаемой части потока в текущем времени будет (подставляя 6.7 в 2.4):
$$ \dot{\mathcal{H}}^2 = \tilde{X}_0^2 = \frac{\dot{S}^2 + \dot{L}^2}{k^2 \tilde{X}_0^2} \quad (7.2) $$
Или также:
$$  k^2 \ \tilde{X}_0^4 = \dot{S}^2 + \dot{L}^2 \quad (7.3) $$
Подставляя соответвующие компоненты, получим:
$$  k^2 \ \tilde{X}_0^4 = \begin{pmatrix} \tilde{X}_0^2 & 0 & 0 & 0 \\ 0 & \tilde{X}_1^2 & 0 & 0 \\ 0 & 0 & \tilde{X}_2^2 & 0 \\ 0 & 0 & 0 & \tilde{X}_3^2 \end{pmatrix} k^2 \tilde{X}_0^2 + \begin{pmatrix} 0 & 0 & 0 & 0 \\ 0 & \tilde{X}_1^2 & 0 & 0 \\ 0 & 0 & \tilde{X}_2^2 & 0 \\ 0 & 0 & 0 & \tilde{X}_3^2 \end{pmatrix} k^2 \dot{X}_0^2 \quad (7.3) $$
