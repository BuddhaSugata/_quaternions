# Униметрия. Фазовый формализм

Мы ощущаем эту реальность как данность, как нулевой уровень дериватива, но что если она вторична?
Что если реальность, какой мы привыкли её воспринимать - лишь [пещера Платона](https://ru.wikipedia.org/wiki/%D0%9C%D0%B8%D1%84_%D0%BE_%D0%BF%D0%B5%D1%89%D0%B5%D1%80%D0%B5)?  
В статье рассмотрена возможность варьирования метрики пространства-времени с помощью концепции, в которой время и пространство вторичны относительно изменения некоторого единого прото-параметра - фазы, с целью получения математической модели, воспроизводящей эффекты специальной теории относительности в более фундаментальном представлении, и оставляющей возможность для объединения с другими актуальными теориями.  
Мы предполагаем, что метрика пространства-времени - это своего рода оболочка, презентующая законы природы в тех терминах, которые нам наиболее естественны - представление о которых, сложилось исторически натурально. Но позволим себе взглянуть на вещи иначе и предположим, что время - это мера относительной скорости протекания циклов, а не самостоятельная фундаментальная величина. Тогда темпоральную ось теорий относительности вполне законномерно можно представить длиной, которую составные частицы массивного объекта проходят в ненаблюдаемой части пространства. Однако длина тоже не является фундаментальной величиной. Она зависит от масштаба и выбора базисного вектора, а значит по сути относительна.  
Отсюда вытекает допущение, давшее название статье - и время, и пространство могут и должны быть заданы единой величиной, чтобы дать место существованию математического объяснения постоянства скорости света и сохранения интервала в гиперболическом пространстве.  
В статье мы ищем формализм, в котором СТО естественно вытекает из математики более фундаментального параметра, и оба эффекта воспроизводятся как результат сохранения фазовых свойств объектов. Фундаментальной величиной при этом, очевидно, будет служить фаза, потому что угол абсолютен в евклидовом пространстве, которое, следовательно, также принято фундаментальным.  

## 1. Время и пространство как производные
Предположим, что существует переменная величина, назовём её обезличенно $ \chi $, изменение которой порождает наблюдаемые эффекты времени и пространства. Алгебраически это означает просто, что время-пространственные базовые векторы (единицы, меры) являются скоростью - первой производной какого-то прото-движения:
$$ dx_0 = \frac{d H}{ d\chi } d\chi = \tilde{H} \ d\chi, \ dx_\mathcal{l} = \frac{d L}{ d\chi } d\chi = \tilde{L} \ d\chi, \, l = 1,2,3 \qquad (1.1) $$
Можно сказать, что мы динамизируем тензор пространства-времени через единую общую переменную, введя прото-скорость интервала СТО $ ds = \tilde{S} \ d\chi $ аналогично ранее представленым прото-скоростям:
$$ \tilde{S}^2 = \frac{ds^2}{d\chi^2} = \frac{g_{ij} dx^i dx^j}{d\chi^2} = g(\chi) = \tilde{H}^2 - \tilde{L}^2 \quad (1.2) $$
Что равносильно:
$$ \tilde{\mathcal{H}}^2 = \tilde{S}^2 + \tilde{L}^2  \qquad (1.3) $$
С математической точки зрения можно сказать, что (1.3) просто утверждает существование параметра, по которому темпоральная компонента интервала может быть разложена как сумма двух проекций на перпендикулярные оси в евклидовом пространстве, то есть если представить, что время-пространство индуцировано изменением единого параметра, то величина скорости по данному параметру вдоль темпоральной координатной оси может быть представлена как сумма скоростей вдоль двух отдельных компонент - пространственной и вне-пространственной (наблюдаемой и ненаблюдаемой, внешней и внутренней).  
Или, упрощая, (1.3) выглядит как формула сложения скоростей в евклидовом векторном пространстве. Предположим, что так и есть, но по какой-то причине при повороте "угла обзора" сохраняется не норма $ \tilde{H} $, а проекция на вне-пространственную $ \tilde{S} $.  

## 2. Концепт потока
Предположим, что слева в (1.3) - величина, являющаяся источником всей динамики времени-пространства __некоторого объекта__. Будем называть эту величину __потоком__, а векторное пространство (в математическом смысле, не в физическом), в котором поток существует - __прото-пространство__ [__Хора__ (Платон)](https://en.wikipedia.org/wiki/Kh%C3%B4ra?utm_source=chatgpt.com).  
Геометрически мы определим два базисных вектора назвав их (__время__, __пространство__), не разделяя пока пространство далее, чтобы не усложнять: $ (\mathbf{e}_h, \, \mathbf{e}_l) $; получив в результате представление потока $\tilde{H} = \tilde{H} (\tilde{S}, \tilde{L})$ как вектора в данном базисе.  
Введём название и обозначение некоторой __прото-функции__, задающей характер потока объекта как вектор в обозначенном __прото-пространстве__:
$$ H = \{ \mathbf{e}_n \}, \, \ H \in \mathbb{R}^n \qquad (2.1) $$
И далее рассмотрим какими свойствами она должна обладать, чтобы подобрать её форму.

### 2.1. Скорость потока
Мы стали обозначать скорость тильдой вместо точки или штриха, чтобы избежать возможной путаницы в будущем, потому что величина с точкой - это скорость по локальному времени объекта, со штрихом - по длине, а тильда - по прото-параметру:
$$ [ \tilde{H} ], [ \tilde{S} ], [ \tilde{L} ]   = \left[ \frac{ \mathtt{прото-пространство} }{ \mathtt{прото-параметр} } \right] $$
Будем называть её также __прото-скорость__.  

### 2.2. Объект
Мы предполагаем, что прото-параметр - мера динамики объектов, поэтому любая ненулевая величина скорости потока ($ \tilde{H} \ne 0 $) будет называться __объектом__. Если объект производит в нашем восприятии собственную систему отсчёта, в которой он находится в покое:
$$ \tilde{S} = \tilde{H} \qquad (2.2.1) $$
то можно сказать, что весь поток объекта является ненаблюдаемым (для самого объекта) - внутренним. Будем говорить о таком объекте как о __сложном__ (__массивном__), в противопоставление объекту, который не может находится в покое:
$$ \tilde{L} = \tilde{H} \qquad (2.2.2) $$
и весь поток которого внешний. Будем называть его __простым__ объектом.  
Таким образом, сложным объектом на языке современной физики будет являться любое __событие__ в смысле СТО или череда таковых - мировая линия, а простым - вектор подпространства нулевого интервала, содержащий информацию о __событии__.  
В общем случае, объект без взаимодействий это постоянная величина количества прото-движения, которое может быть представлено вектором в пространстве прото-скоростей $ (\mathbf{e}_h, \, \mathbf{e}_l) $, которое удобно будет задать комплексным: $ \tilde{V} = e^{z}, z \in \mathbb{C}: \Re(\tilde{V}) = \tilde{S}; \ \Im(\tilde{V}) = \tilde{L}; \ |\tilde{V}| = \tilde{H} $.  
Из (2.2.1) и (2.2.2) совместно следует, что поток сложного объекта всегда движется вдоль реальной оси с той же прото-скоростью $ \tilde{H} $, с которой простые объекты движутся перпендикулярно ему - вдоль мнимой.  
В представлении униметрии мы предлагаем рассматривать сложный объект как совокупность множества элементарных замкнутых траекторий с проекциями на наблюдаемое пространство для реализации эффектов создаваемых полей в его окреснтости (гравитационное, электро-магнитное, сильное и слабое взаимодействия), что для моделирования эффектов ОТО может быть представлено гладко изменяющимся скалярным полем над прото-пространством, а для СТО - принято постоянным. Кажется, это всё, что стоит постулировать на данном этапе.  
Кроме того, подобная конструкция позволит объяснить эффекты изменения геометрии тел при наличии скорости, например, просто как пересечение прото-образа (вероятно, полисферы) сложного объекта с подпространством простых объектов, которое зависит от состояния приёмника. В конце концов, если подпространство простых объектов всегда находится внутри множества пространств всех сложных объектов, то простой объект окажется свойством сложного.  
Эффекты СТО достигаются трансляцией точки в пространстве $ (\mathbf{e}_h, \, \mathbf{e}_l) $ вдоль мнимой оси (с сохранением реальной части - интервала). При таком преобразовании теряется ортогональность, и норма изменяется. Это и создаёт известное противоречие с привычным пониманием относительности - норма вектора $ |\tilde{V}| = \tilde{H} $ увеличится.  
Противоречие возникает оттого, что если объект А движется относительно объекта Б, то норма скорости потока А будет больше, но вместе с тем, объект Б движется в ИСО объекта А, где норма потока А должна быть меньше, чем норма потока Б. Эта неурядица с нормами наряду с эмпирически установленным постоянством скорости света привела к использованию здесь псевдо-евклидового инструментария пространства Минковского.  
В предлагаемом концепте, очевидно, необходимо принять, что норма не изменяется - меняется её отображение как результат действия физического процесса, возникающего при изменении относительной скорости объекта и наблюдателя, то есть в итоге существует математический механизм, приводящий к выравниванию по вне-пространственной оси, не по полной евклидовой норме прото-скорости.  

### 2.3. Время и пространство
Масштабирование времени скоростью потока можно задать по цепному правилу через нормирующий фактор:
$$ dx_0 = \tilde{H} \ d\chi = \frac{dx_0}{d\chi} \textcolor{blue}{ \frac{d\chi}{d\tau} } \ d\tau = \tilde{H} \textcolor{blue}{ \dot{\chi} } \ d\tau = \dot{H} d\tau = \mathtt{c} d\tau \qquad (2.3.1) $$
Отсюда видно, что мы принимаем скорость потока по локальному времени константой скорости света $ \mathtt{c} $.  
Аналогично для пространственной части:
$$ dx_l = \tilde{L} \ d\chi = \frac{dx_l}{d\chi} \textcolor{green}{ \frac{d\chi}{dl} } \ dl = \tilde{L} \textcolor{green}{ \chi' } \ dl = L' dl, \, l = 1,2,3 \qquad (2.3.2) $$
Из равенства для светового нулевого интервала: $ dx_0 = dx_l \Rightarrow \mathtt{c} = \tilde{L}' \frac{dl}{d\tau} $. Hесложно видеть, что при масштабировании темпоральной оси скоростью света пространственный масштаб всегда будет единичным $ \tilde{L}' = 1 $. Это удобно, но предположим, что у этого натурально возникающего удобства есть причина в виде единой прото-скорости масштабирующей нашу воспринимаемую реальность. Зафиксируем и вернёмся к этой мысли позже.  

### 2.4. Скорость света
Выразим интервал (1.2) представив обе координаты через прото-параметр:
$$ ds^2 = \frac{\mathtt{c}^2}{ \dot{\chi}^2} d\chi^2 - \frac{1}{ \chi'^2 } d\chi^2 \qquad (2.4.1) $$
Примая интервал равным нулю:
$$ \mathtt{c} = \frac{\dot{\chi}}{ \chi' } = \frac{dl}{ d\tau } \ne f(\chi) \qquad (2.4.2) $$
Откуда видно, что скорость света является параметром, связывающим движения по осям __вне зависимости от значения прото-параметра__.  
Во-первых, если посмотреть на (2.4.2) следующим образом:
$$ \mathtt{c} = \frac{\dot{\chi}}{ \chi' } = \left(\frac{d\chi}{d\tau} \right) \left[ \frac{dl}{ d\chi } \right] \sim (\nu) \ [\lambda] \qquad (2.4.3) $$
То можно заметить сходство с характеристиками фотона _частота_ и _длина волны_, произведение которых даёт скорость света. Следовательно, далее будем исследовать фазу в качестве претендента на прото-параметр: $ \chi = \phi $.  
Во-вторых, из (2.4.1) и (2.4.2) следует:
$$ ds^2 = \mathtt{c}^2 \left( \frac{d\chi^2}{ \dot{\chi}^2}  - \frac{d\chi^2}{ \dot{\chi}^2} \right) = 0, \forall \dot{\chi} \qquad (2.4.4) $$
где решением для единичной частоты будет $ \tau = \chi $, значит в терминах фазы можно составить интервал для каждого отдельного фотона, где собственным временем фотона будет являться течение его фазы. Это в некотором смысле подтверждает идею о вторичности времени относительно фазы как средства измерить относительность произвольных циклов.  
Также важным наблюдением является факт "попытки" нормирования интервала фотона его частотой, однако в связи с отсутствием у фотона проекции на ось потока $ \mathbf{e}_h $, эта попытка реализуется лишь как линейное преобразование фазы фотона пропорционально изменению фазы потока, которая, фактически, выполняет роль времени сложного объекта:
$$
\frac{\chi_1}{\chi_0} = \frac{\tau_1}{\tau_0} \qquad (2.4.5)
$$
Можно сказать, что у фотона просто нет собственного течения по времени в нашем понимании последнего, поэтому он масштабируется только нашим ощущением времени сложного объекта. Такое восприятие несколько непривычно, но получается, что у фотона нет частоты - только некоторая величина фазы. Частота появляется только как отношение этой фазы к собственной фазе времени наблюдателя, где последняя, фактически, выполняет роль собственного времени.  

### 2.5. Лоренц-фактор
Это ключевой пункт теории. Здесь мы в первую очередь рассматриваем случай, когда фоновый вектор (гравитации) относительно слаб, тогда мы можем утверждать, что условие сохранения количества фазового движения обеспечивает наличие локальной системы отсчёта, где течение времени - собственной фазы объекта - будет максимальным: $ \vec{V}_0 = \tilde{H} \, \mathbf{e}_h $; а отклонение от этой нормали $ \vec{V}_1 = \tilde{H} \, (\cos\theta \ \mathbf{e}_h + \sin \theta \ \mathbf{e}_l) = \tilde{H} \, (\cos\theta \ + \sin \theta \ \mathbf{u}) =\tilde{H} \, e^{\theta \ \mathbf{u}} $ даст видимое перемещение в наблюдаемом пространстве вместе с относительным замедлением времени.   
Мы знаем, что в результате преобразования прото-скоростей:
$$ \tilde{\mathcal{H}}^2 = \tilde{S}^2 + \tilde{L}^2 \overset{\cdot \dot{\chi}^2}{\longmapsto} \dot{\mathcal{H}}^2 = \dot{S}^2 + \dot{L}^2  \qquad (2.5) $$
Происходит переход нормы от $ \tilde{\mathcal{H}} $ к норме $ \dot{S} $. Тогда единственной возможнотью для реализации этого перехода является, очевидно, нормирующий фактор - отношение фазы той системы, из которой переходим, к той, в которую переходим. Чтобы сохранение евклидовой нормы выглядело сохранением псевдо-евклидовой нормы, он должен быть отношением прото-скоростей:
$$ \dot{\chi} = \frac{\chi_{\mathtt{obs.}}}{\chi_{\tau}} = \frac{\tilde{H}}{\tilde{S}} = \frac{1}{\sqrt{1-\frac{\tilde{L}^2}{\tilde{H}^2}}} \sim \gamma = \frac{1}{\sqrt{1-\frac{V^2}{\mathbf{c}^2}}}  \qquad (2.5.1) $$
Принцип сохранения полной фазы сложного объекта (равно как и для суммы его частей) даёт аналогичную по структуре зависимость:
$$ \chi_{\mathtt{obs.}}^2 = \chi_{\tau}^2 + \chi_l^2 \Rightarrow \frac{\chi_{\mathtt{obs.}}}{\chi_{\tau}} = \frac{1}{\sqrt{1 - \frac{\chi_l^2}{\chi_{\mathtt{obs.}}^2}}} \sim \gamma  \qquad (2.5.2) $$
Однако прото выразить отношение проекций потока (или отношения фаз) через угол прото-пространственного отклонения не достаточно, чтобы получить нужную зависимость:
$$
\dot{\chi} = \frac{1}{\cos\theta} = \sec\theta \ne \cosh \theta \quad (2.5.3)
$$
Дело в том, что секанс здесь выражает мгновенный фактор, но чтобы получить глобальный эффект, нужно интегрировать по всей траектории отклонения - нормирующий фактор меняет фазу на каждом бесконечно малом повороте, что можно выразить через утверждение равенства малого прироста некоторой величины секансу угла прото-пространственного отклонения или интегрирование:
$$
d\eta = {\sec{\theta} \ d\theta} \quad \iff \quad \eta \ (\theta) = \int{\dot{\chi} \ d\theta} = \int{\sec{\theta} \ d\theta}   \qquad (2.5.4)
$$
Интегралом секанса является функция структурно идентичная полному продольному релятивистскому преобразованию частоты фотонов (эффекту Доплера):
$$
\eta \ (\theta) = \ln \left( \sqrt{ \left| \frac{1+\sin\theta}{1-\sin\theta} \right| } \right) \ \iff \ e^{\eta \ (\theta)} = \sqrt{ \left| \frac{1+\sin\theta}{1-\sin\theta} \right| } \quad \sim \quad r = \sqrt{ \frac{1+\beta}{1-\beta} } \qquad (2.5.5)
$$
Ранее мы предположили отсутствие частоты как свойства фотонов, следовательно, всё изменение их частоты должно быть результатом изменения фазы сложного объекта, а правая часть (2.5.5) как раз включает оба эффекта применимых к движению - Доплер и релятивистское замедление времени.  
Осталось только отсюда получить зависимость угла поворота фазы от скорости:
$$
\theta = \arcsin \left( \frac{e^{2 \eta} - 1}{e^{2 \eta} + 1} \right) = \arcsin(\tanh \eta) \qquad (2.5.6)
$$
Секанс которого будет равен гиперболическому косинусу угла $ \eta $:
$$
\cosh \eta = \sec(\arcsin(\tanh(\eta))) \qquad (2.5.7)
$$
Теперь по порядку от наблюдаемой реальности к фазовому пространству. Когда объект набирает скорость его наблюдаемая фаза меняется в зависимости от отношения его скорости к скорости света $ \beta = V / \mathtt{c} $ по следующему закону:
$$
r = e^{\eta} = \sqrt{ \frac{1+\beta}{1-\beta} } \qquad (2.5.8)
$$
где $ r $ - отношение частот фотонов для неподвижного источника и движущегося приёмника, а $ \eta $ - быстрота ([rapidity](https://en.wikipedia.org/wiki/Rapidity)) - угол в гиперболическом аппарате СТО. Угол поворота скорости фазы $ \theta $ таков, что его синус равен гиперболическому тангенсу быстроты:
$$
\sin\theta = \tanh \eta = \beta = \frac{V}{\mathtt{c}} \qquad (2.5.9)
$$
Тогда Лоренц-фактор будет секансом этого угла:
$$
\boxed{ \gamma = \frac{1}{\cos \theta} = \frac{1}{ \sqrt{1 - \tanh^2 \eta}} = \cosh \eta } \qquad (2.5.10)
$$
Отметим, что Лоренц-фактор возникает здесь не как самодостаточный постулируемый конструкт, а как результат изменения фазы. Более того, полное изменение частоты фотонов наблюдаемого движущегося объекта также возникает натурально при интегрировании поворота потока.  

### 2.6. Инвариантность
В стандартной СТО интервал $ds^2$ является инвариантом группы Лоренца. В фазовом формализме роль интервала выполняет величина полной фазы наблюдателя $ \tilde{H} $.  
Для сложных объектов всегда существует ИСО, в которой они покоятся - _собственная_ ИСО, в прообразе которой прото-скорость объекта не имеет фазового отклонения: $ |\vec{v}| = \tilde{H} $. Множество взаимно неподвижных сложных объектов образуют плоское (в смысле нулевой кривизны) пространство фотонов, ввиду отсутствия искажения их фазовых характеристик.  
Наличие относительного пространственного перемещения одного сложного объекта относительно другого представляется поворотом прото-скорости в фазовом пространстве, на угол $ \theta = \arcsin \left( \frac{V}{\mathtt{c}} \right) $, так что сохранение полной фазы движущегося объекта даёт: $ \vec{v}' = \vec{v} \cdot e^{\theta \ \mathbf{u}} = \tilde{H} (\cos\theta + \mathbf{u} \ \sin\theta) = \tilde{S} + \tilde{L} \ \mathbf{u} $.  
Проекция фазовой прото-скорости движущегося объекта на ось потока наблюдателя $ \tilde{S} = \tilde{H} \cos\theta $, является дифференциальной формой отклонения "внутренней" фазы (связанной с течением времени) обусловленного поворотом "внешней" - пространственной (связанной с пространственным перемещением).  
Если мы предположим, что в восприятии наблюдателя физически обнаружимы только "дотированные" величины - нормированные с учётом угла его фазового течения, то сохраняемой при фазовом повороте окажется именно проекция "внутренней" фазы, растянутая нормированием: $ \dot{S} = \tilde{H} = \tilde{S} / \cos{\theta} $.  
В физических величинах наблюдателя сохранение такой формы даст:
$$
\dot{S}^2 = \dot{H}^2 - \dot{L}^2 = \left( \frac{1}{\cos^2 \theta} - \tan^2 \theta \right) \ \tilde{H}^2 \qquad (2.6)
$$
где $ \dot{H} = \mathtt{c} $ - метрика времени и $ \dot{L} = 1 $ - метрика пространства для объектов коллинеарного направления фазы, а $ \dot{H} = \mathtt{c} \gamma $ и $ \dot{L} = \frac{V}{\mathtt{c}} \gamma = \beta \gamma $, соответственно, для наблюдения объектов в изменённом фазовом состоянии с фактором $ \gamma = \sec \theta $ - от угла поворота прото-скорости.  

## Заключение

Предложенный евклидовый фазовый формализм позволяет рассматривать релятивистские эффекты как проекцию более фундаментальной динамики потоков в протопространстве. Такой подход открывает возможность для дальнейшего исследования связи между геометрией, фазой и наблюдаемой физической реальностью. В перспективе униметрия может служить мостом к новым представлениям о времени, массе и структуре материи, сохраняя при этом простоту евклидовой геометрии.



### 2.7. Ускорение
Преобразования масштабов проецируемого базиса происходят с возникновением ускорения - изменения угла прото-скорости $ \theta $, то есть ускорение - это ротор.  
Имеет смысл соотнести его с количеством прото-скорости:

### 2.7. Ускорение
Преобразования масштабов проецируемого базиса происходят с возникновением ускорения — изменения угла прото-скорости $ \theta $.  
Иными словами, ускорение можно рассматривать как ротор в прото-пространстве, то есть как скорость изменения фазового угла:

$$
a = \frac{dV}{dt} \;\;\; \longleftrightarrow \;\;\; \alpha = \frac{d\theta}{d\tau}
\qquad (2.7.1)
$$

где $ a $ — привычное ускорение в пространстве-времени, а $ \alpha $ — фазовое ускорение, описывающее изменение угла прото-скорости относительно собственного времени объекта.

---

#### 2.7.1. Связь с рапидностью
Ранее мы установили, что угол $ \theta $ связан с гиперболической быстротой (рапидностью) $ \eta $ через соотношение:

$$
\sin \theta = \tanh \eta, 
\qquad
\gamma = \cosh \eta = \frac{1}{\cos \theta}.
\qquad (2.7.2)
$$

Тогда фазовое ускорение можно выразить через производную быстроты:

$$
\alpha = \frac{d\theta}{d\tau} 
= \frac{d\theta}{d\eta} \cdot \frac{d\eta}{d\tau}.
\qquad (2.7.3)
$$

Из (2.7.2) получаем:

$$
\frac{d\theta}{d\eta} = \frac{1}{\cosh \eta} = \frac{1}{\gamma}.
\qquad (2.7.4)
$$

Следовательно:

$$
\alpha = \frac{1}{\gamma} \frac{d\eta}{d\tau}.
\qquad (2.7.5)
$$

---

#### 2.7.2. Связь с собственным ускорением
В теории относительности вводится **собственное ускорение** $a_{\ast}$ как производная быстроты по собственному времени:

$$
a_{\ast} = \frac{d\eta}{d\tau}.
\qquad (2.7.6)
$$

Сопоставив (2.7.5) и (2.7.6), получаем:

$$
\alpha = \frac{a_{\ast}}{\gamma}.
\qquad (2.7.7)
$$

Таким образом, фазовое ускорение — это нормированная версия собственного ускорения, где нормировка задаётся Лоренц-фактором.  

---

#### 2.7.3. Физический смысл
1. В привычной СТО ускорение описывает изменение пространственной скорости.  
2. В фазовом формализме ускорение естественно возникает как **изменение угла прото-скорости**, то есть вращение потока в прото-пространстве.  
3. Постоянное собственное ускорение соответствует **равномерному вращению в фазовом пространстве**, что отображается в пространстве-времени как гиперболическое движение (траектория — мировая гипербола).  

---

Итак, ускорение в униметрии имеет более фундаментальное объяснение: это просто скорость фазового вращения потока объекта, которое при проекции на пространство-время проявляется как гиперболическое движение с релятивистскими ограничениями.


Тогда можно ввести преобразование нормирующее  
которую необъодимо нормировать преобразованием фазы так, как если бы геометрического поворота не происходило (из 2.5):  
$$
\dot{S}^2 = \tilde{H}^2 = \frac{1}{\cos^2\theta} (\dot{H}^2 - \dot{L}^2) = \tilde{H}^2 (1 - \tan^2 \theta) \qquad (2.6)
$$

$$
\vec{V}' = \vec{\tilde{H}}' = \vec{V} \ e^{\theta \ \mathbf{u}} \quad \overset{ 1: \ \cdot \cos \theta }{\longmapsto} \quad \vec{V}'' = \frac{1}{\cos\theta} \tilde{H} ( \cos\theta + \mathbf{u} \ \sin{\theta} ) = \tilde{H} + \mathbf{u} \ \tan{\theta} \ \tilde{H} \qquad (2.6.1)
$$



При этом мы можем работать с фазой как с геометрическим объектом в прото-пространстве в качестве прообраза реального объекта - отклонение полной фазы покоя для перемещения ведёт к уменьшению _проекции_ фазы времени, следовательно, к его наблюдаемому замедлению. Причём взаимные проекции двух объектов одинаковой нормы будут равноуменьшены - время будет казаться замедленным в обеих системах. 

В классической СТО преобразование Лоренца выражается через гиперболические функции:  

$$
\cosh x = \frac{e^x + e^{-x}}{2}, \quad \sinh x = \frac{e^x - e^{-x}}{2}.
$$

Однако можно показать, что то же преобразование допускает эквивалентную запись через евклидовый поворот в фазовом пространстве.  
Ключевая связь задаётся тождеством:

$$
\boxed{\cosh x = \frac{1}{\cos\left(\arcsin\left(\frac{e^{2x} - 1}{e^{2x} + 1}\right)\right)}}
$$

Здесь:

- аргумент $x$ — это **рапидность** (гиперболический угол Лоренца);  
- $\frac{e^{2x} - 1}{e^{2x} + 1} = \tanh(x)$ — это **относительная скорость в долях $c$**;  
- вложенные функции $\arcsin$ и $\cos$ восстанавливают угол в евклидовом пространстве, для которого гиперболическое растяжение эквивалентно обычному повороту.  

Таким образом, можно интерпретировать гиперболический сдвиг вдоль временной оси как евклидово вращение в фазовом пространстве:

$$
\cosh x \quad \longleftrightarrow \quad \frac{1}{\cos \varphi}, \quad \text{где } \sin \varphi = \tanh x.
$$

Это соотношение устанавливает явный **мост между гиперболическим и евклидовым описанием**:  
- в СТО стандартное преобразование выражается через $\cosh$ и $\sinh$,  
- в евклидовой фазовой модели то же самое выражается через тригонометрические функции с углом $\varphi$, зависящим от скорости.  



#### 2.5.1.  Перемещение потока
Зададим два потока через прото-скорости:
- $ \vec{V}_A = \tilde{A} \ \mathbf{e}_{hA} = \tilde{a}_0 \ \mathbf{e}_{hB} + \tilde{a}_1 \ \mathbf{e}_{l} = \vec{V}_B \ e^{\theta \mathbf{u}} $
- $ \vec{V}_B = \tilde{B} \ \mathbf{e}_{hB} = \tilde{b}_0 \ \mathbf{e}_{hB} + \tilde{b}_1 \ \mathbf{e}_{l} = \vec{V}_A \ e^{-\theta \mathbf{u}} $
- $ \tilde{H} = |\vec{V}_A| = |\vec{V}_B| = \sqrt{\tilde{a}_0^2 + \tilde{a}_1^2} = \sqrt{\tilde{b}_0^2 + \tilde{b}_1^2} = |\tilde{A}| = |\tilde{B}| = \vec{A} \ \bar{\vec{A}} = \vec{B} \ \bar{\vec{B}} $  

Если перемещением вектора $ \vec{V}_A $ в позцию $ \vec{V}_B $ является поворот на $ \theta $ в прото-пространстве, то воспринимаемое пространство для объекта будет просто:
$$
dl = \left\Vert \frac{\vec{V}}{d\theta} \right\Vert d\theta \quad \overset{\left\Vert \mathbf{u} \right\Vert = 1}{\Longrightarrow} \quad  dl = |\vec{V}| d\theta 
$$
Следовательно, "пространство" есть многообразие всех возможных ориентаций $ \vec{V} $ при фиксированной норме - окружность для 2D, но принимая во внимание трёхмерность наблюдаемого пространства - $ S^3 $.  
А время связано с внутренним поворотом (циклом внутри $ \vec{V} $), то есть с компонентой, ортогональной внешнему повороту:
$$
\mathtt{c} d\tau = |\vec{V}| d\chi_\tau
$$
Нормирующий фактор возникает как отношение внутренних (темпоральных) фаз объектов, в котором фаза наблюдателя принята равной единице:
$$ 
\dot{\chi} = \frac{1}{\cos\theta} = \frac{1}{\sqrt{1 - \sin^2\theta}} = \frac{1}{\sqrt{1-\beta^2}} = \frac{1}{\sqrt{1-\frac{V^2}{\mathbf{c}^2}}}
$$
Тогда можено показать как гиперболическая форма интервала:
$$
ds^2 = \mathtt{c}^2 d\tau^2 - dl^2 = \mathtt{c}^2 dt^2
$$
Возникает из евклидового поворота прото-скорости нормированного изменением частоты:
$$
\mathtt{c}^2 d\tau^2 \ \dot{\chi}^2 = \mathtt{c}^2 dt^2 \ \dot{\chi}^2 + dl^2 \ \dot{\chi}^2
$$

Резюмируя выше написанное, можно утверждать, что гиперболическое преобразование специальной теории относительности в терминах времени-пространства может быть представлено поворотом в евклидовом фазовом пространстве.  

Покажем на простом примере. Пусть собственная фаза Алисы, находящейся на поверхности Земли, $ \chi_A = 1 \ [\mathtt{рад}] \ (\mathbf{e}_{hA}) $. Боб летит на Альфу Центавра за Джумбо-багетом. Его скорость в ИСО Алисы $ \beta = V / \mathtt{c} = 0.5 [] $, что в единицах фазы Алисы $ \tilde{\beta} = \chi_A \cdot \beta = 1 \cdot 0.5 \ (\mathbf{e}_{l}) = 0.5 \ [\mathtt{рад}] \ (\mathbf{e}_{l}) $. Фаза Боба в восприятии Алисы будет корнем из евклидовой разности частоты Алисы и своей скорости в её восприятии: $ \chi_B = \sqrt{\chi_A^2 - \tilde{\beta}^2} = \sqrt{1 - 0.25} = 0.866 \ [\mathtt{рад.}] \ (\mathbf{e}_{hA}) $, что значит его время течёт в 0,866 раз медленнее.
<div class="alert alert-block alert-info"> <b>Примечание.</b> Если принимать фазу Алисы отличной от единицы, относительную скорость Боба пришлось бы <i>перенормировать в единицах фазы Алисы</i> - за большую фазу он пройдёт больше пространства: \(\tilde{\beta}=\chi_A\,\beta\). Тогда \(\chi_B=\sqrt{\chi_A^2-\tilde{\beta}^2} =\chi_A\sqrt{1-\beta^2}\), и, следовательно, \(\chi_B/\chi_A=\sqrt{1-\beta^2}=1/\gamma\). С единицей всё арифметически проще.  
</div>

### 2.6. Парадокс близнецов
В представлении униметрии парадокс близнецов разрешается через наличие движения относительно гравитирующих объектов. Близнец менее подвижный в гравитационном поле состарится больше - его частота выше, так как он не тратит фазу на движение поперёк потока.

В случае рассмотрения лишь двух потоков с угловым отклонением при гравитационном фоне близком к нулю, концептуальным является решение о равноправности / исключительности потока одного из объектов. СТО утверждает равноправие систем отсчёта, но достаточных эмпирических данных пока нет. Иными словами, мы не можем пока точно сказать, является ли изменение скорости течения времени объекта реальным процессом (исключительная система существует) или лишь "дефектом" восприятия (системы равноправны).  
В потоковом представлении есть возможность свести наличие объективного замедления времени (уменьшения фазы) движущегося объекта к наличию движения относительно направления вектора гравитационного потока, который существует всегда: если гипотетически представить, что весь существующий поток составляют только два идентичных объекта, движущиеся относительно друг друга, то суммарный вектор гравитационного потока, создаваемого ими будет геометрической суммой векторов их потоков, разделённой пополам. 




### 2.7. Фазовая разность излучения

#### 2.7.1. Определения

- $\phi_s$ — фаза излучения в системе источника  
- $\phi_0$ — фаза, измеренная наблюдателем  
- $\omega_s = \frac{d\phi_s}{d\tau_s}$ — частота источника  
- $\omega_0 = \frac{d\phi_0}{d\tau_0}$ — наблюдаемая частота  
- $\beta = V / \mathtt{c}$, $\gamma = 1/\sqrt{1-\beta^2}$  

Продольный релятивистский Доплер:

$$
r \equiv \frac{\omega_0}{\omega_s} = \sqrt{\frac{1-\beta}{1+\beta}}
$$

#### 2.7.2. Связь фазы и частоты

Частота — это изменение фазы по собственному времени:

$$
\omega_s = \frac{d\phi_s}{d\tau_s}, \quad \omega_0 = \frac{d\phi_0}{d\tau_0}
$$

Тогда:

$$
r = \frac{\omega_0}{\omega_s} = \frac{d\phi_0 / d\tau_0}{d\phi_s / d\tau_s} = \frac{d\phi_0}{d\phi_s} \cdot \frac{d\tau_s}{d\tau_0}
$$

С учётом релятивистского замедления времени:

$$
\frac{d\tau_s}{d\tau_0} = \gamma \quad \Rightarrow \quad \frac{d\phi_0}{d\phi_s} = \frac{r}{\gamma}
$$

#### 2.7.3. Подставим выражения

$$
\frac{d\phi_0}{d\phi_s} = \frac{ \sqrt{\frac{1-\beta}{1+\beta}} }{ 1/\sqrt{1-\beta^2} } = \sqrt{\frac{1-\beta}{1+\beta}} \cdot \sqrt{1-\beta^2} = 1 - \beta
$$

#### 2.7.4. Формула фазовой разности

Определим фазовую разность как

$$
\Delta \phi = \phi_s - \phi_0
$$

Тогда при малых интервалах:

$$
d\Delta \phi = d\phi_s - d\phi_0 = d\phi_s - (1-\beta)d\phi_s = \beta \, d\phi_s
$$

Или в виде со "скоростью":

$$
 \frac{ \Delta \phi }{\phi_s} = \frac{ V }{\mathtt{c}} \qquad (2.7)
$$

#### 2.7.5. Интерпретация

- $\beta = 0 \Rightarrow d\Delta \phi = 0$ — фаза инвариантна  
- $\beta \neq 0$ — относительное движение приводит к изменению фазы пропорционально $\beta$  
- $ V < 0 $ — объект движется к наблюдателю, фаза увеличивается
- $ V > 0 $ — объект движется от наблюдателя, фаза уменьшается

### 2.7.6. Фазовый нормализующий фактор
Мы предположили, что фотон не имеет частоты, которая возникает только как результат его восприятия сложным объектом. Тогда можем предположить, что полное релятивистское отношение частот является преобразованием фазы самого объекта. Тогда здесь возникает противоречие:
$$
    \dot{\chi} = \frac{\chi_{\mathtt{obs.}}}{\chi_{\tau}} = \frac{\tilde{H}}{\tilde{S}} = \frac{\tilde{H}}{\tilde{H} \, \cos \zeta} = \frac{1}{\cos \zeta}
$$
где параметризирующий угол цета связан со скоростным поворотом следующим образом: $ \sin \zeta = \beta $, значит:
$$
    \dot{\chi} = \frac{1}{\cos \zeta}
$$


Следовательно, далее мы будем исходить из (1) фундаментальности прото-параметра и вторичности времени и пространства и (2) геометрической интерпретации времени.  
Первое означает, что время-пространство создаётся изменением в отношениях характеристик объектов. Второе - темпоральную ось можно представить замкнутой осью варьируемой величины (прото-скорость), тогда изменение течения времени будет происходить в результате переноса одной и той же прото-длины в системы отсчёта с меняющийся геометрической структурой.  
Предположим, что прото-функция может быть описана как бесконечно малый элемент длины в полярных координатах: 
$$ dH^2 = dR^2 + R^2 d\phi^2 \iff dH^2 = d\tilde{H}^2 + \tilde{H}^2 d \chi^2 \qquad (2.5.1) $$
Принимая, что правая часть (1.3) состоит из "внутренней" $ \tilde{S} $ и "внешней" $ \tilde{L} $ динамики, разложим угловую компоненту по двум осям $ d\chi^2 = d\chi_s^2 + d\theta^2 $, и подставим в (2.5.1):
$$ dH^2 = dR^2 + R^2 d\phi_s^2 + R^2 d\phi_l^2 \iff dH^2 = d\tilde{H}^2 + \tilde{H}^2 d \chi_s^2 + \tilde{H}^2 d \theta^2 \qquad (2.5.2) $$
Мы получили тот же объект, но с альтерацией по углу обзора, с естественным сохранением нормы при поворотах (ортогональных преобразованиях).
Как мы заметили в 2.3 скорость потока должна быть единой для обеспечения правильного проецирования Хоры в метрику Минковского, что вполне уместно геометрически представляется как сохранение радиуса в (2.5.2). Теперь мы можем разделить эффекты, на возникающие при постоянстве радиуса и варьировании полной фазы по внутренней - эффекты СТО, и прочие вариации - эффекты ОТО.

$$ \begin{matrix} \left[ \frac{ \mathtt{прото-пространство} }{ \mathtt{фаза} } \right] \ \hat{\mathbf{e}}_{L} = \frac{1}{2\pi} \left[ \frac{\mathtt{длина}}{\mathtt{фаза}} \right] & \Rightarrow & 2 \pi \ [\mathtt{прото-пространство}] \ \hat{\mathbf{e}}_{L} = [\mathtt{длина}] & \Rightarrow \\ 2 \pi \  [\mathtt{прото-пространство}] \ \hat{\mathbf{e}}_{S} = [\mathtt{длина}] & \Rightarrow & \frac{2 \pi}{ \mathtt{c}} \  [\mathtt{прото-пространство}] \ \hat{\mathbf{e}}_{S} = [\mathtt{время}] \end{matrix} \qquad (2.6.1)  $$

Равно, как и для преобразования времени в длину фаза не требуется, потому что отношение времени и пространства фиксировано через константу скорости света, которая в представлении униметрии является прото-скоростью реализации динамики объектов (скоростью полной фазы), приведённой к локальным единицам времени.

### 2.8.1. Кватернион излучения
Мы показали, что эффект СТО воспроизводим в терминах униметрии, но задачей перехода к нативным метрическим единицам было упростить всё, а не наоборот. Поэтому попытаемся теперь взглянуть на мир через призму униметрии - фазу - напрямую. Удобно будет это сделать на примере.  
Положим на расстоянии $ D [m] $ от приёмника находится источник излучения. Примем декартову 3-ёхмерную систему координат с приёмником в начале, тогда произвольное направление на объект может быть задано единичным мнимым кватернионом, скажем, $ \mathbf{ u } \in \mathbb{H}: |\mathbf{u}| = 1, \Re (\mathbf{u}) = 0 $.  
Представим, что приняли с этого направления фотон с длиной волны $  \tilde{H}_1 \equiv \lambda_1  $ равной прото-скорости фотона. Если записать кватернион в таком виде:  
$$
\mathbf{p} = \cos \Phi_1 + \mathbf{u} \sin \Phi_1 = e^{\Phi_1 \mathbf{u}}, \quad \Phi_1 = \frac{2 \pi \ D}{ \lambda_1 }
$$
то получим некоторый кватернион "положения", в котором роль расстояния выполняет фаза.  
Однако, в таком виде фотону каждой длины волны будет соответствовать особый кватернион, что неудобно, поэтому дополнительно нормируем его длиной волны $ \lambda_1 / 2\pi = e^{h_1} $, тогда получится:
$$
\mathbf{P} = \frac{\lambda_1}{2\pi} \ \mathbf{p} = \frac{\lambda_1}{2\pi} \ e^{\Phi_1 \mathbf{u}} = e^{h_1 + \Phi_1 \mathbf{u}} = e^{\mathbf{z}_1}: \quad \int_0^{\Phi_1}{ \left| \frac{ d\mathbf{Q} }{ d\Phi } \right| d\Phi  } = \int_0^{\Phi_1}{ \frac{\lambda_1}{2\pi} d\Phi  } = \frac{\lambda_1 }{2\pi} \Phi_1 = \frac{\lambda_1 }{2\pi} \cdot \frac{2 \pi \ D}{ \lambda_1 } = D, \forall \lambda_1
$$
что даст универсальный кватернион, который можно подстраивать по нужной длине волны, получая соответствующую ей фазу, но для любого случая длина траектории точки на 3-сфере по изменению фазы сохраняется постоянной и равна расстоянию в привычном понимании.  
Используя эту функцию в качестве нормы на пространстве кватернионов можем ввести ввести отображение:
$$
\mathbf{Q} \mapsto \vec{D}
$$
извлекающее дистанцию как:
$$
\left| \vec{D} \right| = \int_0^{\Phi_1}{ \left| \frac{ d\mathbf{Q} }{ d\Phi } \right| d\Phi  } = D
$$
Иначе говоря, кватернион положения после нормировки напрямую задаёт наблюдаемую дистанцию и направление на объект.  
Очевидно, тогда верно:
$$
\tilde{H} = \frac{d\vec{D}}{ d\Phi } = \left| \frac{ d\mathbf{Q} }{ d\Phi } \right| 
$$
Посмотрим на разложение квадратичной формы данного кватерниона на билинейные отображения:
$$
\tilde{H}^2 = \left| \frac{ d\mathbf{Q} }{ d\Phi } \right|^2 = \frac{\lambda_1^2}{4 \pi^2} (-\sin \Phi_1 + \mathbf{u} \cos \Phi_1)^2 
$$
