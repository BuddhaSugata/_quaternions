# Униметрия. Прото-функция

Мы привыкли воспринимать эту реальность как окончательную, как «нулевой уровень» бытия. Но что, если она — лишь производная?  
Что, если всё, что мы видим и ощущаем, — всего лишь тени, как в [пещере Платона](https://ru.wikipedia.org/wiki/%D0%9C%D0%B8%D1%84_%D0%BE_%D0%BF%D0%B5%D1%89%D0%B5%D1%80%D0%B5), а подлинная первооснова скрыта глубже?  

Униметрия рассматривает реальность не как фундамент, а как результат — отображение динамики более простого, но недоступного нам напрямую «прото-пространства».  
Я называю его *Хора* — средой, где нет привычных нам понятий времени и пространства. Здесь нет событий в привычном смысле — есть лишь потоки.  
Эти потоки не разворачиваются «во времени» и не занимают «место»; они существуют как векторные сущности, чьи ориентации относительно друг друга и порождают то, что мы ощущаем как пространство и время.  

В Хоре можно выделить два типа потоков:  
- **внутренние компоненты** — формируют локальное (собственное) время;  
- **внешние компоненты** — формируют протяжённость, которую мы называем пространством.  

Поток в Хоре движется с определённой *прото-скоростью* (~v), которая описывается через ориентацию его внутренних и внешних составляющих.  
Меняя эту ориентацию, мы получаем различие между фотоном и массивным объектом, между быстрым и медленным течением времени, между локальной и глобальной геометрией.  

Таким образом, привычная метрика пространства-времени — это не фундаментальный объект, а производная, конформно масштабированная проекция движения в Хоре.  
И если принять эту точку зрения, то релятивистские эффекты можно описать не через гиперболическую геометрию, а через евклидову, просто вводя масштабный фактор, зависящий от ориентации потока.

$$ d\chi = k \tilde{X}_0 \ d\tau \qquad (6.8) $$
В том смысле, что если слева величина единая для течения времени и пространства, которая будет сохраняться при её интегрировании в любых обстоятельствах, то выражение справа - тоже инвариант. Можно показать это несколько иначе, заметив, что единый прото-параметр задаёт величину, которая будет инвариантна для преобразований затрагивающих и время, и пространство (из 6.4):
$$ d\chi = k \tilde{X}_0 \ d\tau \qquad (6.8) $$
В том смысле, что если слева величина единая для течения времени и пространства, которая будет сохраняться при её интегрировании в любых обстоятельствах, то выражение справа - тоже инвариант. И как видно из сравнения (6.4) и (6.6), частотная сущность воспринимаемого времени ведёт к инвариантности относительно себя именно "вывернутой" формы - интервала, а не полной энергии.  
Также из (6.8) следует, что для объекта фиксированной сложности ($ k = const $) величина $ \tilde{X}_0 $ должна компенсировать изменение течения времени. Нагляднее будет его переписать так:
$$ d\chi =  \frac{ \dot{X}_0 \ d\tau }{\tilde{X}_0}  \qquad (6.9) $$
Скорость света в единицах локального времени постоянна, пока постоянна скорость потока относительно прото-параметра.  


Из наблюдений мы можем заключить, что сама метрика при рассмотрении пространственной скорости сохраняется, то есть полная скорость потока $ \tilde{\mathcal{H}} = const $ перераспределяется внутри континуума. Но этого самого по себе недостаточно, чтобы сохранить $ \dot{X}_0 = const = k \tilde{X}_0^2 $

    ## 4. Частота
Возьмём релятивистскую формулу доплеровского изменения частоты как зависимость, описывающую относительное изменение частоты, и включающую в себя все затрагиваемые эффекты (доплеровский эффект и релятивистское движение, не затрагивает вращение и гравитацию):
$$ \omega_1 = \omega_0 \sqrt{ \frac{ 1 - \frac{ v }{ \mathtt{c} } }{ 1 + \frac{ v }{ \mathtt{c} } } } $$
($ v $ положительна в направлении приёмника)  
и выразим в обратную сторону как отношение скорости к скорости света через отношение частот:
$$ \frac{ v }{ \mathtt{c} } = \frac{ 1 - \frac{ \omega_1^2 }{\omega_0^2} }{ 1 + \frac{ \omega_1^2 }{\omega_0^2} } $$
Теперь выразим гамма-фактор в терминах частот:
$$ \gamma = \frac{1}{\sqrt{1 - \frac{ v^2 }{ \mathtt{c}^2}}} = \frac{ \omega_1^2 + \omega_0^2 }{ 2 \omega_1 \omega_0 } $$
С другой стороны мы знаем, что гамма-фактор - это отношение скорости течения времени для неподвижной и подвижной ИСО:
$$ \gamma = \frac{ d\tau_0 }{ d\tau_1 } $$
$ d\tau_0 > d\tau_1 $ - наблюдаемое время обстоятельств источника течёт медленнее, чем приёмника, и частота наблюдаемого сигнала $ \omega_1 $, соответственно, ниже испускаемого $ \omega_0 $.  
Подставим:
$$ \frac{ d\tau_0 }{ d\tau_1 } = \frac{ \omega_1^2 + \omega_0^2 }{ 2 \omega_1 \omega_0 } = \frac{1}{2} \left( \frac{\omega_1}{\omega_0} + \frac{\omega_0}{\omega_1} \right) $$
Примем
$$ \omega = \frac{d\phi}{d\tau} $$
Из чего после некоторых несложных преобразований получим:
$$ \gamma^{-2} = \frac{d\tau_1^2}{d\tau_0^2} = 2 \frac{ d\phi_1 }{ d\phi_0 } - \frac{ d\phi_1^2 }{ d\phi_0^2 } $$
Что с другой стороны:
$$ \gamma^{-2} = 1 - \frac{ V^2 }{ \mathtt{c}^2 } $$
Следовательно:
$$ \frac{ V^2 }{ \mathtt{c}^2 } = 1 - 2 \frac{ d\phi_1 }{ d\phi_0 } + \frac{ d\phi_1^2 }{ d\phi_0^2 } = \left( 1 - \frac{ d\phi_1 }{ d\phi_0 } \right)^2 $$
Избавимся от квадратов и введём $ \Delta = |\phi_0 - \phi_1| $:
$$ \mathtt{c} d \Delta = V d\phi_0 $$

## 3. Простой объект
Простой объект (фотон) движется только по внешней оси Хоры, а сложный — имеет проекцию на внутреннюю: частицы сложного объекта должны иметь проекции на внешние оси для формирования наблюдаемого образа объекта. Вместе с тем, в зависимости от фазовых характеристик наблюдателя внешний образ создаваемый фотонами будет меняться. Можно предположить, что невидимый на нашем уровне восприятия образ обеспечивает эффект гравитации - скорость потока в точке пространства является скоростью течения фаз объектов, что влияет на их временные характеристики.
Иначе говоря, фотоны формируют наблюдаемое 3-мерное пространство, однако его характеристики зависят от самого объекта-приёмника, потому что и фотон, и приёмник связаны изменением единого фактора - скоростью потока.
Фотон имеет собственную прото-скорость, равную его длине волны. Масштаб восприятия фотона приёмником задаётся отношением их прото-скоростей.

### 3.1. Аналитическое представление
Будем называть простой объект _фотоном_ и характеризовать его количественно прото-скоростью $ \tilde{v}_p = dx_p / d\chi $, что с использованием (2.3.1) будет:
$$ \textcolor{blue}{ \dot{\chi}_p } = \left(\frac{d\chi_p}{d\tau_p} \right) = (\nu_p) =  \frac{ \mathtt{c} }{ \tilde{H} } \overset{\tilde{H} = \tilde{v}_p}{\implies} (\nu_p) =  \frac{ \mathtt{c} }{ \tilde{v}_p } = \frac{ dx_h }{ dx_p } \frac{ d \chi_p }{ d\tau_p } \implies dx_h = dx_p = dx_l; \ [\lambda_p] = \tilde{v}_p \qquad (3.1.1) $$
И при условии $ \tau = \chi \Rightarrow (\nu_p) = 1 $ из (2.4.3) даст собственный интервал фотона (задающий время, по которому скорость фотона в прото-пространстве равна скорости света), необходимо лишь формально разделить оси.  
Такой интервал задаёт прото-длину, на которую может успеть сместиться сложный объект по внутренней оси пока фотон смещается по внешней, равную смещению фотона, __если прото-скорости объекта и фотона равны__. Как если бы фотон принимался равноценным фотоном, но движущимся "ненаблюдаемо" (то есть уже массивным объектом).  
Также здесь видно, что прото-скорость фотона это просто его длина волны. Отсюда, наконец, можно связать единицы измерения длины с прото-пространством:

А время, получается, должно бы являться функцией скорости фазы, но так как наблюдение и времени, и пространства зависят от одного параметра, то движение по темпоральной оси гораздо проще выразить через константу скорости света.  
Теперь мы можем переписать интервал (2.3.3) для фотона с использованием новых терминов:
$$ ds^2 = \left(  \lambda_p^2 \hat{\mathbf{e}}_{S}^2 - \nu_p^2 \hat{\mathbf{e}}_{L}^2 \right) \ d\chi^2 = \mathtt{c}^2  \left( \nu_p^{-2} \hat{\mathbf{e}}_{S}^2 - \lambda_p^{-2} \hat{\mathbf{e}}_{L}^2 \right) \ d\chi^2 \qquad (3.1.3) $$
Однако первый компонент не является частью активности фотона, он лишь эквивалент его движения в пространственной части, и по сути вторичен, когда сам вектор несёт информацию о частоте или длине волны.  
В остальном это весьма и весьма вычурная и нежизнеспособная модель, потому что такой массивный объект получился бы, вероятно, нестабилным, как минимум, ввиду трёхмерности наблюдаемого пространства, поэтому введём произвольный сложный объект-наблюдатель (приёмник), обстоятельства приёма которого характеризуются прото-скоростью только по ненаблюдаемой оси (массивный объект, находящийся в покое в собственной ИСО): $ \tilde{H}_o = \frac{ dx_h }{ d\chi } $. Тогда его частота относительно себя будет также единичной, порождая собственное локальное восприятие времени: 
$$ \textcolor{blue}{ \dot{\chi}_o } = \left(\frac{d\chi}{d\tau_o} \right) = (\nu_o) = \frac{ \mathtt{c} }{ \tilde{H} } \overset{\tilde{H} = \tilde{H}_o}{\implies} (\nu_o) = \frac{ \mathtt{c} }{ \tilde{H}_o } = \frac{ dx_h }{ dx_o } \frac{ d \chi }{ d\tau_o } \implies dx_h = dx_o \qquad (3.1.4) $$
Однако, частота фотона $ p $ в восприятии объекта $ o $ будет просто:
$$ \mathtt{c} =  \nu_o \ \lambda_o = \nu_p \ \lambda_p \ \Rightarrow \ (\nu_{p\rightarrow o}) |_{\lambda_o = 1} = \frac{\mathtt{c}}{ \lambda_p } = \frac{\mathtt{c}}{ \tilde{v}_p } \qquad (3.5) $$
Грубо говоря, собственная прото-скорость $ \tilde{H}_o = \lambda_o $ сложного объекта масштабирует принимаемый фотон своим течением фазы. Следовательно, область применимости специальной теории относительности в терминах униметрии - объекты одинаковой собственной фазовой скорости потока при наличии взаимной скорости в наблюдаемом пространстве.   

### 3.2. Геометрическое представление
Естественно будет представить фазу углом, что в совокупности с трёхмерностью наблюдаемого пространства даёт 3-сферу некоторого радиуса. Причём если принять образ прото-функции $ H = R \phi $, то  радиус будет прото-скоростью:  
$$ \tilde{H} = \frac{dH}{d\phi} = R \qquad (3.2.1) $$
Тогда представим прото-функцию через произвольное перемещение в радиальных координатах:
$$ dH^2 = dR^2 + R^2 d\phi \ \Rightarrow \ \breve{H}^2 = \frac{dH^2}{dR^2} = \left( 1 - \frac{ R^2 d\phi^2 }{ dH^2 } \right)^{-1} = \left( 1 - \frac{ R^2 \dot{\phi}^2 }{ \dot{H}^2 } \right)^{-1} \qquad (3.2.2) $$
Мы ранее в (2.3.1) уже приняли изменение прото-функции по времени скоростью света, осталось заметить, что наличие фазовой скорости в таком представлении является не чем иным, как перемещением в наблюдаемом пространстве, а в данном контексте, перемещением на 3-сфере, совершаемом за некоторое локальное время, тогда очевидно производная прото-функции по радиусу даст гамма-фактор:
$$ \breve{H}^2 = \frac{dH^2}{dR^2} = \left( 1 - \frac{ R^2 \dot{\phi}^2 }{ \dot{H}^2 } \right)^{-1} \sim \left( 1 - \frac{ V^2 }{ \mathtt{c}^2 } \right)^{-1} = \gamma^2 \qquad (3.2.3) $$
И не повлияет на истинность предыдущих умозаключений по фазе ввиду принятого ранее условия относительной неизменности прото-скорости - радиуса. Иначе говоря, (3.2.2) даёт (3.2.1) при $ R = \mathtt{const} $.  
Наличие фазовой скорости объекта изменяет радиальную прото-скорость при условии сохранения полного количества движения. Можно сказать, что внешне-пространственное движение отнимает часть движения внутри-пространственного.  
Получилось неплохо, значит, во-первых, у нас есть второй прото-параметр $ \chi_2 = R $, а во-вторых, если запишем выражение полного количества движения в Хоре:
$$ dH^2 = \frac{\partial H^2}{\partial R^2} dR^2 + \frac{\partial H^2}{\partial \phi^2} d\phi^2 = \breve{H}^2 dR^2 + \tilde{H}^2 d\phi^2 = \frac{dH^2}{d\tau^2} d\tau^2 = \mathtt{c}^2 d\tau^2 \qquad (3.2.4) $$
И покажем его в виде соответствующем современному интервалу:
$$ \breve{H}^2 dR^2 = \mathtt{c}^2 d\tau^2 - \tilde{H}^2 d\phi^2 \sim ds^2 = \mathtt{c}^2 d\tau^2 - dl^2 \qquad (3.2.5) $$
То получится, что радиальное движение объекта представлет внутреннее, ненаблюдаемое движение объекта, формирующее его собственную циклическую структуру, которая влияет на восприятие времени. Однако происходит это иначе, чем в фазовой части. Характер связи между радиусом и локальным временем таков:
$$ d\phi_1 > d\phi_0 \iff \ d\tau_1 < d\tau_0, \ \ dR_1 > dR_0 \iff \ d\tau_1 > d\tau_0 \qquad (3.2.6) $$
Если фаза была обратна локальному времени, то радиус - пропорционален: чем больше радиус, тем медленнее происходят идентичные процессы (больше дифференциал времени), и из подобия (3.2.3) можно попробовать утверждать просто $ R = k \tau , \ k \in \mathbb{R} $.  
Тогда отношение радиусов (прото-скоростей) есть отношение времён:
$$ \frac{R_0}{R_1} = \frac{\tau_0}{\tau_1} = \frac{\breve{H}_1}{\breve{H}_0} \qquad (3.2.7) $$
Дальше, можем также связать фазовую и радиальную производные прото-функции через скорость света и выразить радиальную производную:
$$ dx_h = \tilde{H} \ \dot{\phi} \ d\tau = \breve{H} \ \dot{R} \ d\tau = \mathtt{c} \ d\tau \quad \Rightarrow \quad \dot{R} = \frac{\tilde{H}}{\breve{H}} \dot{\phi} \qquad (3.2.8)  $$
Значит, в собственной системе отсчёта с единичной частотой $ \nu = \dot{\phi} = 1 $ радиальная нормировка аналогичного объекта относительной пространственной скорости ($ R d \Delta(\phi) = 0 $) можно представить так:
$$ \dot{R} = (\tilde{H}) \left[ \frac{1}{\breve{H}} \right] \sim  \ \mathtt{???} \qquad (3.2.9) $$

### 4.1. Фаза времени (сложного объекта)
Представим, что существуют два количественно разных сложных объекта: $ H_0 = k H_1, \ k > 1 $. В собственных системах, где они находятся в покое, их течение времени будет отличаться прямо пропорционально их количеству прото-движения:
$$ dH_0 = \mathtt{c} \tau_0 , \ dH_1 = \mathtt{c} \tau_1, \ dH_0 = k dH_1 \Rightarrow d\tau_0 = k d\tau_1 \qquad (4.1.1) $$
Что означает, что у большего объекта время течёт медленнее в k раз.  


\sqrt{ \left(\frac{\partial H}{\partial \chi_s} \chi_s \right)^2 + \left( \frac{\partial H}{\partial \chi_l} \chi_l \right)^2 }, \ H \in \mathbb{R}^n


:
$$ dH^2 = \breve{H}^2 d\phi_\tau^2 + \tilde{H}^2 d\phi_l^2 = \frac{dH^2}{d\tau^2} d\tau^2 = \mathtt{c}^2 d\tau^2 \qquad (4.1.1) $$
Это уже 4-сфера, но мы утрируем её за счёт невовлечения незатрагиваемых осей пространства. Остаётся просто окружность, где по мнимой оси развёрнута скорость внешней фазы, а по вещественной - внутренней (времени). Тогда отношение их прото-скоростей можно снова параметрировать углом, а из постоянства полного количества протодвижения объекта выводятся следующие эффекты.
Ранее мы уже получили гамма-фактор в качестве зависимости изменения потока по времени наблюдателя от изменения внешней фазы:
$$ \gamma = \frac{dH}{\breve{H} d\phi_\tau} = \left( 1 - \frac{\tilde{H}^2 d\phi_l^2}{\mathtt{c}^2 d\tau^2} \right)^{-1/2} = \left( 1 - \frac{V^2}{\mathtt{c}^2} \right)^{-1/2} \qquad (4.1.2) $$
Теперь можем получить закон изменения потока в наблюдаемом пространстве сложного объекта:
$$ \tilde{H}^2 d\phi_l^2 = \mathtt{c}^2 d\tau^2 \left( 1- \frac{\breve{H}^2 d\phi_\tau^2}{\mathtt{c}^2 d\tau^2} \right) \sim g_{00} = \left( 1 - \frac{2 G M}{\mathtt{c}^2 r} \right)^{1/2} \qquad (4.1.3) $$
$$ \xi = \frac{dH}{\tilde{H} d\phi_l} = \left( 1 - \frac{\breve{H}^2 d\phi_\tau^2}{\mathtt{c}^2 d\tau^2} \right)^{-1/2} \sim \left( 1 - \frac{2 G M}{\mathtt{c}^2 r} \right)^{-1/2} \qquad (4.1.3) $$
Мы предполагаем, что эффект гравитационного взаимодействия достигается наложением множества потоков сложных объектов:
$$ \breve{H} = \breve{H}_{ex} + \breve{H}_{self} \qquad (4.1.3) $$
Тогда собственный поток одного из них в точке своей локации будет составлять некоторую долю от общего потока:
$$ \breve{H}_{self} d\phi_\tau = \mathtt{c} d\tau - \breve{H}_{ex} d\phi_\tau = \mathtt{c} d\tau \left( 1 - \frac{\breve{H}_{ex} d\phi_\tau}{\mathtt{c} d\tau} \right) \qquad (4.1.4) $$

Наметим также механизм моделирования эффектов ОТО. Закон сохранения количества полной фазы по внешней фазе в условиях различных прото-скоростей будет формировать сферу вокруг объекта с радиально изменяющимися параметрами:
$$ ] \ dR \ne 0, R_s \ne R_l: \quad dH^2 = dR^2 + R_s^2 d\phi_s^2 + R_l^2 d\phi_l^2 \ \iff \ \frac{d\chi}{d\chi_l} = \left( 1 - \frac{d\chi_s^2}{d\chi^2} \right)^{-1/2} \sim g_{00} \qquad (2.5.4) $$
Изменение фазы потока сложного объекта в его окресности (при отклонении по внешней фазе) будет  
В рамках моделирования эффектов СТО примем радиус в (2.1) постоянным, тогда скорости по прото-параметру будут:
$$ \tilde{H} = \frac{dH}{d\chi} = R, \quad \frac{\partial H}{\partial \chi_s} = R, \quad \frac{\partial H}{\partial \chi_l} = R \qquad (2.1.1) $$
Не сложно видеть, что условие количественного постоянства потока $ H $ в (2.2) накладывает ограничение на его распределение по осям. Воспользуемся этим:
$$ ] \ d\chi_s = \cos \alpha \ d\alpha, \ d\chi_l = \imath \sin \alpha \ d\alpha: \quad \frac{dH}{d\alpha} = R e^{\alpha \imath} = e^{r + \alpha \imath}  \qquad (2.1.2)  $$
В нашем случае и проблема, и её решение заключаются в том, что векторы $ d\chi_s, d\chi_l $ являются для нас базовыми и задают не только масштаб воспринимаемого пространства, но и восприятие времени.

```python
# Code for the Jupyter-style insertion: derive and visualize the bridge rotation(i*phi) -> boost
# - symbolic derivation of eta(chi) = ∫ sec chi dchi = ln(sec+tan)
# - doppler r = exp(-eta) = 1/(sec+tan) = sqrt((1-beta)/(1+beta)) with beta = sin chi
# - numeric checks and plots
# - matrix demonstration: R(i*phi) similarity -> Lorentz boost (real)

import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import pandas as pd
from caas_jupyter_tools import display_dataframe_to_user

# Symbolic derivation
chi = sp.symbols('chi', real=True)
beta = sp.sin(chi)
sec = 1/sp.cos(chi)
tan = sp.tan(chi)

eta_expr = sp.integrate(sp.sec(chi), (chi, 0, chi))  # definite integral from 0 to chi
eta_simpl = sp.simplify(eta_expr)
r_expr = sp.exp(-eta_simpl)
r_simpl = sp.simplify(r_expr)

# Alternative expression using beta
beta_sym = sp.symbols('beta', real=True)
r_from_beta = sp.sqrt((1-beta_sym)/(1+beta_sym))

# Show symbolic results
symbolic_results = {
    'eta(chi)': sp.pretty(eta_simpl),
    'r(chi)': sp.pretty(r_simpl),
    'r(beta)': sp.pretty(sp.simplify(r_from_beta))
}

# Numeric checks
chi_vals = np.linspace(-0.9, 0.9, 11)  # stay within (-pi/2,pi/2) and avoid endpoints
rows = []
for ch in chi_vals:
    b = np.sin(ch)
    secv = 1/np.cos(ch)
    tanv = np.tan(ch)
    eta_num = float(np.log(secv + tanv))
    r_num = float(np.exp(-eta_num))
    r_via_beta = float(np.sqrt((1-b)/(1+b)))
    rows.append({'chi': ch, 'beta=sin(chi)': b, 'eta': eta_num, 'r (from chi)': r_num, 'r (from beta)': r_via_beta})

df = pd.DataFrame(rows)

# Display a small table
display_dataframe_to_user("chi_eta_doppler_table", df)

# Plot chi -> eta and chi -> r
chi_plot = np.linspace(-1.2, 1.2, 400)
beta_plot = np.sin(chi_plot)
eta_plot = np.log(1/np.cos(chi_plot) + np.tan(chi_plot))
r_plot = np.exp(-eta_plot)
r_plot_beta = np.sqrt((1-beta_plot)/(1+beta_plot))

plt.figure(figsize=(8,4))
plt.plot(chi_plot, eta_plot)
plt.title("rapidity η(χ) = ∫ sec χ dχ = ln(secχ + tanχ)")
plt.xlabel("χ (rad)")
plt.ylabel("η")
plt.grid(True)
plt.show()

plt.figure(figsize=(8,4))
plt.plot(chi_plot, r_plot, label='r from χ (1/(sec+tan))')
plt.plot(chi_plot, r_plot_beta, linestyle='--', label='r from β = sinχ (sqrt((1-β)/(1+β)))')
plt.title("Doppler factor r(χ) and verification r(β)")
plt.xlabel("χ (rad)")
plt.ylabel("r")
plt.legend()
plt.grid(True)
plt.show()

# Matrix demonstration: rotation -> analytic continuation -> similarity -> boost
phi = sp.symbols('phi', real=True)
J = sp.Matrix([[0, -1],[1, 0]])
R_theta = sp.exp(phi * J)  # rotation matrix e^{phi J}
# analytic continuation phi -> I*varphi produces complex matrix; show symbolic
varphi = sp.symbols('varphi', real=True)
R_i_varphi = sp.exp(sp.I * varphi * J)
S = sp.diag(1, sp.I)
S_inv = sp.diag(1, -sp.I)  # inverse of diag(1,i) is diag(1,-i)
similar = sp.simplify(S * R_i_varphi * S_inv)

# Simplify to standard boost form
similar_simpl = sp.simplify(similar)

# Numeric example for varphi=0.7
varphi_val = 0.7
R_i_num = np.array(sp.N(sp.exp(sp.I * varphi_val * J), 15), dtype=complex)
S_num = np.array([[1,0],[0,1j]], dtype=complex)
sim_num = S_num.dot(R_i_num).dot(np.linalg.inv(S_num))
sim_num_simpl = np.real_if_close(sim_num, tol=1e-9)

# Prepare output dict
output = {
    'symbolic_results': symbolic_results,
    'similarity_transform_result_symbolic': sp.srepr(similar_simpl),
    'similarity_transform_numeric_varphi': sim_num_simpl.tolist()
}

output

```

# Евклидова интерпретация гиперболического преобразования

В специальной теории относительности преобразования Лоренца традиционно записываются через гиперболические функции.  
Например, для быстроты $\xi = \tanh^{-1}\beta$ (где $\beta = v/c$) имеем:

$$
\cosh \xi = \gamma = \frac{1}{\sqrt{1-\beta^2}}.
$$

Здесь появляется гиперболический косинус.  

Однако можно построить явный мост между гиперболическим и евклидовым представлениями.

---

## Шаг 1. Связь через экспоненту

Используем стандартное определение:

$$
\tanh x = \frac{e^{2x}-1}{e^{2x}+1}.
$$

---

## Шаг 2. Вставка в тригонометрический угол

Рассмотрим выражение:

$$
\cosh x = \frac{1}{\cos\!\left(\arcsin\!\left(\frac{e^{2x}-1}{e^{2x}+1}\right)\right)}.
$$

Тем самым гиперболическая функция переписывается как обратный косинус угла, определённого через арксинус от экспоненциального отношения.

---

## Шаг 3. Интерпретация

- В гиперболическом пространстве: растяжение за счёт $\cosh x$ и $\sinh x$.  
- В евклидовом фазовом пространстве: это же отображение можно понимать как поворот, если параметр $x$ трактовать как «фазовый угол» (связан с $\chi$).  

Таким образом, лоренцево преобразование можно переформулировать как **поворот в евклидовом фазовом пространстве**, при этом гиперболический множитель $\gamma$ выражается через тригонометрическую функцию.

---

## Вывод

Мы получили явный мост:

$$
\cosh x \;\;\longleftrightarrow\;\; \frac{1}{\cos(\theta)}, \quad 
\theta = \arcsin\!\left(\frac{e^{2x}-1}{e^{2x}+1}\right).
$$

Это показывает, что гиперболическое преобразование СТО имеет эквивалентное евклидовое представление, если перейти к описанию через фазовый угол.
