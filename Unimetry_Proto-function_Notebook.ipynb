# Униметрия. Прото-функция

Мы привыкли воспринимать эту реальность как окончательную, как «нулевой уровень» бытия. Но что, если она — лишь производная?  
Что, если всё, что мы видим и ощущаем, — всего лишь тени, как в [пещере Платона](https://ru.wikipedia.org/wiki/%D0%9C%D0%B8%D1%84_%D0%BE_%D0%BF%D0%B5%D1%89%D0%B5%D1%80%D0%B5), а подлинная первооснова скрыта глубже?  

Униметрия рассматривает реальность не как фундамент, а как результат — отображение динамики более простого, но недоступного нам напрямую «прото-пространства».  
Я называю его *Хора* — средой, где нет привычных нам понятий времени и пространства. Здесь нет событий в привычном смысле — есть лишь потоки.  
Эти потоки не разворачиваются «во времени» и не занимают «место»; они существуют как векторные сущности, чьи ориентации относительно друг друга и порождают то, что мы ощущаем как пространство и время.  

В Хоре можно выделить два типа потоков:  
- **внутренние компоненты** — формируют локальное (собственное) время;  
- **внешние компоненты** — формируют протяжённость, которую мы называем пространством.  

Поток в Хоре движется с определённой *прото-скоростью* (~v), которая описывается через ориентацию его внутренних и внешних составляющих.  
Меняя эту ориентацию, мы получаем различие между фотоном и массивным объектом, между быстрым и медленным течением времени, между локальной и глобальной геометрией.  

Таким образом, привычная метрика пространства-времени — это не фундаментальный объект, а производная, конформно масштабированная проекция движения в Хоре.  
И если принять эту точку зрения, то релятивистские эффекты можно описать не через гиперболическую геометрию, а через евклидову, просто вводя масштабный фактор, зависящий от ориентации потока.

$$ d\chi = k \tilde{X}_0 \ d\tau \qquad (6.8) $$
В том смысле, что если слева величина единая для течения времени и пространства, которая будет сохраняться при её интегрировании в любых обстоятельствах, то выражение справа - тоже инвариант. $$ \dot{X}_0^2 \left( d\tau^2 -\frac{dx_1^2}{dx_0^2} d\tau^2 - \frac{dx_2^2} {dx_0^2}d\tau^2 - \frac{dx_3^2}{dx_0^2} d\tau^2 \right) = \left| dx_0 = \dot{X}_0 \ d\tau \right| = \dot{X}_0^2 d\tau^2 - dx_1^2 - dx_2^2 - dx_3^2 \quad (6.7) $$
Значит, можно утверждать, что сохранение гиперболической формы интервала представимо поворотом вектора прото-скорости в евклидовом пространстве с учётом соответствующей геометризации времени.  
Можно показать это несколько иначе, заметив, что единый прото-параметр задаёт величину, которая будет инвариантна для преобразований затрагивающих и время, и пространство (из 6.4):
$$ d\chi = k \tilde{X}_0 \ d\tau \qquad (6.8) $$
В том смысле, что если слева величина единая для течения времени и пространства, которая будет сохраняться при её интегрировании в любых обстоятельствах, то выражение справа - тоже инвариант. И как видно из сравнения (6.4) и (6.6), частотная сущность воспринимаемого времени ведёт к инвариантности относительно себя именно "вывернутой" формы - интервала, а не полной энергии.  
Также из (6.8) следует, что для объекта фиксированной сложности ($ k = const $) величина $ \tilde{X}_0 $ должна компенсировать изменение течения времени. Нагляднее будет его переписать так:
$$ d\chi =  \frac{ \dot{X}_0 \ d\tau }{\tilde{X}_0}  \qquad (6.9) $$
Скорость света в единицах локального времени постоянна, пока постоянна скорость потока относительно прото-параметра.  


Из наблюдений мы можем заключить, что сама метрика при рассмотрении пространственной скорости сохраняется, то есть полная скорость потока $ \tilde{\mathcal{H}} = const $ перераспределяется внутри континуума. Но этого самого по себе недостаточно, чтобы сохранить $ \dot{X}_0 = const = k \tilde{X}_0^2 $

    ## 4. Частота
Возьмём релятивистскую формулу доплеровского изменения частоты как зависимость, описывающую относительное изменение частоты, и включающую в себя все затрагиваемые эффекты (доплеровский эффект и релятивистское движение, не затрагивает вращение и гравитацию):
$$ \omega_1 = \omega_0 \sqrt{ \frac{ 1 - \frac{ v }{ \mathtt{c} } }{ 1 + \frac{ v }{ \mathtt{c} } } } $$
($ v $ положительна в направлении приёмника)  
и выразим в обратную сторону как отношение скорости к скорости света через отношение частот:
$$ \frac{ v }{ \mathtt{c} } = \frac{ 1 - \frac{ \omega_1^2 }{\omega_0^2} }{ 1 + \frac{ \omega_1^2 }{\omega_0^2} } $$
Теперь выразим гамма-фактор в терминах частот:
$$ \gamma = \frac{1}{\sqrt{1 - \frac{ v^2 }{ \mathtt{c}^2}}} = \frac{ \omega_1^2 + \omega_0^2 }{ 2 \omega_1 \omega_0 } $$
С другой стороны мы знаем, что гамма-фактор - это отношение скорости течения времени для неподвижной и подвижной ИСО:
$$ \gamma = \frac{ d\tau_0 }{ d\tau_1 } $$
$ d\tau_0 > d\tau_1 $ - наблюдаемое время обстоятельств источника течёт медленнее, чем приёмника, и частота наблюдаемого сигнала $ \omega_1 $, соответственно, ниже испускаемого $ \omega_0 $.  
Подставим:
$$ \frac{ d\tau_0 }{ d\tau_1 } = \frac{ \omega_1^2 + \omega_0^2 }{ 2 \omega_1 \omega_0 } = \frac{1}{2} \left( \frac{\omega_1}{\omega_0} + \frac{\omega_0}{\omega_1} \right) $$
Примем
$$ \omega = \frac{d\phi}{d\tau} $$
Из чего после некоторых несложных преобразований получим:
$$ \gamma^{-2} = \frac{d\tau_1^2}{d\tau_0^2} = 2 \frac{ d\phi_1 }{ d\phi_0 } - \frac{ d\phi_1^2 }{ d\phi_0^2 } $$
Что с другой стороны:
$$ \gamma^{-2} = 1 - \frac{ V^2 }{ \mathtt{c}^2 } $$
Следовательно:
$$ \frac{ V^2 }{ \mathtt{c}^2 } = 1 - 2 \frac{ d\phi_1 }{ d\phi_0 } + \frac{ d\phi_1^2 }{ d\phi_0^2 } = \left( 1 - \frac{ d\phi_1 }{ d\phi_0 } \right)^2 $$
Избавимся от квадратов и введём $ \Delta = |\phi_0 - \phi_1| $:
$$ \mathtt{c} d \Delta = V d\phi_0 $$
относительная скорость - это изменение фазы.

есть фаза времени, фаза пространства, фаза фотона. получается, что когда объект движется, то фаза времени меняется независимо от направления движения, потому что там (фаза фотона - дельта)(фаза фотона + дельта), а фаза фотона - зависимо, потому что доплер эффект зависит от направления скорости. то есть фазу времени от фазы фотона отличает только дополнительный множитель, где дельта фазы всегда с обратным знаком. вроде как также получается, что фаза фотона совпадает с фазой пространства.
то есть фазу пространства от фазы времени отличает множитель.
