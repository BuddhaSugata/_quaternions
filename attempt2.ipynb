# Фазовый формализм СТО - Униметрия

Мы привыкли воспринимать время и пространство как данность. Фазовый формализм униметрии предлагает посмотреть [иначе](https://ru.wikipedia.org/wiki/%D0%9C%D0%B8%D1%84_%D0%BE_%D0%BF%D0%B5%D1%89%D0%B5%D1%80%D0%B5): время и пространство — это производные от единого параметра, который мы называем фазой.

В этой концепции релятивистские эффекты — замедление времени, доплеровский сдвиг и т.д. — возникают как естественные проявления движения потоков в фазовом пространстве. Формализм не вводит новые законы физики, он лишь предлагает удобный язык и инструмент для их описания. Подобно лагранжевому и гамильтоновому подходу в классической механике, это попытка упростить структуру уже известных зависимостей и сделать явными причинно-следственные связи.

## 1. Время и пространство как производные
Предположим, что существует переменная величина, назовём её __фаза__ $\vec{\chi} \in \mathbb{C}$, изменение которой порождает наблюдаемые эффекты времени и пространства. Алгебраически это будет означать, что время-пространственные базовые векторы (единицы, меры) являются фазовой скоростью - производной фазы по направлениям:
$$
\hat{h} \, dx_0 = \frac{\partial \vec{\chi}}{\partial \chi_h} \frac{d\chi_h}{d\chi} d\chi = \tilde{H} \ d\chi, \ \mathbf{l} \, dx_l = \frac{\partial \vec{\chi}}{\partial \chi_l} \frac{d\chi_l}{d\chi} d\chi = \tilde{L} \ d\chi, \, l = 1,2,3 \qquad (1.1)
$$
Можно сказать, что мы динамизируем тензор пространства-времени через единую общую переменную, тогда, введя фазовую скорость интервала СТО $ds = \tilde{S} \ d\chi$ аналогично ранее представленым прото-скоростям, можем записать закон сохранения интервала в следующем виде:
$$
\tilde{S}^2 = \frac{ds^2}{d\chi^2} = \frac{g_{ij} dx^i dx^j}{d\chi^2} = g(\chi) = \tilde{H}^2 - \tilde{L}^2 \qquad (1.2)
$$
Что равносильно:
$$
\tilde{H}^2 = \tilde{S}^2 + \tilde{L}^2  \qquad (1.3)
$$
Такая форма предоставляет возможность выразить фазовые скорости $\tilde{S}, \ \tilde{L}$ через $\tilde{H}$:
$$ 
\tilde{S} = \tilde{H} \cos\theta, \ \tilde{L} = \tilde{H} \sin\theta \qquad (1.4)
$$
где $\theta$ - угол отклонения фазовой скорости от вещественной оси в фазовом пространстве.  
С математической точки зрения можно сказать, что (1.3) просто утверждает существование параметра, по которому темпоральная компонента интервала может быть разложена как сумма двух проекций на перпендикулярные оси в евклидовом пространстве, то есть если представить, что время-пространство индуцировано изменением единого параметра, то величина скорости по данному параметру вдоль темпоральной координатной оси может быть представлена как сумма скоростей вдоль двух отдельных компонент - пространственной и вне-пространственной (наблюдаемой и ненаблюдаемой, внешней и внутренней).  
Или, упрощая, (1.3) выглядит как формула сложения скоростей в евклидовом векторном пространстве. Предположим, что так и есть, но по какой-то причине при повороте "угла обзора" сохраняется не норма $ \tilde{H} $, а проекция на вне-пространственную $ \tilde{S} $.  

## 2. Фазовое пространство
Пусть математическое векторное пространство называется [__хора__ (Платон)](https://en.wikipedia.org/wiki/Kh%C3%B4ra?utm_source=chatgpt.com) и имеет свойства обыкновенного комплексного пространства $\mathbb{C}$.  
Геометрически мы определим два базисных вектора назвав их (__время__, __пространство__): $ (\hat{h}, \, \mathbf{l}) \equiv \mathbb{C}$; получив в результате представление фазы $\vec{\chi} = R \ e^{\theta \mathbf{l}}, \ \theta \in [-\pi, \pi]$ как вектора в данном базисе со следуюшими отношениями:
$$
\tilde{H} = R, \quad \tilde{S} = R \cos\theta, \quad \tilde{L} = R \ \mathbf{l} \sin\theta \qquad (2.1)
$$
и (1.3) становится __законом сохранения__ полной фазы объекта при повороте вектора фазы.  
Отметим, что вместе с тем, фазовое пространство, скорости которого являются базисами для наблюдаемого пространства-времени, связано с последним интегральным преобразованием вида:
$$
x^i = \int{ \tilde{X}^i(\chi) \ d\chi}, \quad i = 0,1,2,3 \qquad (2.2)
$$
где $\tilde{X}^i(\chi)$ — фазовые скорости, являющиеся проекциями производной фазы $\vec{\chi}$ на соответствующие направления базиса $(\hat{h}, \mathbf{l})$.  
Таким образом, привычные координаты времени и пространства представляют собой интегральные образы фазовых скоростей, а сами фазовые скорости — это «мост» между скрытым фазовым параметром и наблюдаемым многообразием Минковского.  
Ввиду отношений (1.1) и (1.2) фазовое пространство $\mathbb{C}$ выступает как порождающее пространство, а наблюдаемое пространство-время — как его проекционное отображение.  

## 3. Объекты
Любая __фундаментальная частица__ является __элементарным объектом__ униметрии и характеризуется ненулевой величиной фазы ($\vec{\chi} \ne 0$).  
Составные __объекты__ являются фазовыми конфигурациями, для полного представления которых в фазовом пространстве в основном требуется его расширение дополнительными измерениями, кроме случая __фотона__, фаза которого всегда направлена вдоль мнимой оси хоры:
$$
\mathbf{p} = \frac{d\vec{\chi}}{d\chi_l} = p \mathbf{l} \in \Im \qquad (3.1)
$$
Выделим все явления, отличные от перемещения фотонов, в отдельную категорию, аналогично СТО получив фотонное подпространство, которое в униметрии представлено мнимой осью, а остальные, более __сложные__ в этом контексте объекты, характеризуются наличием вещественной проекции фазы и ненулевой массой.
Тогда проводя параллель с СТО, сложный объект униметрии ассоциирован с любым __событием__ или чередой таковых - с мировой линией, а фотон - с точкой подпространства нулевого интервала, содержащей информацию о __событии__.  
Фаза любого объекта может быть приведена поворотом в хоре в _нулевое_ направление (положение нулевого отклонения вектора фазы) вдоль вещественной оси:
$$
\vec{\chi}_o = R \in \Re \qquad (3.2)
$$
Объект A, движущийся относительно покоящегося наблюдателя, будет представлен вектором фазы находящейся под углом к вектору фазы наблюдателя:
$$
\vec{\chi}_a = R \ e^{(\theta_A \, \mathbf{l})}, \, \sin \theta_a = \frac{V}{\mathtt{c}} \qquad (3.3)
$$
таким, что синус угла поворота его вектора фазы равен отношению взаимной скорости объектов к скорости света: $\theta > 0$ - отдаление, $\theta < 0$ - сближение, для перпендикулярного движения знак не важен (фактор квадратичен в уравнении).  
Таким образом, множество относительно неподвижных объектов формируют друг для друга фотонное пространство нулевой кривизны. Такие объекты будем называть _синфазными_.  

### 3.1. Пространство
Из определения (1.4) для пространственной составляющей следует, что в нулевом направлении компонента скорости фазы по мнимой оси хоры равна нулю как и базисный вектор пространства объекта, что противоречит наблюдаемой действительности - как минимум для обеспечения собственной проекции в наблюдаемое пространство у объекта должна существовать проекция фазы на мнимую ось, поэтому примем такую формулировку вырожденной и применимой лишь для малых объектов.  
Формализуем это требование следующим образом. Сложим два вектора фазы с равными пространственными отклонениями, но с разным знаком, и разделим результат на два, чтобы компенсировать норму. Получим:
$$
\vec{\chi}^\pm = R \, e^{\pm \zeta \ \mathbf{l}} \quad \chi_l = \frac{\vec{\chi}^+-\vec{\chi}^-}{2} = R \, \mathbf{l} \sin\zeta   \qquad (3.1.1)
$$
где $\vec{\chi}_l$ выступает в роли пространственного вектора фазы.

Таким образом, полная фаза объекта в _локальном_ направлении представляется как пара:
$$
\vec{\chi}_o = \vec{\chi}_\tau + \vec{\chi}_l = R \ \cos\zeta + R \ \mathbf{l} \sin\zeta \qquad (3.1.2)
$$
что эквивалентно разложению на темпоральную и пространственную компоненты.
При этом:

$\vec{\chi}_t = R \cos\zeta$ — временной базис,

$\vec{\chi}_l = R \, \mathbf{l} \, \sin \zeta$ — пространственный базис.

Если нормировать их на полный радиус $R$, то получаем ортогональные единичные базисы для фазового пространства:
$$
\hat{h} = \cos\zeta, \, \hat{\mathbf{l}} = \sin \zeta \qquad (3.1.3)
$$
Тогда __для единой нормы__ скорости фазы возникает три направления и соответствующих им вещественных значения, которые мы хотели бы обозначить.

### 3.2. Абсолютное, локальное и наблюдаемое время
Нулевому положению фазы соответвует __абсолютное__ время $t = t(\tilde{H})$ - самое медленное. Из 3.1 следует, что не существует объекта, который бы функционировал по этому времени. Однако оно может быть использовано для нормирования при переходе между обстоятельствами фаз разных скоростей.  
В (1.1) мы ввели базисный вектор времени как производную фазы объекта по вещественной компоненте, тогда в локальном направлении вектор фазы всегда даёт его максимально возможное для объекта значение:
$$
dx_0 = \frac{ d\vec{\chi}_o }{ d\chi } d\chi = \frac{\vec{\chi}^+ + \vec{\chi}^-}{2} d\chi = \cos\zeta \, d\chi = d\tau \qquad (3.2.1)
$$
формируя _локальное_ время объекта (в радианах).  
Для значений $\zeta \rightarrow 0: \ d\tau \rightarrow dt$.  
Объект А, перемещающийся относительно наблюдателя, будет иметь _собственное_ - __наблюдаемое__ - время относительно покоящегося наблюдателя в виде вещественной части производной собственного вектора фазы по вектору фазы наблюдателя:
$$
\tilde{H}_a = \Re \left(\frac{d\vec{\tau}_a}{d\vec{\tau}_o} \right) = \Re \left( \frac{d\vec{\chi}_a}{d\vec{\chi}_o} \right) = \frac{\cos\theta_a \ d\chi_o}{d\chi_o} = \cos\theta_a = \sqrt{1-\sin^2\theta_a} = \sqrt{1-\frac{V^2}{\mathtt{c}^2}} = \frac{1}{\gamma}  \qquad (3.1.2)
$$
что находится в согласовании с (3.3), так как является следствием сохранения фазы, а также соответствует экспериментально известному преобразованию времени и симметрично для восприятия объектом А перемещения наблюдателя за счёт геометрического представления (в собственной системе координат объекта А проекция фазы наблюдателя на вещественную ось тоже $\cos\theta_a$).  

### 3.3. Нормирование
Пусть локальное течение времени сложного объекта задано какой-то конекретной частотой $d\tau=d\chi_o$, тогда масштабирование времени можно задать по цепному правилу через нормирующий фактор:
$$ dx_0 = \tilde{H} \ d\chi = \frac{dx_0}{d\chi} \textcolor{blue}{ \frac{d\chi}{d\tau} } \ d\tau = \tilde{H} \textcolor{blue}{ \dot{\chi} } \ d\tau = \dot{H} d\tau = \mathtt{c} d\tau \qquad (3.1.1) $$
Отсюда видно, что мы принимаем скорость потока по локальному времени константой скорости света $ \mathtt{c} $.  
Аналогично для пространственной части:
$$ dx_l = \tilde{L} \ d\chi = \frac{dx_l}{d\chi} \textcolor{green}{ \frac{d\chi}{dl} } \ dl = \tilde{L} \textcolor{green}{ \chi' } \ dl = L' dl, \, l = 1,2,3 \qquad (3.1.2) $$
Из равенства (1.4) для света: $dx_0 = dx_l \Rightarrow \mathtt{c} = \tilde{L}' \frac{dl}{d\tau}$; несложно видеть, что при масштабировании темпоральной оси скоростью света пространственный масштаб всегда будет единичным $ \tilde{L}' = 1 $.  

### 3.4. Свет
Продолжим с (1.4) выразив обе координаты через фазу и нормирующие факторы:
$$ \frac{\mathtt{c}}{ \dot{\chi}} d\chi = \frac{1}{ \chi' } d\chi \qquad (3.4.1) $$
Получается:
$$ \mathtt{c} = \frac{\dot{\chi}}{ \chi' } = \frac{dl}{ d\tau } \ne f(\chi) \qquad (3.4.2) $$
Откуда видно, что в нашей модели скорость света является параметром, связывающим движения по осям __вне зависимости от локального изменения фазы__.  
Во-первых, если посмотреть на (3.4.2) следующим образом:
$$ \mathtt{c} = \frac{\dot{\chi}}{ \chi' } = \left(\frac{d\chi}{d\tau} \right) \left[ \frac{dl}{ d\chi } \right] \sim (\nu) \ [\lambda] \qquad (3.4.3) $$
То можно заметить сходство с характеристиками фотона _частота_ и _длина волны_, произведение которых даёт скорость света.  
Во-вторых, из (3.4.1) и (3.4.2) следует:
$$ ds^2 = \mathtt{c}^2 \left( \frac{d\chi^2}{ \dot{\chi}^2}  - \frac{d\chi^2}{ \dot{\chi}^2} \right) = 0, \forall \dot{\chi} \qquad (3.4.4) $$
где решением для единичной частоты будет $\tau = \chi$ - собственное время фотона это его фаза.
