# Фазовый формализм СТО - Униметрия

Мы привыкли воспринимать время и пространство как данность. Фазовый формализм униметрии предлагает посмотреть [иначе](https://ru.wikipedia.org/wiki/%D0%9C%D0%B8%D1%84_%D0%BE_%D0%BF%D0%B5%D1%89%D0%B5%D1%80%D0%B5): время и пространство — это производные от единого параметра, который мы называем фазой.

В этой концепции релятивистские эффекты — замедление времени, доплеровский сдвиг и т.д. — возникают как естественные проявления движения потоков в фазовом пространстве. Формализм не вводит новые законы физики, он лишь предлагает удобный язык и инструмент для их описания. Подобно лагранжевому и гамильтоновому подходу в классической механике, это попытка упростить структуру уже известных зависимостей и сделать явными причинно-следственные связи.

## 1. Время и пространство как производные
Предположим, что существует переменная величина, назовём её __фаза__ $\vec{\chi} \in \mathbb{C}$, изменение которой порождает наблюдаемые эффекты времени и пространства. Алгебраически это будет означать, что время-пространственные базовые векторы (единицы, меры) являются фазовой скоростью - производной фазы по направлениям в фазовом пространстве:
$$
\hat{h} \, dx_0 = \frac{\partial \vec{\chi}}{\partial \chi_h} \frac{d\chi_h}{d\chi} d\chi
= \tilde{H} \ d\chi, \ 
\mathbf{l} \, dx_l = \frac{\partial \vec{\chi}}{\partial \chi_l} \frac{d\chi_l}{d\chi} d\chi
= \tilde{L} \ d\chi, \,
l = 1,2,3
\qquad (1.1)
$$
Можно сказать, что мы динамизируем тензор пространства-времени через единую общую переменную, тогда, введя фазовую скорость интервала СТО $ds = \tilde{S} \ d\chi$ аналогично ранее представленым прото-скоростям, можем записать закон сохранения интервала в следующем виде:
$$
\tilde{S}^2 = \frac{ds^2}{d\chi^2}
= \frac{g_{ij} dx^i dx^j}{d\chi^2}
= g(\chi)
= \tilde{H}^2 - \tilde{L}^2
\qquad (1.2)
$$
Что равносильно:
$$
\tilde{H}^2 = \tilde{S}^2 + \tilde{L}^2 
\qquad (1.3)
$$
Такая форма предоставляет возможность выразить фазовые скорости $\tilde{S}, \ \tilde{L}$ через $\tilde{H}$:
$$ 
\tilde{S} = \tilde{H} \cos\theta, \ \tilde{L}
= \tilde{H} \sin\theta
\qquad (1.4)
$$
где $\theta$ - угол отклонения фазовой скорости от вещественной оси в фазовом пространстве.  
С математической точки зрения можно сказать, что (1.3) просто утверждает существование параметра, по которому темпоральная компонента интервала может быть разложена как сумма двух проекций на перпендикулярные оси в евклидовом пространстве, то есть если представить, что время-пространство индуцировано изменением единого параметра, то величина скорости по данному параметру вдоль темпоральной координатной оси может быть представлена как сумма скоростей вдоль двух отдельных компонент - пространственной и вне-пространственной (наблюдаемой и ненаблюдаемой, внешней и внутренней).  
Или, упрощая, (1.3) выглядит как формула сложения скоростей в евклидовом векторном пространстве. Предположим, что так и есть, но по какой-то причине при повороте "угла обзора" сохраняется не норма $\tilde{H}$, а проекция на вне-пространственную $\tilde{S}$.  

## 2. Фазовое пространство
Пусть математическое векторное пространство называется [__хора__ (Платон)](https://en.wikipedia.org/wiki/Kh%C3%B4ra?utm_source=chatgpt.com) и имеет свойства обыкновенного комплексного пространства $\mathbb{C}$.  
Геометрически мы определим два базисных вектора назвав их (__время__, __пространство__): $(\hat{h}, \, \mathbf{l}) \equiv \mathbb{C}$; получив в результате представление фазы $\vec{\chi} = R \ e^{\theta \mathbf{l}}, \ \theta \in [-\pi, \pi]$ как вектора в данном базисе со следуюшими отношениями:
$$
\tilde{H} = R, \quad \tilde{S} = R \cos\theta, \quad \tilde{L} = R \ \mathbf{l} \sin\theta \qquad (2.1)
$$
и (1.3) становится __законом сохранения__ полной фазы объекта при повороте вектора фазы.  
При этом мы задали фазовое пространство так, что частные производные равны единице, поэтому (1.1) можно переписать так:
$$
\hat{h} \, dx_0 = \frac{d\chi_h}{d\chi} d\chi = \tilde{H} \ d\chi, \ \mathbf{l} \, dx_l = \frac{d\chi_l}{d\chi} d\chi = \tilde{L} \ d\chi, \, l = 1,2,3 \qquad (2.2)
$$
Отметим, что вместе с тем, фазовое пространство, скорости которого являются базисами для наблюдаемого пространства-времени, связано с последним интегральным преобразованием вида:
$$
x^i = \int{ \tilde{X}^i(\chi) \ d\chi}, \quad i = 0,1,2,3 \qquad (2.3)
$$
где $\tilde{X}^i(\chi)$ — фазовые скорости, являющиеся проекциями производной фазы $\vec{\chi}$ на соответствующие направления базиса $(\hat{h}, \mathbf{l})$.  
Таким образом, привычные координаты времени и пространства представляют собой интегральные образы фазовых скоростей, а сами фазовые скорости — это «мост» между скрытым фазовым параметром и наблюдаемым многообразием Минковского.  
Ввиду отношений (2.2) и (2.3) фазовое пространство $\mathbb{C}$ выступает как порождающее пространство, а наблюдаемое пространство-время — как его проекционное отображение.  

## 3. Объекты
Любая __фундаментальная частица__ является __элементарным объектом__ униметрии и характеризуется ненулевой величиной фазы ($\vec{\chi} \ne 0$).  
Составные __объекты__ являются фазовыми конфигурациями, для полного представления которых в фазовом пространстве в основном требуется его расширение дополнительными измерениями, кроме случая __фотона__, фаза которого всегда направлена вдоль мнимой оси хоры:
$$
\mathbf{p} = \frac{d\vec{\chi}}{d\chi_l} = p \mathbf{l} \in \Im \qquad (3.1)
$$
Выделим все явления, отличные от перемещения фотонов, в отдельную категорию, аналогично СТО получив фотонное подпространство, которое в униметрии представлено мнимой осью, а остальные, более __сложные__ в этом контексте объекты, характеризуются наличием вещественной проекции фазы и ненулевой массой.
Тогда проводя параллель с СТО, сложный объект униметрии ассоциирован с любым __событием__ или чередой таковых - с мировой линией, а фотон - с точкой подпространства нулевого интервала, содержащей информацию о __событии__.  
Фаза любого объекта может быть приведена поворотом в хоре в _нулевое_ направление (положение нулевого отклонения вектора фазы) вдоль вещественной оси:
$$
\vec{\chi}_o = R \in \Re \qquad (3.2)
$$
Объект A, движущийся относительно покоящегося наблюдателя, будет представлен вектором фазы находящейся под углом к вектору фазы наблюдателя:
$$
\vec{\chi}_a = R \ e^{(\theta_A \, \mathbf{l})}, \, \sin \theta_a
= \frac{V}{\mathtt{c}}
\equiv \beta
\qquad (3.3)
$$
таким, что синус угла поворота его вектора фазы равен отношению взаимной скорости объектов к скорости света: $\theta > 0$ - отдаление, $\theta < 0$ - сближение, для перпендикулярного движения знак не важен (фактор квадратичен в уравнении).  
Таким образом, множество относительно неподвижных объектов формируют друг для друга фотонное пространство нулевой кривизны. Такие объекты будем называть _синфазными_.  

### 3.1. Пространство
Из определения (1.4) для пространственной составляющей следует, что в нулевом направлении компонента скорости фазы по мнимой оси хоры равна нулю как и базисный вектор пространства объекта, что противоречит наблюдаемой действительности - как минимум для обеспечения собственной проекции в наблюдаемое пространство у объекта должна существовать проекция фазы на мнимую ось, поэтому примем такую формулировку вырожденной и применимой лишь для малых объектов.  
Формализуем это требование следующим образом. Сложим два вектора фазы с равными пространственными отклонениями, но с разным знаком, и разделим результат на два, чтобы компенсировать норму. Получим:
$$
\vec{\chi}^\pm = R \, e^{\pm \zeta \ \mathbf{l}} \quad \chi_l = \frac{\vec{\chi}^+-\vec{\chi}^-}{2} = R \, \mathbf{l} \sin\zeta   \qquad (3.1.1)
$$
где $\vec{\chi}_l$ выступает в роли пространственного вектора фазы, а $\zeta$ - угол локального восприятия: не зависит от скорости в пространстве, является внутренним аттрибутом объекта (функционально связанным с массой и плотностью).

Таким образом, полная фаза объекта в _локальном_ направлении представляется как пара:
$$
\vec{\chi}_o = \vec{\chi}_\tau + \vec{\chi}_l = R \ \cos\zeta + R \ \mathbf{l} \sin\zeta \qquad (3.1.2)
$$
что эквивалентно разложению на темпоральную и пространственную компоненты.
При этом:

$\vec{\chi}_\tau = R \cos|\zeta|$ — временной базис,

$\vec{\chi}_l = R \, \mathbf{l} \, \sin|\zeta|$ — пространственный базис.

Если нормировать их на полный радиус $R$, то получаем ортогональные единичные базисы для фазового пространства:
$$
\hat{h} = \cos\zeta, \, \hat{\mathbf{l}} = \sin \zeta \qquad (3.1.3)
$$
Тогда __для единой нормы__ $\tilde{H}$ скорости фазы возникает три направления и соответствующих им вещественных значения, которые мы хотели бы обозначить.  

### 3.2 Абсолютное, локальное и наблюдаемое время  
Нулевому положению фазы соответствует **абсолютное** время $t = t(\tilde{H})$ - самое медленное. Из §3.1 следует, что не существует объекта, который бы функционировал по этому времени.  
Однако оно может быть использовано, например, для нормирования при переходе между обстоятельствами фаз разных скоростей.  
В §1.1 мы ввели базисный вектор времени как производную фазы объекта по вещественной компоненте.  
В локальном направлении вектор фазы всегда даёт его максимально возможное для объекта значение - __локальное__ время:  
$$
dx_0 = \frac{d}{d\chi}\,\Re(\vec{\chi}) \, d\chi 
= \frac{\vec{\chi}^+ + \vec{\chi}^-}{2}\, d\chi 
= \cos \zeta \, d\chi 
= d\tau 
\qquad (3.2.1)
$$
Здесь возникает **двойная аналогия углов**:  

- $\zeta$ — *внутренний угол объекта*  
  (характеризует его массу, собственную "плотность" потока).  
- $\theta$ — *внешний угол движения*  
  (характеризует скорость относительно наблюдателя).  

Таким образом, замедление времени у объекта имеет **двойную структуру**:  

- ограничено **внутренним состоянием** (через множитель $\cos\zeta$),  
- и дополнительно — **внешним движением** (через множитель $\cos\theta$).  

**Ключевой вывод:**  
наблюдаемое время всегда является результатом *двух наложенных преобразований* - внутреннего (природа объекта) и внешнего (его движение).  

Объект А, перемещающийся относительно наблюдателя, будет иметь _собственное_ - __наблюдаемое__ - время относительно покоящегося наблюдателя в виде вещественной части производной собственного вектора фазы по вектору фазы наблюдателя:
$$
\tilde{H}_a = \Re \left( \frac{d\vec{\chi}_a}{d\vec{\chi}_o} \right)
= \frac{\cos\theta_a \ d\chi_o}{d\chi_o}
= \cos\theta_a
= \sqrt{1-\sin^2\theta_a}
= \sqrt{1-\frac{V^2}{\mathtt{c}^2}}
= \frac{1}{\gamma} 
\qquad (3.2.2)
$$
что находится в согласовании с (3.3), так как является следствием сохранения фазы, а также соответствует экспериментально известному преобразованию времени и симметрично для восприятия объектом А перемещения наблюдателя за счёт геометрического представления - в собственной системе координат объекта А проекция фазы наблюдателя на вещественную ось тоже $\cos\theta_a$.  

### 3.3. Нормирование
Пусть локальное течение времени сложного объекта параметризовано **фазой**: вводим опорную частоту $\nu_0$ и связь
$$
d\tau=\frac{1}{\nu_0}\,d\chi_o.
$$
По цепному правилу:
$$
dx_0=\tilde{H}\,d\chi=\frac{dx_0}{d\chi_o}\,\frac{d\chi_o}{d\tau}\,d\tau
=\tilde{H}\,\dot{\chi}\,d\tau=\dot{H}\,d\tau,
\qquad (3.1.1)
$$
где введены обозначения
$$
\nu:=\frac{d\chi}{d\tau},\qquad \dot{\chi}:=\frac{\nu}{\nu_0},\qquad \dot{H}:=\tilde{H}\,\dot{\chi}.
$$
Отсюда видно, что мы принимаем скорость течения по локальному времени калибровкой: при нормировке $\dot{H}\equiv\mathtt{c}$ имеем $dx_0=\mathtt{c}\,d\tau$.

Аналогично для пространственной части:
$$
dx_l=\tilde{L}\,d\chi
=\frac{dx_l}{d\chi_o}\,\frac{d\chi_o}{dl}\,dl
=\tilde{L}\,\chi'\,dl
= L'\,dl,\quad l=1,2,3,
\qquad (3.1.2)
$$
где $\chi':=d\chi/dl$, $L':=\tilde{L}\,\chi'$. Из равенства (1.4) для света $dx_0=dx_l$ следует
$$
\mathtt{c}=\tilde{L}'\,\frac{dl}{d\tau},
$$
и при темпоральной нормировке на скорость света пространственный масштаб становится единичным: $\tilde{L}'=1$.  

### 3.4. Свет
Продолжим с (1.4) выразив обе координаты через фазу и нормирующие факторы:
$$
\frac{\mathtt{c}}{ \dot{\chi}} d\chi = \frac{1}{ \chi' } d\chi
\qquad (3.4.1)
$$
Получается:
$$
\mathtt{c} = \frac{\dot{\chi}}{ \chi' }
= \frac{dl}{ d\tau } \ne f(\chi)
\qquad (3.4.2)
$$
Откуда видно, что в нашей модели скорость света является параметром, связывающим движения по осям __вне зависимости от локального изменения фазы__.  
Во-первых, если посмотреть на (3.4.2) следующим образом:
$$
\mathtt{c} = \frac{\dot{\chi}}{ \chi' }
= \left(\frac{d\chi}{d\tau} \right) \left[ \frac{dl}{ d\chi } \right] \sim (\nu) \ [\lambda]
\qquad (3.4.3)
$$
То можно заметить сходство с характеристиками фотона _частота_ и _длина волны_, произведение которых даёт скорость света, где $\chi$ действительно соответствует фазе фотона.  
Во-вторых, из (3.4.1) и (3.4.2) следует:
$$
ds^2 = \mathtt{c}^2 \left( \frac{d\chi^2}{ \dot{\chi}^2}  - \frac{d\chi^2}{ \dot{\chi}^2} \right) = 0, \forall \dot{\chi}
\qquad (3.4.4)
$$
где решением для единичной частоты будет $\tau = \chi$ - собственное время фотона это его фаза. Интересно, что в такой нормировке длина вектора фазовой скорости фотона численно равна его длине волны $\tilde{H}_p = \lambda$.  
В-третьих, если и время, и пространство заданы единым параметром, то сохранение скорости света при преобразованиях - естественное следствие такой параметризации.  
Если скорость равна наклону траектории на диаграмме «время–пространство» в фазовых координатах:
$$
\frac{dx_l}{dx_0} 
= \frac{\tilde{L}\,d\chi}{\tilde{H}\,d\chi}
= \sin \theta
= \frac{V}{\mathtt{c}}
\equiv \beta
\qquad (3.4.5)
$$
Для фотона $\theta = \pi/2$, и потому:
$$
\frac{dx_l}{dx_0} = 1 \quad \Rightarrow \quad V = c.  
\qquad (3.4.6)
$$
Скорость света в униметрии возникает как отношение, задающее пространственный базис в зависимости от нормы $\tilde{H}$ скорости потока фазы.

### 3.5. Лоренц-фактор
Мы знаем, фактически, что в результате преобразования направления фазовой скорости вида:
$$
\tilde{H}^2 = \tilde{S}^2 + \tilde{L}^2 \overset{\cdot \dot{\chi}^2}{\longmapsto} \dot{H}^2 = \dot{S}^2 + \dot{L}^2 
\qquad(3.5.1)
$$
должен происходить переход нормы от $\tilde{H}$ к норме $\dot{S}$.  
_Лемма_: переход $\tilde{H}\rightarrow\dot{S}$ является проявлением эволюции фазовой скорости объекта по собственному времени $\dot{H}(\tau(\chi))$.
То есть перенормировка происходит в каждый момент приобретения дополнительной скорости объектом, в то время как величина $\tilde{S}$ является математической и проективной, когда в реальности объект имеет только величину своей скорости фазы $\tilde{H}$, восприятие которой меняется ввиду изменения скорости течения локального времени, то есть физически мотивированным было бы переписать (3.5.1) как:
$$
\dot{H}^2 = \tilde{H}^2 + \dot{L}^2
\qquad(3.5.2)
$$
которое читается так: фазовая скорость движущегося объекта $\tilde{H}$ в обстоятельствах наблюдателя, обусловленных его фазовой скоростью $(\dot{H},\dot{L})$, сохраняется.  
Докажем лемму, описав мгновенную перенормировку, соответствующую повороту фазовой скорости - как замену переменной $\chi \mapsto \tau(\chi)$ с локальным Якобианом:
$$
\frac{d\tau}{d\chi}=\cos\zeta(\chi)\,\cos\theta(\chi)
\quad\Longrightarrow\quad
\mathcal{J}(\zeta,\theta):=\frac{d\chi}{d\tau}=\frac{1}{\cos\zeta\,\cos\theta}.
\qquad(3.5.3)
$$

В результате компоненты фазовой скорости мгновенно масштабируются:
$$
\dot{H}
=\frac{dx_0}{d\tau}
=\frac{dx_0}{d\chi}\frac{d\chi}{d\tau}
=\tilde{H}\,\mathcal{J}
=\frac{\tilde{H}}{\cos\zeta\,\cos\theta},
\qquad
\dot{L}
=\tilde{L}\,\mathcal{J}
=\frac{\tilde{L}}{\cos\zeta\,\cos\theta}.
\qquad(3.5.4)
$$

В инфинитезимальной форме:
$$
d\ln\dot{H}=d\ln\mathcal{J}
=\tan\zeta\,d\zeta+\tan\theta\,d\theta
\;\Rightarrow\;
d\dot{H}=\dot{H}\,(\tan\zeta\,d\zeta+\tan\theta\,d\theta).
\qquad(3.5.5)
$$

- **Чистый буст** ($d\zeta=0$, $d\tilde{H}=0$):
  $$
  d\dot{H}=\dot{H}\,\tan\theta\,d\theta.
  $$
- **Чистое внутреннее изменение** ($d\theta=0$):
  $$
  d\dot{H}=\dot{H}\,\tan\zeta\,d\zeta.
  $$

> Если в нормировке §3.3 фиксировать $\dot{H}\equiv \mathtt{c}$, то (3.5.4) означает, что «изменение» достигается умножением на текущий Якобиан $\mathcal{J}$: физического скачка нет, есть локальная смена параметра.

Таким образом, (3.5.2) с учётом перепараметризации при чистом бусте $d\zeta=0$ (здесь и далее в статье внутренние изменения не рассматриваются) даёт псевдо-евклидову форму интервала в пространстве Минковского:
$$
\tilde{H}^2 = \dot{H}^2 - \dot{L}^2
= \sec^2\theta \ (\tilde{H}^2 - \tilde{L}^2)
= \gamma^2 \ (\tilde{H}^2 - \tilde{L}^2)
\qquad(3.5.6)
$$

**Вывод.**  
Переход нормы следует из инфинитезимального поворота фазы и пересчёта меры времени.  
- В хоре сохраняется полная скорость $\tilde{H}$ (евклидова норма).  
- В наблюдаемом времени сохраняется $\dot{S}$ (минковская норма).  
- Эти величины идентичны:
$$
\boxed{\;\tilde{H}=\dot{S}\;}
\qquad(3.5.7)
$$
так что наблюдаемое течение времени есть переинтерпретация полной скорости фазы в собственной системе объекта.

### 3.6. Быстрота и угол фазового поворота

По определению в униметрии скорость выражается через угол фазового поворота:
$$
\beta \equiv \frac{V}{\mathtt{c}} = \sin\theta.
\qquad(3.6.1)
$$

Быстрота задаётся соотношением $\tanh\eta=\beta$, откуда
$$
d\eta=\frac{d\beta}{1-\beta^2}.
\qquad(3.6.2)
$$

Дифференцируя $\beta=\sin\theta$ и подставляя в (3.6.2), получаем
$$
d\beta=\cos\theta\,d\theta,\qquad 1-\beta^2=\cos^2\theta
\;\;\Rightarrow\;\;
d\eta=\sec\theta\,d\theta.
\qquad(3.6.3)
$$

Интегрирование даёт
$$
\eta(\theta)=\int \sec\theta\,d\theta
=\ln\!\bigl|\sec\theta+\tan\theta\bigr|
=\tfrac12 \ln\!\left|\frac{1+\sin\theta}{1-\sin\theta}\right|.
\qquad(3.6.4)
$$

Эквивалентно (фиксируя $\eta(0)=0$):
$$
e^{\eta(\theta)}=\sqrt{\frac{1+\sin\theta}{1-\sin\theta}}.
\qquad(3.6.5)
$$

Подставляя $\sin\theta=\beta$, получаем продольный релятивистский доплер-фактор:
$$
e^{\eta}=\sqrt{\frac{1+\beta}{1-\beta}} \;\equiv\; r.
\qquad(3.6.6)
$$

Обратная зависимость угла от быстроты:
$$
\theta=\arcsin(\tanh\eta)
=\arcsin\!\left(\frac{e^{2\eta}-1}{e^{2\eta}+1}\right).
\qquad(3.6.7)
$$

Лоренц-фактор в трёх эквивалентных формах:
$$
\gamma=\frac{1}{\sqrt{1-\beta^2}}
=\frac{1}{\sqrt{1-\sin^2\theta}}
=\sec\theta
=\cosh\eta.
\qquad(3.6.8)
$$

Связь с перенормировкой темпоральной компоненты (чистый буст):
$$
d\ln \dot{H}=\tan\theta\,d\theta
=\sin\theta\,d\eta.
\qquad(3.6.9)
$$

### 3.7 Сложение скоростей

Быстрота аддитивна:
$$
\eta_{12}=\eta_1+\eta_2 
\qquad(3.7.1)
$$

Так как $\tanh\eta=\beta$ , имеем релятивистский закон сложения скоростей:
$$
\beta_{12}
=\tanh(\eta_1+\eta_2)
=\frac{\tanh\eta_1+\tanh\eta_2}{1+\tanh\eta_1\,\tanh\eta_2}
=\frac{\beta_1+\beta_2}{1+\beta_1\beta_2}.
\qquad (3.7.2)
$$

Через фазовый угол (напоминание: $\beta=\sin\theta$):
$$
\gamma=\frac{1}{\sqrt{1-\beta^2}}=\sec\theta,
\qquad
\gamma\beta=\tanh\eta=\sin\theta\,\sec\theta=\tan\theta.
\qquad(3.7.3)
$$

Отсюда, используя (3.7.1), удобно вычислять композицию как
$$
\gamma_{12}=\cosh(\eta_1+\eta_2)=\gamma_1\gamma_2(1+\beta_1\beta_2),
\quad
\gamma_{12}\beta_{12}=\sinh(\eta_1+\eta_2)=\gamma_1\gamma_2(\beta_1+\beta_2).
\qquad(3.7.4)
$$

### 3.8. Доплеровский сдвиг

Определим наблюдаемую частоту как скорость нарастания фазы по собственному времени наблюдателя:
$$
\nu := \frac{d\chi}{d\tau}
\qquad (3.8.1)
$$

Для двух последовательных фронтов волны приращение фазы одинаково, поэтому
$$
\frac{\nu_{\text{obs}}}{\nu_{\text{src}}}
=\frac{d\chi/d\tau_{\text{obs}}}{d\chi/d\tau_{\text{src}}}
=\frac{d\tau_{\text{src}}}{d\tau_{\text{obs}}}
\qquad (3.8.2)
$$

**Продольный случай.**  
Пусть источник движется со скоростью $V$ вдоль прямой к наблюдателю (подход) или от него (удаление). Интервал между эмиссиями в системе наблюдателя равен $\gamma\, d\tau_{\text{src}}$, а источник за это время смещается на $\pm V\,\gamma\, d\tau_{\text{src}}$ (знак «$+$» — удаление, «$-$» — подход). Тогда
$$
d\tau_{\text{obs}}
=\gamma\, d\tau_{\text{src}}\,(1\pm \beta),
\qquad
\beta=\frac{V}{\mathtt{c}}
\qquad (3.8.3)
$$
Отсюда продольный доплеровский фактор:
$$
\boxed{\;
\frac{\nu_{\text{obs}}}{\nu_{\text{src}}}
=\frac{1}{\gamma(1\pm \beta)}
\;}
\quad
\begin{cases}
\text{подход: } (-)\\[2pt]
\text{удаление: } (+)
\end{cases}
\qquad (3.8.4)
$$

Эквивалентные формы (с $\beta=\sin\theta,\ \gamma=\sec\theta$ и быстротой $\eta$):
$$
\frac{\nu_{\text{obs}}}{\nu_{\text{src}}}
=\sqrt{\frac{1\mp \beta}{1\pm \beta}}
=\sec\theta\, (1\mp \sin\theta)
=e^{\mp \eta}
\qquad (3.8.5)
$$

- **Подход (синее смещение):**
$$
\frac{\nu_{\text{obs}}}{\nu_{\text{src}}}
=\frac{1}{\gamma(1-\beta)}
=\sqrt{\frac{1+\beta}{1-\beta}}
=\sec\theta\,(1+\sin\theta)
=e^{+\eta}
\qquad (3.8.6)
$$

- **Удаление (красное смещение):**
$$
\frac{\nu_{\text{obs}}}{\nu_{\text{src}}}
=\frac{1}{\gamma(1+\beta)}
=\sqrt{\frac{1-\beta}{1+\beta}}
=\sec\theta\,(1-\sin\theta)
=e^{-\eta}
\qquad (3.8.7)
$$

Соответственно, отношение длин волн — обратное:
$$
\frac{\lambda_{\text{obs}}}{\lambda_{\text{src}}}
=\frac{1}{r},
\quad
r:=\frac{\nu_{\text{obs}}}{\nu_{\text{src}}}
=e^{\pm\eta}
\qquad (3.8.8)
$$

**Поперечный доплер.**  
При $\varphi=90^\circ$ (направление на источник перпендикулярно скорости в системе наблюдателя):
$$
\frac{\nu_{\text{obs}}}{\nu_{\text{src}}}
=\frac{1}{\gamma}
=\cos\theta
\qquad (3.8.9)
$$

**Общий угловой случай.**  
Если $\varphi$ — угол между вектором скорости источника и направлением на наблюдателя **в системе наблюдателя**, то
$$
\boxed{\;
\frac{\nu_{\text{obs}}}{\nu_{\text{src}}}
=\gamma\,(1-\beta\cos\varphi)
\;}
\qquad (3.8.10)
$$

> В терминах фазы (§3.4) доплеровский сдвиг — это пересчёт меры времени между источником и наблюдателем при неизменности счёта фазовых циклов: сдвигается не «сама волна», а масштаб параметра, по которому измеряется её фаза.

## 4. Заключение

Фазовый формализм униметрии показывает, что релятивистские эффекты естественно следуют из евклидовой геометрии **фазового пространства**, где время и пространство — это **производные проекции** единого параметра (фазы).

Главные итоги:

- **Интервал Минковского** возникает как сохранение нормы фазовой скорости при смене параметра (повороте фазы).
- **Замедление времени** и **сокращение длины** — геометрические следствия двойной структуры углов: внутреннего $\zeta$ (связанного с природой объекта) и внешнего $\theta$ (связанного с относительным движением).
- **Скорость света** выступает универсальной константой, связывающей темпоральное и пространственное масштабирование независимо от деталей фазовой динамики.
- **Лоренц-фактор** и **закон сложения скоростей** выводятся напрямую из геометрии поворотов на комплексной окружности, без явного введения гиперболической метрики.
- **Доплеровский сдвиг** неизбежно возникает как *масштабирование меры времени* между источником и наблюдателем при неизменности счёта фазовых циклов:  
  продольный фактор $r=\nu_{\text{obs}}/\nu_{\text{src}}=\frac{1}{\gamma(1\pm\beta)}=\sqrt{\frac{1\pm\beta}{1\mp\beta}}=e^{\pm\eta}$,  
  поперечный — $r=1/\gamma$.

Таким образом, униметрия сохраняет все проверенные экспериментом следствия СТО, но предлагает **единый фазовый язык**, в котором:
1) релятивистские эффекты — это проекции одной и той же фазовой динамики;
2) внутренние свойства объектов (масса, «плотность» потока) входят через угол $\zeta$;
3) время и пространство понимаются как интегральные образы фазовых скоростей, что создаёт мост к описанию областей, где «метрика» не задана априори.

**Перспективы.** Связать фазовую картину с квантованием (дискретизация фазы), обобщить к кривизне (ОТО) и исследовать коллективные фазовые конфигурации (поля, взаимодействия, когерентность), используя доплеровские соотношения как операциональные связки между наблюдаемыми частотой, длиной волны и фазовой быстротой.

## 5. Связь с известными теориями

В этом разделе показано, как фазовый формализм униметрии стыкуется с квантовой теорией, калибровочными полями, ОТО и космологией. Во всех случаях остаёмся в рамке: время и пространство — интегральные образы фазовой динамики (§§1–3), а релятивистские эффекты — геометрия поворота фазы.

### 5.1. Квантование как дискретизация фазы
Гипотеза: элементарная «ячейка» возникает как условие периодичности
$$
\Delta\chi=2\pi n,\quad n\in\mathbb{Z}.
$$
Связь с квантовой фазой достигается идентификацией
$$
\chi\;\hat{=}\;\frac{S}{\hbar}.
$$
Тогда интерференция — интерференция **фазовых траекторий** в хоре, а амплитуды — веса таких траекторий. Из §3.4 (частота $\nu=d\chi/d\tau$) следует
$$
E=\hbar\,\frac{d\chi}{d\tau},\qquad p_l=\hbar\,\frac{d\chi_l}{dl},
$$
что согласуется с де Бройлем при $E=\hbar\omega$, $p=\hbar k$. Квантование — ограничения на допустимые **замкнутые фазы** (условия типа Бора–Зоммерфельда) в конфигурациях с компактной $\chi$.

### 5.2. Калибровочные симметрии как перепараметризации $\chi$
Глобальный сдвиг $\chi\mapsto\chi+\chi_0$ ненаблюдаем. Разрешая **локальные** перепараметризации $\chi\mapsto\chi+\alpha(x)$, требуем когерентности сравнения фаз в разных точках — это порождает связность (калибровочное поле). На уровне волновой функции $\psi\sim e^{i\chi}$:
$$
\psi(x)\to e^{i\alpha(x)}\psi(x),\qquad \partial_\mu\to D_\mu=\partial_\mu-iA_\mu.
$$
Здесь $A_\mu$ — геометрическая «связка» между фазовыми слоями; минимальное взаимодействие возникает как требование согласованности фазового переноса.

### 5.3. Масса и внутренний угол $\zeta$
В §3.1 введён внутренний угол $\zeta$, задающий разложение фазы. Гипотеза: **масса** — инвариант, связанный с неустранимой вещественной проекцией $\cos\zeta\neq0$.
- безмассовые: $\zeta=\pm\frac{\pi}{2}$ (локальное время «заменяется» пространственным бегом; фотонное подпространство);
- массивные: $|\zeta|<\frac{\pi}{2}$, собственное время существует; внешнее замедление через $\gamma=\sec\theta$ наслаивается на внутренний множитель $\cos\zeta$ (§3.2).
Это открывает путь к фазовой трактовке массогенеза (сопоставление $\zeta$ механизмам типа Хиггса).

### 5.4. ОТО как неоднородная перепараметризация $\chi$
В §1.2 интервал задан через $g(\chi)$. При неоднородном распределении фазы Якобиан перепараметризации (§3.5) становится полем:
$$
\mathcal{J}(x)=\frac{d\chi}{d\tau}(x),\qquad dx^\mu=\tilde{X}^\mu(\chi)\,d\chi,
$$
и метрика наблюдаемого многообразия приобретает кривизну из-за вариаций $\mathcal{J}$ и $\tilde{X}^\mu$. Геодезические — траектории экстремального «фазового времени» (стационарной $\chi$). Гравитационное красное смещение трактуется как доплероподобное масштабирование меры времени в неоднородном поле $\mathcal{J}(x)$.

### 5.5. Космология и «абсолютный» калибр
Калибр «абсолютного времени» фиксируем космологически (см. Appendix): **комовинговая система** с нулевым диполем CMB. Тогда космологическое время FLRW задаёт глобальный калибр $t$; космологический красный сдвиг — накопленный фазовый (доплеровский) фактор на расширяющемся фоне; локально сохраняется лоренц-инвариантность (§§3.6–3.8).

### 5.6. Экспериментальные опоры
Фазовая интерпретация воспроизводит:
- продольный/поперечный **доплер** (§3.8);
- замедление времени (мионы, атомные часы, GPS);
- релятивистское сложение скоростей (§3.7).
Дальнейшие тесты: высокоточная спектроскопия при контролируемых $\theta$ и, по возможности, с управляемой эффективной «плотностью» (углом $\zeta$) в конденсированных системах/оптических решётках.

**Итог.** Униметрия не меняет законов, а переупорядочивает их в едином фазовом языке. В нём доплеровский сдвиг, лоренц-фактор, калибровочная связь и (эвристически) гравитация оказываются аспектами одной и той же геометрии перепараметризации $\chi$.

## Физическая мотивация и онтология фазового вектора χ

Идея проста: **фаза — это то, что реально можно посчитать тиками собственного осциллятора** (атомные часы, «де Бройлевские» часы и т.п.). Мы рассматриваем **скорость изменения этой фазы** как геометрический объект и договариваемся, как из него извлекаются наблюдаемые «время» и «пространство».

### Операциональные постулаты

**P1 (локальный отсчёт).** У любого объекта есть монотонный локальный счётчик фазы (тикер). В пределе идеальной стабильности его приращение пропорционально собственному времени: \(d\sigma \propto d\tau\). Мы будем использовать \(\sigma\) как «фазовую дугу».

**P2 (фазовый поток как вектор).** Изменение фазы задаётся **вектором фазового потока** \(\boldsymbol\chi\) в евклидовом прото-пространстве \(\mathcal E\). Наблюдатель выбирает ортонормированную тройку \((\mathbf e_1,\mathbf e_2,\mathbf e_3)\) и тем самым задаёт наблюдаемое 3-пространство \(S\subset\mathcal E\) и его ортогональное дополнение \(S^\perp\).

**P3 (как получить наблюдаемые).** Проецируем \(\boldsymbol\chi\) на \(S\) и \(S^\perp\):
\[
\mathbf l:=P_S\boldsymbol\chi,\qquad
t:=\|P_{S^\perp}\boldsymbol\chi\|.
\]
Дальше упаковываем это в **реальный кватернион**
\[
q \;=\; t+\mathbf l\ \in\ \mathbb H,\qquad |q|^2=t^2+\|\mathbf l\|^2=\|\boldsymbol\chi\|^2.
\]
Кинематика — это **роторы** (кватернионные повороты) потока \(\boldsymbol\chi\) относительно тетрады наблюдателя.

### Почему это не «пустая переупаковка»

1) **Физичность χ.** \(\boldsymbol\chi\) — не «мистическая сущность», а компактный носитель того, что реально измеряется двумя приборами:
   - скаляр \(t\): число тиков собственного осциллятора (по сути — \(d\tau\) после нормировки);
   - вектор \(\mathbf l\): то, как **ориентация наблюдателя** «снимает» пространственные компоненты фазового потока (IMU/гиротетрада задаёт оси).

2) **Евклидовость как удобный калькулятор.** Работать в \(\mathcal E\) удобно: буст — это обычный 2D-поворот в плоскости \((\mathbf e_t,\hat{\mathbf u})\):
\[
d=\cos\frac{\psi}{2}+\hat{\mathbf u}\sin\frac{\psi}{2},\qquad q\mapsto dqd,
\]
а **гипербола СТО** появляется простой заменой параметра
\[
\sin\psi\leftrightarrow\tanh\eta,\quad \cos\psi\leftrightarrow\operatorname{sech}\eta,
\]
т.е. это не новая физика, а **другая координатизация** тех же инвариантов.

3) **Добавленная ценность — алгоритм.** В этой упаковке извлечение «поворота Вигнера» становится буквально одной строкой
\[
d_{12}=\sqrt{L_{12}L_{21}},\quad r_W=\bar d_{12}\,L_{12},
\]
что даёт простой, стабильный и быстрый код (и мы показываем численный контроль). Это не претензия на новый эффект — это **рабочий рецепт**, который хорошо живёт в вычислениях.

### Почему кватернионы, а не «просто рапидность»

- **Рапидность \(\eta\)** — скаляр 1D-бустов. В 3D композиция бустов порождает **поворот** (Вигнера). Кватернионный **ротор** держит и ось, и угол в одном объекте; композиция — это просто умножение.
- **Единый объект для R и D.** Пространственный поворот \(r\) и «буст» \(d\) — один и тот же тип единичного кватерниона (меняется плоскость вращения). Это снимает разрыв «гипербола vs сфера» и делает код однородным.
- **Численная устойчивость.** Нет гимбал-локов, нормализация — одно деление, выбор главной ветви корня для \(d_{12}\) — очевиден (\(\Re d_{12}\ge0\)).

### Что с «khōra» и «комплексным срезом»

- Термин *khōra* — **метафора** для интуиции «носителя фазы»; его легко убрать/перенести в примечание, чтобы не создавать философского тумана.
- **Комплексный срез не обязателен.** В этой работе всё делается **реальными кватернионами**. «Гиперболический» вид появляется **только** как перепараметризация \(\psi\to\eta\); он не требует комплексирования.

### Мини-связь с наблюдаемыми

Выбирая «часы» эволюции, получаем привычные формулы:
\[
dt=\gamma\,d\tau,\quad \beta=\sin\psi,\quad \gamma=\frac1{\cos\psi}.
\]
Дальше следуют стандартные эффекты (дилатация, доплер, Томас/Вигнер). То есть «онтология» здесь строго операциональна: **тик-счётчик + тетрада** \(\Rightarrow\) \(q=t+\mathbf l\) \(\Rightarrow\) роторная кинематика.


```python
import numpy as np

# ----- базовая кватернионная алгебра -----

def qmul(a, b):
    w1, x1, y1, z1 = a
    w2, x2, y2, z2 = b
    return np.array([
        w1*w2 - x1*x2 - y1*y2 - z1*z2,
        w1*x2 + x1*w2 + y1*z2 - z1*y2,
        w1*y2 - x1*z2 + y1*w2 + z1*x2,
        w1*z2 + x1*y2 - y1*x2 + z1*w2,
    ])

def qconj(q):
    w, x, y, z = q
    return np.array([w, -x, -y, -z])

def qnorm(q):
    return np.linalg.norm(q)

def qnormalize(q, eps=1e-15):
    n = qnorm(q)
    return q if n < eps else q/n

def qsqrt_unit(q, eps=1e-15):
    """
    Главная ветвь sqrt для единичного кватерниона q=[w,v], Re>=0.
    """
    q = qnormalize(q)
    w, vx, vy, vz = q
    v = np.array([vx, vy, vz])
    vnorm = np.linalg.norm(v)
    if w >= 0:
        s = np.sqrt(0.5*(1.0 + w))
        vec = v/(2.0*s) if s > eps else (v/(vnorm+eps))
        return qnormalize(np.array([s, *vec]))
    else:
        # w ~ отриц., корень почти чисто-векторный
        axis = v/(vnorm+eps) if vnorm > eps else np.array([1.0,0.0,0.0])
        return np.array([0.0, *axis])

def raxis_angle(r, eps=1e-15):
    r = qnormalize(r)
    w, x, y, z = r
    v = np.array([x, y, z])
    vnorm = np.linalg.norm(v)
    angle = 2.0*np.arctan2(vnorm, w)
    axis = np.array([1.0,0.0,0.0]) if vnorm < eps else v/vnorm
    return axis, angle

# ----- параметризация D-ротора -----

def psi_from_beta(beta):
    """
    Униметрический угол psi из β:  β=sinψ,  γ=1/cosψ.
    Численно устойчиво через полуугол: tan(ψ/2) = γβ/(γ+1).
    """
    beta = float(beta)
    if not (0 <= beta < 1):
        raise ValueError("beta must be in [0,1)")
    gamma = 1.0/np.sqrt(1.0 - beta*beta)
    t = (gamma*beta)/(gamma+1.0)
    return 2.0*np.arctan(t)

def drotor_from_velocity(v, c=1.0):
    v = np.array(v, dtype=float)
    speed = np.linalg.norm(v)
    if speed == 0:
        return np.array([1.0, 0.0, 0.0, 0.0])
    u = v/speed
    beta = speed/float(c)
    psi = psi_from_beta(beta)
    c2, s2 = np.cos(psi/2.0), np.sin(psi/2.0)
    return np.array([c2, *(s2*u)])

# ----- Wigner из двух скоростей по униметрии -----

def wigner_rotor_from_velocities(v1, v2, c=1.0):
    d1 = drotor_from_velocity(v1, c)
    d2 = drotor_from_velocity(v2, c)
    L12 = qmul(d2, d1)
    L21 = qmul(d1, d2)
    # из условия d12 * e_t * d12 = L12 * e_t * L21, где e_t ≡ 1, получаем d12^2 = L12 L21
    d12 = qsqrt_unit(qmul(L12, L21))
    rW  = qmul(qconj(d12), L12)
    return qnormalize(rW)

# ----- Классическая формула угла -----

def wigner_angle_classic(v1, v2, c=1.0):
    v1 = np.array(v1, float); v2 = np.array(v2, float)
    u1 = v1/np.linalg.norm(v1) if np.linalg.norm(v1) > 0 else np.array([1.0,0.0,0.0])
    u2 = v2/np.linalg.norm(v2) if np.linalg.norm(v2) > 0 else np.array([1.0,0.0,0.0])
    cos_th = float(np.clip(np.dot(u2, u1), -1.0, 1.0))
    sin_th = np.sqrt(1.0 - cos_th*cos_th)

    beta1 = np.linalg.norm(v1)/float(c)
    beta2 = np.linalg.norm(v2)/float(c)
    eta1  = np.arctanh(beta1)  # rapidities
    eta2  = np.arctanh(beta2)

    sh1 = np.sinh(eta1/2.0); ch1 = np.cosh(eta1/2.0)
    sh2 = np.sinh(eta2/2.0); ch2 = np.cosh(eta2/2.0)

    num = sin_th * sh1 * sh2
    den = ch1*ch2 + cos_th * sh1 * sh2
    return 2.0 * np.arctan2(num, den)

def axis_classic(v1, v2):
    v1 = np.array(v1, float); v2 = np.array(v2, float)
    u1 = v1/np.linalg.norm(v1) if np.linalg.norm(v1) > 0 else np.array([1.0,0.0,0.0])
    u2 = v2/np.linalg.norm(v2) if np.linalg.norm(v2) > 0 else np.array([1.0,0.0,0.0])
    a = np.cross(u2, u1)
    n = np.linalg.norm(a)
    return a/n if n > 0 else np.array([1.0,0.0,0.0])

# ----- Демонстрация -----

if __name__ == "__main__":
    c = 1.0
    v1 = np.array([0.99, 0.0, 0.0]) * c     # 0.6c по X
    v2 = np.array([0.0, 0.99, 0.0]) * c     # 0.5c по Y

    # униметрический расчёт
    rW = wigner_rotor_from_velocities(v1, v2, c)
    axis_u, phi_u = raxis_angle(rW)

    # классическая формула
    phi_cl = wigner_angle_classic(v1, v2, c)
    axis_cl = axis_classic(v1, v2)

    # сравнение (ось - до знака; сравниваем |dot|)
    axis_align = abs(np.dot(axis_u, axis_cl))
    angle_err  = abs(phi_u - phi_cl)

    print(f"phi_unimetry (deg) = {np.rad2deg(phi_u):.8f}")
    print(f"phi_classic (deg)  = {np.rad2deg(phi_cl):.8f}")
    print(f"|axis dot|         = {axis_align:.12f}   (1.0 = совпало)")
    print(f"|angle diff| (deg) = {np.rad2deg(angle_err):.12e}")

    # простая проверка точности
    assert axis_align > 1 - 1e-10, "Оси не совпадают (с точностью до знака)."
    assert angle_err < 1e-10, "Углы не совпали в пределах машинной точности."
```

## Физическая мотивация и онтология фазового вектора χ

Идея проста: **фаза — это то, что реально можно посчитать тиками собственного осциллятора** (атомные часы, «де Бройлевские» часы и т.п.). Мы рассматриваем **скорость изменения этой фазы** как геометрический объект и договариваемся, как из него извлекаются наблюдаемые «время» и «пространство».

### Операциональные постулаты

**P1 (локальный отсчёт).** У любого объекта есть монотонный локальный счётчик фазы (тикер). В пределе идеальной стабильности его приращение пропорционально собственному времени: \(d\sigma \propto d\tau\). Мы будем использовать \(\sigma\) как «фазовую дугу».

**P2 (фазовый поток как вектор).** Изменение фазы задаётся **вектором фазового потока** \(\boldsymbol\chi\) в евклидовом прото-пространстве \(\mathcal E\). Наблюдатель выбирает ортонормированную тройку \((\mathbf e_1,\mathbf e_2,\mathbf e_3)\) и тем самым задаёт наблюдаемое 3-пространство \(S\subset\mathcal E\) и его ортогональное дополнение \(S^\perp\).

**P3 (как получить наблюдаемые).** Проецируем \(\boldsymbol\chi\) на \(S\) и \(S^\perp\):
\[
\mathbf l:=P_S\boldsymbol\chi,\qquad
t:=\|P_{S^\perp}\boldsymbol\chi\|.
\]
Дальше упаковываем это в **реальный кватернион**
\[
q \;=\; t+\mathbf l\ \in\ \mathbb H,\qquad |q|^2=t^2+\|\mathbf l\|^2=\|\boldsymbol\chi\|^2.
\]
Кинематика — это **роторы** (кватернионные повороты) потока \(\boldsymbol\chi\) относительно тетрады наблюдателя.

### Почему это не «пустая переупаковка»

1) **Физичность χ.** \(\boldsymbol\chi\) — не «мистическая сущность», а компактный носитель того, что реально измеряется двумя приборами:
   - скаляр \(t\): число тиков собственного осциллятора (по сути — \(d\tau\) после нормировки);
   - вектор \(\mathbf l\): то, как **ориентация наблюдателя** «снимает» пространственные компоненты фазового потока (IMU/гиротетрада задаёт оси).

2) **Евклидовость как удобный калькулятор.** Работать в \(\mathcal E\) удобно: буст — это обычный 2D-поворот в плоскости \((\mathbf e_t,\hat{\mathbf u})\):
\[
d=\cos\frac{\psi}{2}+\hat{\mathbf u}\sin\frac{\psi}{2},\qquad q\mapsto dqd,
\]
а **гипербола СТО** появляется простой заменой параметра
\[
\sin\psi\leftrightarrow\tanh\eta,\quad \cos\psi\leftrightarrow\operatorname{sech}\eta,
\]
т.е. это не новая физика, а **другая координатизация** тех же инвариантов.

3) **Добавленная ценность — алгоритм.** В этой упаковке извлечение «поворота Вигнера» становится буквально одной строкой
\[
d_{12}=\sqrt{L_{12}L_{21}},\quad r_W=\bar d_{12}\,L_{12},
\]
что даёт простой, стабильный и быстрый код (и мы показываем численный контроль). Это не претензия на новый эффект — это **рабочий рецепт**, который хорошо живёт в вычислениях.

### Почему кватернионы, а не «просто рапидность»

- **Рапидность \(\eta\)** — скаляр 1D-бустов. В 3D композиция бустов порождает **поворот** (Вигнера). Кватернионный **ротор** держит и ось, и угол в одном объекте; композиция — это просто умножение.
- **Единый объект для R и D.** Пространственный поворот \(r\) и «буст» \(d\) — один и тот же тип единичного кватерниона (меняется плоскость вращения). Это снимает разрыв «гипербола vs сфера» и делает код однородным.
- **Численная устойчивость.** Нет гимбал-локов, нормализация — одно деление, выбор главной ветви корня для \(d_{12}\) — очевиден (\(\Re d_{12}\ge0\)).

### Что с «khōra» и «комплексным срезом»

- Термин *khōra* — **метафора** для интуиции «носителя фазы»; его легко убрать/перенести в примечание, чтобы не создавать философского тумана.
- **Комплексный срез не обязателен.** В этой работе всё делается **реальными кватернионами**. «Гиперболический» вид появляется **только** как перепараметризация \(\psi\to\eta\); он не требует комплексирования.

### Мини-связь с наблюдаемыми

Выбирая «часы» эволюции, получаем привычные формулы:
\[
dt=\gamma\,d\tau,\quad \beta=\sin\psi,\quad \gamma=\frac1{\cos\psi}.
\]
Дальше следуют стандартные эффекты (дилатация, доплер, Томас/Вигнер). То есть «онтология» здесь строго операциональна: **тик-счётчик + тетрада** \(\Rightarrow\) \(q=t+\mathbf l\) \(\Rightarrow\) роторная кинематика.


### Как «убрать» $c^2$ из формул (три эквивалентных способа)

Причина появления $c^2$ проста: в уравнении Фридмана $\Lambda$ имеет размерность $[{\rm длина}]^{-2}$, а $\Omega_u^2$ — $[{\rm время}]^{-2}$. Чтобы согласовать размерности, в стандартной записи стоит множитель $c^2$. Убрать его можно одним из трёх путей.

---

**A) Натуральные/геометрические единицы ($c\equiv 1$)**

Просто полагаем $c=1$. Тогда
$$H^2=\frac{8\pi G}{3}\rho-\frac{k}{a^2}+\frac{\Lambda}{3},$$
а HD-форма
$$\Omega_u^2=\frac{u^2}{4}\Bigl(\frac{8\pi G}{3}\rho\Bigr)-\frac{k}{4u^2}+\frac{\Lambda}{12}\,u^2.$$
Соответствие «$\Lambda$ из шероховатости» становится
$$\boxed{\ \Lambda=3\,\kappa_{\rm HD}\,\sigma_\chi^2\ }\quad\text{или при 3D-дисперсии}\quad \boxed{\ \Lambda=\kappa_{\rm HD}\,\sigma_{\rm vec}^2\ }.$$

---

**B) Переопределить константу: $\hat\Lambda\equiv \Lambda c^2$**

Держим SI-единицы, но переобозначим:
$$\Omega_u^2=\frac{u^2}{4}\Bigl(\frac{8\pi G}{3}\rho\Bigr)-\frac{k\,c^2}{4u^2}+\frac{\hat\Lambda}{12}\,u^2,\qquad \hat\Lambda\equiv\Lambda c^2.$$
Тогда соответствие записывается вообще **без** $c^2$:
$$\boxed{\ \hat\Lambda=3\,\kappa_{\rm HD}\,\sigma_\chi^2=\kappa_{\rm HD}\,\sigma_{\rm vec}^2\ }.$$
Если нужна «классическая» $\Lambda$, просто делим на $c^2$ в самом конце: $\Lambda=\hat\Lambda/c^2$.

---

**C) Перейти к безразмерным плотностям (через $H_0$)**

Вводим $\rho_{c0}\equiv 3H_0^2/(8\pi G)$ и $\Omega_{x0}\equiv \rho_x^0/\rho_{c0}$. Тогда HD-«лестница» записывается без $c$:
$$
\boxed{\ \Omega_u^2=\frac{H_0^2}{4}\Bigl(\Omega_{r0}\,u^{-6}+\Omega_{m0}\,u^{-4}+\Omega_{k0}\,u^{-2}+\Omega_{\Lambda0}\,u^{2}\Bigr)\ },
$$
где $\Omega_{\Lambda0}\equiv \Lambda c^2/(3H_0^2)$.

Связь с дисперсией:
- если используем **проекционную** дисперсию по одному направлению $\sigma_{\rm proj}^2$, то из $\Lambda c^2=3\,\kappa_{\rm HD}\,\sigma_{\rm proj}^2$ получаем
  $$\boxed{\ \Omega_{\Lambda0}=\kappa_{\rm HD}\,\frac{\sigma_{\rm proj}^2}{H_0^2}\ }.$$
- если используем **полную 3D-дисперсию** $\sigma_{\rm vec}^2=3\,\sigma_{\rm proj}^2$, то из $\Lambda c^2=\kappa_{\rm HD}\,\sigma_{\rm vec}^2$ получаем
  $$\boxed{\ \Omega_{\Lambda0}=\frac{\kappa_{\rm HD}}{3}\,\frac{\sigma_{\rm vec}^2}{H_0^2}\ }.$$

> На практике удобно фиксировать изотропию и работать с одной из форм:  
> — «в проекции»: $\ \Omega_{\Lambda0}=\kappa_{\rm HD}\,(\sigma_{\rm proj}^2/H_0^2)$;  
> — «в 3D»: $\ \Omega_{\Lambda0}=(\kappa_{\rm HD}/3)\,(\sigma_{\rm vec}^2/H_0^2)$.  
> При выборе нормировки так, что $\kappa_{\rm HD}=1$, формулы становятся максимально компактными.

---

**Что выбрать?**  
- Для физического смысла и простой алгебры — вариант **B** ($\hat\Lambda$).  
- Для сопоставления с наблюдениями и таблицами параметров — вариант **C** (через $\Omega_{\Lambda0}$).  
- Для компактности в выводах/заметках — **A** ($c=1$).

Во всех трёх вариантах ключевая идея сохраняется: **именно дисперсия микроповоротов** задаёт «$\Lambda$-вклад», а $c^2$ — чисто единичный артефакт, который можно последовательно убрать выбором нормировки.


### 9.x) Первая проверка: **монотонное наращивание** (эффект может только расти)

Идея: шероховатость — это **накопленный** (неотрицательный) разброс микроповоротов. При увеличении «длины пути» $L$ (в $\chi$ или с ростом $z$) накопленная дисперсия не убывает.

**Модель.** Пусть $\delta\omega(\chi)$ — нулевая в среднем стационарная «фазовая угловая скорость»-флуктуация с ковариацией $C(\tau)=\mathbb{E}[\delta\omega(\chi)\delta\omega(\chi+\tau)]\ge 0$ для $|\tau|\lesssim \ell_c$ (короткая корреляция). Тогда фаза
$$
\zeta(L)-\zeta(0)=\int_0^L \delta\omega(\chi)\,d\chi,
$$
и её накопленная дисперсия
$$
\Sigma(L)\;\stackrel{\rm def}{=}\;{\rm Var}\!\left[\int_0^L \delta\omega(\chi)\,d\chi\right]
=\int_0^L\!\!\int_0^L C(|\chi-\chi'|)\,d\chi\,d\chi'
=2\!\int_0^L (L-\tau)\,C(\tau)\,d\tau.
$$
Из последней формы видно:
$$
\boxed{\ \frac{d\Sigma}{dL}=2\!\int_0^L C(\tau)\,d\tau\ \ge 0\quad\Rightarrow\quad \Sigma(L)\ \text{неубывает.}\ }
$$
Для $L\gg \ell_c$: $\ \Sigma(L)\simeq 2L\!\int_0^\infty C(\tau)\,d\tau\ \propto L$ (асимптотически линейный рост).

**Наблюдательный прокси.** Через пункты 5.4–5.5 мы связали «$\Lambda$-вклад» с дисперсией:
- 1D-проекция: $\ \Lambda c^2=3\,\kappa_{\rm HD}\,\sigma_{\rm proj}^2$,
- 3D-вектор: $\ \Lambda c^2=\kappa_{\rm HD}\,\sigma_{\rm vec}^2$.
Если нормировать $\omega$ на $H_0$ и взять $\kappa_{\rm HD}=1$, то
$$
\boxed{\ \Omega_{\Lambda}(\le z)\ \simeq\ \widehat{\sigma}_{\rm proj}^2(\le z)\ \propto\ \Sigma(L(z))\ },
$$
то есть **оценка** $\Omega_{\Lambda}$, восстановленная **накопительно** из данных до красного смещения $z$, должна быть неубывающей функцией $z$ (с точностью до статистического шума).

**Практический тест (минимальный набор шагов):**
1) **Накопительная выборка.** Отсортируйте источники по $z$ и сформируйте вложенные подвыборки $S_k=\{\,z_i\le z_k\,\}$.
2) **Оценка дисперсии.** Постройте в каждой $S_k$ оценку $\widehat{\sigma}_{\rm proj}^2(\le z_k)$, вычитая известные вклады шума (инструментальный, пекулярные скорости, линзирование — либо вычитать модели, либо трактовать как часть «шероховатости»; важно лишь, что итоговая добавка **неотрицательна**).
3) **Монотонность.** Проверьте тренд:
   - непараметрически: тест Манна–Кендалла на неубывание,
   - или через изотонную регрессию (PAVA), сравнив RSS с несвязанной оценкой; значимая **отрицательная** тенденция — сигнал против теории.
4) **Асимптотика.** Приращение $\Delta\widehat{\sigma}^2$ между соседними бинами должно стабилизироваться к константе (рост $\propto L$) при $z\gg z(\ell_c)$; это даёт оценку $\int_0^\infty C(\tau)\,d\tau$.

**Дополнительные монотонности:**
- **По числу независимых лучей** (сильно линзированные системы): дисперсия между путями не убывает при добавлении новых независимых изображений того же источника.
- **По угловому разделению путей:** при фиксированном $z$ растёт с эффективной разностью геометрий лучей (длиннее «маршрут» $\Rightarrow$ больше накопление).

> **Критерий №1 (быстрый):** если в качественных данных оценка $\widehat{\sigma}_{\rm proj}^2(\le z)$ систематически **падает** при переходе к более глубоким подвыборкам (после вычитания известных шумов), то гипотеза «шероховатость как источник $\Lambda$» опровергается. При отсутствии падения и с ростом до плато — гипотеза проходит первичную проверку.


```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
zeta_variance_scan.py — тест монотонности дисперсии по окну в ζ
и карта амплитуды K в зависимости от направления.

Вход: таблица с колонками RA, DEC, z (и, опционально, z_err для photo-z).
Идея: при фиксированной ширине окна Δζ дисперсия по z должна расти ~ z(1+z)^2.
Мы скользим окном по ζ, считаем s_z^2(ζ), фитируем K в модели:
    s_z^2 ≈ K * g(z),   где g(z) = z (1+z)^2,
и проверяем монотонность s_z^2(ζ) (Spearman ρ ≥ 0, нарушения соседних шагов).

Выход:
  • CSV по кэпсам (lon,lat,N,K,K_err, rho, frac_down, …)
  • Карты Mollweide: sphere_K.png, sphere_rho.png, sphere_fracdown.png, sphere_N.png
  • (опц.) сырые траектории s_z^2(ζ) для заданного кэпса

Пример (фибоначчи-сетка, R=15°, Δζ=0.08 рад, шаг 0.02 рад, для z∈[0.02,1.0]):
  python zeta_sigma_map.py --cat DESI_DR1_like.tsv --cols RA DEC Z \
    --frame gal --grid fibonacci --npix 4096 --radius 15 \
    --zmin 0.02 --zmax 1.0 --dzetas 0.06 0.08 0.10 0.12 --step 0.02 \
    --minNcap 1000 --minNwin 120 --subtract-zerr \
    --outdir zeta_sigma_out
"""
import argparse, os, sys, math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ---------- coords ----------
def radec_to_gal(ra_deg, dec_deg):
    ra = np.deg2rad(ra_deg); dec = np.deg2rad(dec_deg)
    x = np.cos(dec)*np.cos(ra); y = np.cos(dec)*np.sin(ra); z = np.sin(dec)
    R = np.array([[-0.0548755604, -0.8734370902, -0.4838350155],
                  [ 0.4941094279, -0.4448296300,  0.7469822445],
                  [-0.8676661490, -0.1980763734,  0.4559837762]])
    xg, yg, zg = (R @ np.vstack([x,y,z]))
    b = np.arcsin(np.clip(zg, -1, 1))
    l = (np.arctan2(yg, xg)) % (2*np.pi)
    return np.rad2deg(l), np.rad2deg(b)

def sph_to_cart(lon_deg, lat_deg):
    L = np.deg2rad(lon_deg); B = np.deg2rad(lat_deg)
    cB = np.cos(B)
    return np.stack([cB*np.cos(L), cB*np.sin(L), np.sin(B)], axis=-1)

def cart_to_sph(xyz):
    x,y,z = xyz[...,0], xyz[...,1], xyz[...,2]
    B = np.arcsin(np.clip(z, -1.0, 1.0))
    L = (np.arctan2(y, x)) % (2*np.pi)
    return np.rad2deg(L), np.rad2deg(B)

# ---------- sphere grids ----------
def grid_fibonacci(npix):
    i = np.arange(npix, dtype=float)
    z = 1 - 2*(i+0.5)/npix
    r = np.sqrt(np.maximum(0.0, 1 - z*z))
    phi = (np.pi * (1 + 5**0.5)) * i
    x = r*np.cos(phi); y = r*np.sin(phi)
    lon = (np.rad2deg(np.arctan2(y, x)) % 360.0)
    lat = np.rad2deg(np.arcsin(z))
    xyz = np.stack([x, y, z], axis=1)
    return lon, lat, xyz

def grid_lonlat(step_lon=5.0, step_lat=5.0):
    lons = np.arange(0.0, 360.0, step_lon)
    lats = np.arange(-90.0, 90.0+1e-9, step_lat)
    LL, BB = np.meshgrid(lons, lats)
    lon = LL.ravel(); lat = BB.ravel()
    xyz = sph_to_cart(lon, lat)
    return lon, lat, xyz

# ---------- zeta transform ----------
def z_to_zeta(z):
    z = np.clip(z, 0, None)
    return 2.0*np.arctan(np.sqrt(z))

def dz_dzeta(z):
    # dz/dζ = sqrt(z)*(1+z)
    z = np.clip(z, 0, None)
    return np.sqrt(z)*(1.0+z)

# ---------- sliding window over ζ ----------
def sliding_windows(z, zeta, z_err=None, dzeta=0.08, step=0.02, zmin=0.0, zmax=np.inf, minNwin=50, subtract_zerr=False):
    """
    Возвращает массивы по окнам: z_c, Nz, var_z, (опц) var_corr.
    var_z — несмещённая выборочная дисперсия по z в окне.
    Если subtract_zerr и есть z_err: var_corr = var_z - median(z_err^2) в окне (>=0).
    """
    m = np.isfinite(z) & np.isfinite(zeta) & (z>=zmin) & (z<=zmax)
    if z_err is not None:
        m &= np.isfinite(z_err)
    z, zeta = z[m], zeta[m]
    z_err = z_err[m] if z_err is not None else None

    order = np.argsort(zeta)
    z, zeta = z[order], zeta[order]
    z_err = z_err[order] if z_err is not None else None

    zeta_min, zeta_max = zeta.min() if zeta.size else 0.0, zeta.max() if zeta.size else 0.0
    if zeta.size==0:
        return np.array([]), np.array([]), np.array([]), np.array([])

    centers = np.arange(zeta_min + dzeta/2, zeta_max - dzeta/2 + 1e-9, step)
    zc_list, N_list, varz_list, varcorr_list = [], [], [], []
    L = zeta.size
    j0 = 0
    for c in centers:
        a, b = c - dzeta/2, c + dzeta/2
        # сдвигаем левую границу
        while j0 < L and zeta[j0] < a: j0 += 1
        j1 = j0
        while j1 < L and zeta[j1] <= b: j1 += 1
        n = j1 - j0
        if n < minNwin:
            zc_list.append(np.nan); N_list.append(n); varz_list.append(np.nan); varcorr_list.append(np.nan)
            continue
        chunk = z[j0:j1]
        # несмещённая выборочная дисперсия
        v = float(np.var(chunk, ddof=1)) if n>1 else np.nan
        if subtract_zerr and (z_err is not None):
            m2 = float(np.median(np.square(z_err[j0:j1])))
            v_corr = max(v - m2, 0.0) if np.isfinite(m2) else np.nan
        else:
            v_corr = np.nan
        zc_list.append(float(np.median(chunk)))   # центр окна — медианный z
        N_list.append(int(n))
        varz_list.append(v)
        varcorr_list.append(v_corr)

    return np.array(zc_list), np.array(N_list), np.array(varz_list), np.array(varcorr_list)

# ---------- fit K in s_z^2 ≈ K * g(z) ----------
def fit_K(zc, varz, weights=None):
    """
    Простая непересекаемая регрессия через начало координат:
        y = K * g(z), y=varz, g(z)=z(1+z)^2.
    Возвращает K, σ_K и число точек.
    """
    m = np.isfinite(zc) & np.isfinite(varz)
    zc, varz = zc[m], varz[m]
    if weights is not None:
        w = weights[m]
    else:
        w = np.ones_like(zc)

    if zc.size < 3:
        return np.nan, np.nan, int(zc.size)

    g = zc * (1.0+zc)**2
    X = g
    # взвешенное МНК через начало: K = (Σ w X y)/(Σ w X^2)
    denom = np.sum(w * X * X)
    if denom <= 0: return np.nan, np.nan, int(zc.size)
    K = float(np.sum(w * X * varz) / denom)
    # дисперсия оценки K (при известной дисперсии y) здесь неочевидна; оценим робастно:
    resid = varz - K*X
    # весим по w, нормируем по числу св.степеней
    dof = max(1, zc.size - 1)
    s2 = float(np.sum(w * resid*resid) / (dof * np.mean(w)))  # грубая шкала
    sigma_K = float(np.sqrt(s2 / denom))
    return K, sigma_K, int(zc.size)

# ---------- monotonicity diagnostics ----------
def monotonic_metrics(varz_series):
    """
    varz_series — последовательность s_z^2 по центрам ζ-окон (в порядке возрастания ζ).
    Считаем:
      • Spearman ρ(varz, индекс)
      • frac_down: доля «соседних падений» (var[i+1] < var[i]) среди валидных пар.
    """
    v = np.asarray(varz_series, float)
    m = np.isfinite(v)
    v = v[m]
    if v.size < 4:
        return np.nan, np.nan
    # Spearman ρ с ранжированием индексов
    ranks_v = v.argsort().argsort().astype(float)
    ranks_i = np.arange(v.size, dtype=float)
    # корреляция Пирсона рангов
    def pearson(x,y):
        x = x - x.mean(); y = y - y.mean()
        den = np.sqrt((x*x).sum() * (y*y).sum())
        return float((x*y).sum()/den) if den>0 else np.nan
    rho = pearson(ranks_v, ranks_i)
    # frac_down
    downs = np.sum(v[1:] < v[:-1])
    frac_down = downs / (v.size - 1.0)
    return rho, float(frac_down)

# ---------- plotting ----------
def mollweide_scatter(lon_deg, lat_deg, val, title, fname, vcenter='sym', cmap='coolwarm'):
    L = (np.asarray(lon_deg) - 180.0) * np.pi/180.0
    B = np.asarray(lat_deg) * np.pi/180.0
    V = np.asarray(val, float)
    fig = plt.figure(figsize=(9,5))
    ax = fig.add_subplot(111, projection='mollweide')
    if vcenter == 'sym':
        vmax = np.nanpercentile(np.abs(V), 98)
        vmin, vmax = -vmax, +vmax
    else:
        vmin = np.nanpercentile(V, 2)
        vmax = np.nanpercentile(V, 98)
    sc = ax.scatter(L, B, c=V, s=8, alpha=0.9, cmap=cmap, vmin=vmin, vmax=vmax)
    ax.grid(True, lw=0.5, alpha=0.4)
    cb = fig.colorbar(sc, ax=ax, shrink=0.8, pad=0.08)
    cb.ax.set_ylabel(title, rotation=90)
    ax.set_title(title)
    fig.tight_layout(); fig.savefig(fname, dpi=160); plt.close(fig)

# ---------- main ----------
def main():
    ap = argparse.ArgumentParser()
    ap.add_argument('--cat', type=str, required=True, help='путь к каталогу (csv/tsv/whitespace)')
    ap.add_argument('--sep', type=str, default=None, help="разделитель: ',', '\\t' или 'auto' (по умолчанию auto/whitespace)")
    ap.add_argument('--cols', nargs='+', required=True, help='имена колонок: например RA DEC Z [ZERR]')
    ap.add_argument('--frame', choices=['gal','equ'], default='gal', help='в какой системе строить сетку')
    ap.add_argument('--grid', choices=['fibonacci','lonlat'], default='fibonacci')
    ap.add_argument('--npix', type=int, default=4096, help='число центров для фибоначчи')
    ap.add_argument('--step-lon', type=float, default=6.0)
    ap.add_argument('--step-lat', type=float, default=4.0)
    ap.add_argument('--radius', type=float, default=15.0, help='радиус кэпса (deg)')

    ap.add_argument('--zmin', type=float, default=0.02)
    ap.add_argument('--zmax', type=float, default=1.0)
    ap.add_argument('--dzeta', type=float, default=0.08, help='ширина окна по ζ (radian)')
    ap.add_argument('--step',  type=float, default=0.02, help='шаг окна по ζ (radian)')
    ap.add_argument('--minNcap', type=int, default=1000, help='минимум объектов в кэпсе')
    ap.add_argument('--minNwin', type=int, default=100, help='минимум объектов в одном окне')
    ap.add_argument('--subtract-zerr', action='store_true', help='вычитать median(z_err^2) из var_z в окнах (для photo-z)')

    ap.add_argument('--outdir', type=str, default='zeta_scan_out')
    args, _ = ap.parse_known_args()
    os.makedirs(args.outdir, exist_ok=True)

    # --- загрузка ---
    if args.sep is None or args.sep.lower()=='auto':
        # пробуем whitespace
        df = pd.read_csv(args.cat, sep=r'\s+', engine='python', comment='#')
    else:
        df = pd.read_csv(args.cat, sep=args.sep, engine='python', comment='#')
    cols = args.cols
    if len(cols) < 3:
        print("Нужно минимум 3 колонки: RA DEC Z [ZERR]", file=sys.stderr); sys.exit(1)
    RA = df[cols[0]].astype(float).values
    DEC= df[cols[1]].astype(float).values
    Z  = df[cols[2]].astype(float).values
    ZERR = df[cols[3]].astype(float).values if len(cols)>3 and cols[3] in df.columns else None

    # фильтр по z
    m = np.isfinite(RA) & np.isfinite(DEC) & np.isfinite(Z) & (Z>=args.zmin) & (Z<=args.zmax)
    if ZERR is not None:
        m &= np.isfinite(ZERR)
    RA,DEC,Z = RA[m],DEC[m],Z[m]
    ZERR = ZERR[m] if ZERR is not None else None

    # координаты к нужной системе
    if args.frame=='gal':
        L,B = radec_to_gal(RA, DEC)
        lon_sn, lat_sn = L, B
    else:
        lon_sn, lat_sn = RA, DEC
    xyz_sn = sph_to_cart(lon_sn, lat_sn)
    zeta = z_to_zeta(Z)

    # сетка центров
    if args.grid=='fibonacci':
        lon_c, lat_c, xyz_c = grid_fibonacci(args.npix)
    else:
        lon_c, lat_c, xyz_c = grid_lonlat(args.step_lon, args.step_lat)

    cosR = np.cos(np.deg2rad(args.radius))

    rows = []
    for i in range(xyz_c.shape[0]):
        c = xyz_c[i]
        mask = (xyz_sn @ c) >= cosR
        Ni = int(mask.sum())
        if Ni < args.minNcap:
            rows.append(dict(lon=lon_c[i], lat=lat_c[i], N=Ni,
                             K=np.nan, K_err=np.nan, rho=np.nan, frac_down=np.nan))
            continue
        z_cap   = Z[mask]
        zeta_cap= zeta[mask]
        zerr_cap= ZERR[mask] if ZERR is not None else None

        zc, Nw, varz, varcorr = sliding_windows(
            z_cap, zeta_cap, z_err=zerr_cap,
            dzeta=args.dzeta, step=args.step,
            zmin=args.zmin, zmax=args.zmax,
            minNwin=args.minNwin,
            subtract_zerr=args.subtract_zerr
        )
        # выбираем какую дисперсию использовать для фита: с коррекцией если доступна
        use_var = np.where(np.isfinite(varcorr), varcorr, varz)

        # веса окна — по количеству объектов в нём
        K, K_err, M = fit_K(zc, use_var, weights=Nw)

        # монотонность по исходной varz:
        rho, frac_down = monotonic_metrics(use_var)

        rows.append(dict(lon=lon_c[i], lat=lat_c[i], N=Ni,
                         K=K, K_err=K_err, rho=rho, frac_down=frac_down))

    out = pd.DataFrame(rows)
    out_csv = os.path.join(args.outdir, "zeta_variance_scan_results.csv")
    out.to_csv(out_csv, index=False)
    print("Saved:", out_csv)

    # карты
    def mw(val, title, name, vcenter='sym', cmap='coolwarm'):
        mollweide_scatter(out['lon'].values, out['lat'].values, val, title,
                          os.path.join(args.outdir, name), vcenter=vcenter, cmap=cmap)
    mw(out['K'].values,       f"K in s_z^2 ≈ K·z(1+z)^2, R={args.radius}°, Δζ={args.dzeta}", "sphere_K.png")
    mw(out['rho'].values,     "Spearman ρ[var_z(ζ), ζ-index]", "sphere_rho.png", vcenter='sym')
    mw(out['frac_down'].values, "fraction of local decreases of var_z", "sphere_fracdown.png", vcenter=None, cmap='viridis')
    mw(out['N'].values,       "N per cap", "sphere_N.png", vcenter=None, cmap='viridis')
    print("Saved maps in:", args.outdir)

if __name__ == "__main__":
    main()
```
