# Фазовый формализм СТО - Униметрия

Мы привыкли воспринимать время и пространство как данность. Фазовый формализм униметрии предлагает посмотреть [иначе](https://ru.wikipedia.org/wiki/%D0%9C%D0%B8%D1%84_%D0%BE_%D0%BF%D0%B5%D1%89%D0%B5%D1%80%D0%B5): время и пространство — это производные от единого параметра, который мы называем фазой.

В этой концепции релятивистские эффекты — замедление времени, доплеровский сдвиг и т.д. — возникают как естественные проявления движения потоков в фазовом пространстве. Формализм не вводит новые законы физики, он лишь предлагает удобный язык и инструмент для их описания. Подобно лагранжевому и гамильтоновому подходу в классической механике, это попытка упростить структуру уже известных зависимостей и сделать явными причинно-следственные связи.

## 1. Время и пространство как производные
Предположим, что существует переменная величина, назовём её __фаза__ $\vec{\chi} \in \mathbb{C}$, изменение которой порождает наблюдаемые эффекты времени и пространства. Алгебраически это будет означать, что время-пространственные базовые векторы (единицы, меры) являются фазовой скоростью - производной фазы по направлениям в фазовом пространстве:
$$
\hat{h} \, dx_0 = \frac{\partial \vec{\chi}}{\partial \chi_h} \frac{d\chi_h}{d\chi} d\chi
= \tilde{H} \ d\chi, \ 
\mathbf{l} \, dx_l = \frac{\partial \vec{\chi}}{\partial \chi_l} \frac{d\chi_l}{d\chi} d\chi
= \tilde{L} \ d\chi, \,
l = 1,2,3
\qquad (1.1)
$$
Можно сказать, что мы динамизируем тензор пространства-времени через единую общую переменную, тогда, введя фазовую скорость интервала СТО $ds = \tilde{S} \ d\chi$ аналогично ранее представленым прото-скоростям, можем записать закон сохранения интервала в следующем виде:
$$
\tilde{S}^2 = \frac{ds^2}{d\chi^2}
= \frac{g_{ij} dx^i dx^j}{d\chi^2}
= g(\chi)
= \tilde{H}^2 - \tilde{L}^2
\qquad (1.2)
$$
Что равносильно:
$$
\tilde{H}^2 = \tilde{S}^2 + \tilde{L}^2 
\qquad (1.3)
$$
Такая форма предоставляет возможность выразить фазовые скорости $\tilde{S}, \ \tilde{L}$ через $\tilde{H}$:
$$ 
\tilde{S} = \tilde{H} \cos\theta, \ \tilde{L}
= \tilde{H} \sin\theta
\qquad (1.4)
$$
где $\theta$ - угол отклонения фазовой скорости от вещественной оси в фазовом пространстве.  
С математической точки зрения можно сказать, что (1.3) просто утверждает существование параметра, по которому темпоральная компонента интервала может быть разложена как сумма двух проекций на перпендикулярные оси в евклидовом пространстве, то есть если представить, что время-пространство индуцировано изменением единого параметра, то величина скорости по данному параметру вдоль темпоральной координатной оси может быть представлена как сумма скоростей вдоль двух отдельных компонент - пространственной и вне-пространственной (наблюдаемой и ненаблюдаемой, внешней и внутренней).  
Или, упрощая, (1.3) выглядит как формула сложения скоростей в евклидовом векторном пространстве. Предположим, что так и есть, но по какой-то причине при повороте "угла обзора" сохраняется не норма $\tilde{H}$, а проекция на вне-пространственную $\tilde{S}$.  

## 2. Фазовое пространство
Пусть математическое векторное пространство называется [__хора__ (Платон)](https://en.wikipedia.org/wiki/Kh%C3%B4ra?utm_source=chatgpt.com) и имеет свойства обыкновенного комплексного пространства $\mathbb{C}$.  
Геометрически мы определим два базисных вектора назвав их (__время__, __пространство__): $(\hat{h}, \, \mathbf{l}) \equiv \mathbb{C}$; получив в результате представление фазы $\vec{\chi} = R \ e^{\theta \mathbf{l}}, \ \theta \in [-\pi, \pi]$ как вектора в данном базисе со следуюшими отношениями:
$$
\tilde{H} = R, \quad \tilde{S} = R \cos\theta, \quad \tilde{L} = R \ \mathbf{l} \sin\theta \qquad (2.1)
$$
и (1.3) становится __законом сохранения__ полной фазы объекта при повороте вектора фазы.  
При этом мы задали фазовое пространство так, что частные производные равны единице, поэтому (1.1) можно переписать так:
$$
\hat{h} \, dx_0 = \frac{d\chi_h}{d\chi} d\chi = \tilde{H} \ d\chi, \ \mathbf{l} \, dx_l = \frac{d\chi_l}{d\chi} d\chi = \tilde{L} \ d\chi, \, l = 1,2,3 \qquad (2.2)
$$
Отметим, что вместе с тем, фазовое пространство, скорости которого являются базисами для наблюдаемого пространства-времени, связано с последним интегральным преобразованием вида:
$$
x^i = \int{ \tilde{X}^i(\chi) \ d\chi}, \quad i = 0,1,2,3 \qquad (2.3)
$$
где $\tilde{X}^i(\chi)$ — фазовые скорости, являющиеся проекциями производной фазы $\vec{\chi}$ на соответствующие направления базиса $(\hat{h}, \mathbf{l})$.  
Таким образом, привычные координаты времени и пространства представляют собой интегральные образы фазовых скоростей, а сами фазовые скорости — это «мост» между скрытым фазовым параметром и наблюдаемым многообразием Минковского.  
Ввиду отношений (2.2) и (2.3) фазовое пространство $\mathbb{C}$ выступает как порождающее пространство, а наблюдаемое пространство-время — как его проекционное отображение.  

## 3. Объекты
Любая __фундаментальная частица__ является __элементарным объектом__ униметрии и характеризуется ненулевой величиной фазы ($\vec{\chi} \ne 0$).  
Составные __объекты__ являются фазовыми конфигурациями, для полного представления которых в фазовом пространстве в основном требуется его расширение дополнительными измерениями, кроме случая __фотона__, фаза которого всегда направлена вдоль мнимой оси хоры:
$$
\mathbf{p} = \frac{d\vec{\chi}}{d\chi_l} = p \mathbf{l} \in \Im \qquad (3.1)
$$
Выделим все явления, отличные от перемещения фотонов, в отдельную категорию, аналогично СТО получив фотонное подпространство, которое в униметрии представлено мнимой осью, а остальные, более __сложные__ в этом контексте объекты, характеризуются наличием вещественной проекции фазы и ненулевой массой.
Тогда проводя параллель с СТО, сложный объект униметрии ассоциирован с любым __событием__ или чередой таковых - с мировой линией, а фотон - с точкой подпространства нулевого интервала, содержащей информацию о __событии__.  
Фаза любого объекта может быть приведена поворотом в хоре в _нулевое_ направление (положение нулевого отклонения вектора фазы) вдоль вещественной оси:
$$
\vec{\chi}_o = R \in \Re \qquad (3.2)
$$
Объект A, движущийся относительно покоящегося наблюдателя, будет представлен вектором фазы находящейся под углом к вектору фазы наблюдателя:
$$
\vec{\chi}_a = R \ e^{(\theta_A \, \mathbf{l})}, \, \sin \theta_a
= \frac{V}{\mathtt{c}}
\equiv \beta
\qquad (3.3)
$$
таким, что синус угла поворота его вектора фазы равен отношению взаимной скорости объектов к скорости света: $\theta > 0$ - отдаление, $\theta < 0$ - сближение, для перпендикулярного движения знак не важен (фактор квадратичен в уравнении).  
Таким образом, множество относительно неподвижных объектов формируют друг для друга фотонное пространство нулевой кривизны. Такие объекты будем называть _синфазными_.  

### 3.1. Пространство
Из определения (1.4) для пространственной составляющей следует, что в нулевом направлении компонента скорости фазы по мнимой оси хоры равна нулю как и базисный вектор пространства объекта, что противоречит наблюдаемой действительности - как минимум для обеспечения собственной проекции в наблюдаемое пространство у объекта должна существовать проекция фазы на мнимую ось, поэтому примем такую формулировку вырожденной и применимой лишь для малых объектов.  
Формализуем это требование следующим образом. Сложим два вектора фазы с равными пространственными отклонениями, но с разным знаком, и разделим результат на два, чтобы компенсировать норму. Получим:
$$
\vec{\chi}^\pm = R \, e^{\pm \zeta \ \mathbf{l}} \quad \chi_l = \frac{\vec{\chi}^+-\vec{\chi}^-}{2} = R \, \mathbf{l} \sin\zeta   \qquad (3.1.1)
$$
где $\vec{\chi}_l$ выступает в роли пространственного вектора фазы, а $\zeta$ - угол локального восприятия: не зависит от скорости в пространстве, является внутренним аттрибутом объекта (функционально связанным с массой и плотностью).

Таким образом, полная фаза объекта в _локальном_ направлении представляется как пара:
$$
\vec{\chi}_o = \vec{\chi}_\tau + \vec{\chi}_l = R \ \cos\zeta + R \ \mathbf{l} \sin\zeta \qquad (3.1.2)
$$
что эквивалентно разложению на темпоральную и пространственную компоненты.
При этом:

$\vec{\chi}_\tau = R \cos|\zeta|$ — временной базис,

$\vec{\chi}_l = R \, \mathbf{l} \, \sin|\zeta|$ — пространственный базис.

Если нормировать их на полный радиус $R$, то получаем ортогональные единичные базисы для фазового пространства:
$$
\hat{h} = \cos\zeta, \, \hat{\mathbf{l}} = \sin \zeta \qquad (3.1.3)
$$
Тогда __для единой нормы__ $\tilde{H}$ скорости фазы возникает три направления и соответствующих им вещественных значения, которые мы хотели бы обозначить.  

### 3.2 Абсолютное, локальное и наблюдаемое время  
Нулевому положению фазы соответствует **абсолютное** время $t = t(\tilde{H})$ - самое медленное. Из §3.1 следует, что не существует объекта, который бы функционировал по этому времени.  
Однако оно может быть использовано, например, для нормирования при переходе между обстоятельствами фаз разных скоростей.  
В §1.1 мы ввели базисный вектор времени как производную фазы объекта по вещественной компоненте.  
В локальном направлении вектор фазы всегда даёт его максимально возможное для объекта значение - __локальное__ время:  
$$
dx_0 = \frac{d}{d\chi}\,\Re(\vec{\chi}) \, d\chi 
= \frac{\vec{\chi}^+ + \vec{\chi}^-}{2}\, d\chi 
= \cos \zeta \, d\chi 
= d\tau 
\qquad (3.2.1)
$$
Здесь возникает **двойная аналогия углов**:  

- $\zeta$ — *внутренний угол объекта*  
  (характеризует его массу, собственную "плотность" потока).  
- $\theta$ — *внешний угол движения*  
  (характеризует скорость относительно наблюдателя).  

Таким образом, замедление времени у объекта имеет **двойную структуру**:  

- ограничено **внутренним состоянием** (через множитель $\cos\zeta$),  
- и дополнительно — **внешним движением** (через множитель $\cos\theta$).  

**Ключевой вывод:**  
наблюдаемое время всегда является результатом *двух наложенных преобразований* - внутреннего (природа объекта) и внешнего (его движение).  

Объект А, перемещающийся относительно наблюдателя, будет иметь _собственное_ - __наблюдаемое__ - время относительно покоящегося наблюдателя в виде вещественной части производной собственного вектора фазы по вектору фазы наблюдателя:
$$
\tilde{H}_a = \Re \left( \frac{d\vec{\chi}_a}{d\vec{\chi}_o} \right)
= \frac{\cos\theta_a \ d\chi_o}{d\chi_o}
= \cos\theta_a
= \sqrt{1-\sin^2\theta_a}
= \sqrt{1-\frac{V^2}{\mathtt{c}^2}}
= \frac{1}{\gamma} 
\qquad (3.2.2)
$$
что находится в согласовании с (3.3), так как является следствием сохранения фазы, а также соответствует экспериментально известному преобразованию времени и симметрично для восприятия объектом А перемещения наблюдателя за счёт геометрического представления - в собственной системе координат объекта А проекция фазы наблюдателя на вещественную ось тоже $\cos\theta_a$.  

### 3.3. Нормирование
Пусть локальное течение времени сложного объекта параметризовано **фазой**: вводим опорную частоту $\nu_0$ и связь
$$
d\tau=\frac{1}{\nu_0}\,d\chi_o.
$$
По цепному правилу:
$$
dx_0=\tilde{H}\,d\chi=\frac{dx_0}{d\chi_o}\,\frac{d\chi_o}{d\tau}\,d\tau
=\tilde{H}\,\dot{\chi}\,d\tau=\dot{H}\,d\tau,
\qquad (3.1.1)
$$
где введены обозначения
$$
\nu:=\frac{d\chi}{d\tau},\qquad \dot{\chi}:=\frac{\nu}{\nu_0},\qquad \dot{H}:=\tilde{H}\,\dot{\chi}.
$$
Отсюда видно, что мы принимаем скорость течения по локальному времени калибровкой: при нормировке $\dot{H}\equiv\mathtt{c}$ имеем $dx_0=\mathtt{c}\,d\tau$.

Аналогично для пространственной части:
$$
dx_l=\tilde{L}\,d\chi
=\frac{dx_l}{d\chi_o}\,\frac{d\chi_o}{dl}\,dl
=\tilde{L}\,\chi'\,dl
= L'\,dl,\quad l=1,2,3,
\qquad (3.1.2)
$$
где $\chi':=d\chi/dl$, $L':=\tilde{L}\,\chi'$. Из равенства (1.4) для света $dx_0=dx_l$ следует
$$
\mathtt{c}=\tilde{L}'\,\frac{dl}{d\tau},
$$
и при темпоральной нормировке на скорость света пространственный масштаб становится единичным: $\tilde{L}'=1$.  

### 3.4. Свет
Продолжим с (1.4) выразив обе координаты через фазу и нормирующие факторы:
$$
\frac{\mathtt{c}}{ \dot{\chi}} d\chi = \frac{1}{ \chi' } d\chi
\qquad (3.4.1)
$$
Получается:
$$
\mathtt{c} = \frac{\dot{\chi}}{ \chi' }
= \frac{dl}{ d\tau } \ne f(\chi)
\qquad (3.4.2)
$$
Откуда видно, что в нашей модели скорость света является параметром, связывающим движения по осям __вне зависимости от локального изменения фазы__.  
Во-первых, если посмотреть на (3.4.2) следующим образом:
$$
\mathtt{c} = \frac{\dot{\chi}}{ \chi' }
= \left(\frac{d\chi}{d\tau} \right) \left[ \frac{dl}{ d\chi } \right] \sim (\nu) \ [\lambda]
\qquad (3.4.3)
$$
То можно заметить сходство с характеристиками фотона _частота_ и _длина волны_, произведение которых даёт скорость света, где $\chi$ действительно соответствует фазе фотона.  
Во-вторых, из (3.4.1) и (3.4.2) следует:
$$
ds^2 = \mathtt{c}^2 \left( \frac{d\chi^2}{ \dot{\chi}^2}  - \frac{d\chi^2}{ \dot{\chi}^2} \right) = 0, \forall \dot{\chi}
\qquad (3.4.4)
$$
где решением для единичной частоты будет $\tau = \chi$ - собственное время фотона это его фаза. Интересно, что в такой нормировке длина вектора фазовой скорости фотона численно равна его длине волны $\tilde{H}_p = \lambda$.  
В-третьих, если и время, и пространство заданы единым параметром, то сохранение скорости света при преобразованиях - естественное следствие такой параметризации.  
Если скорость равна наклону траектории на диаграмме «время–пространство» в фазовых координатах:
$$
\frac{dx_l}{dx_0} 
= \frac{\tilde{L}\,d\chi}{\tilde{H}\,d\chi}
= \sin \theta
= \frac{V}{\mathtt{c}}
\equiv \beta
\qquad (3.4.5)
$$
Для фотона $\theta = \pi/2$, и потому:
$$
\frac{dx_l}{dx_0} = 1 \quad \Rightarrow \quad V = c.  
\qquad (3.4.6)
$$
Скорость света в униметрии возникает как отношение, задающее пространственный базис в зависимости от нормы $\tilde{H}$ скорости потока фазы.

### 3.5. Лоренц-фактор
Мы знаем, фактически, что в результате преобразования направления фазовой скорости вида:
$$
\tilde{H}^2 = \tilde{S}^2 + \tilde{L}^2 \overset{\cdot \dot{\chi}^2}{\longmapsto} \dot{H}^2 = \dot{S}^2 + \dot{L}^2 
\qquad(3.5.1)
$$
должен происходить переход нормы от $\tilde{H}$ к норме $\dot{S}$.  
_Лемма_: переход $\tilde{H}\rightarrow\dot{S}$ является проявлением эволюции фазовой скорости объекта по собственному времени $\dot{H}(\tau(\chi))$.
То есть перенормировка происходит в каждый момент приобретения дополнительной скорости объектом, в то время как величина $\tilde{S}$ является математической и проективной, когда в реальности объект имеет только величину своей скорости фазы $\tilde{H}$, восприятие которой меняется ввиду изменения скорости течения локального времени, то есть физически мотивированным было бы переписать (3.5.1) как:
$$
\dot{H}^2 = \tilde{H}^2 + \dot{L}^2
\qquad(3.5.2)
$$
которое читается так: фазовая скорость движущегося объекта $\tilde{H}$ в обстоятельствах наблюдателя, обусловленных его фазовой скоростью $(\dot{H},\dot{L})$, сохраняется.  
Докажем лемму, описав мгновенную перенормировку, соответствующую повороту фазовой скорости - как замену переменной $\chi \mapsto \tau(\chi)$ с локальным Якобианом:
$$
\frac{d\tau}{d\chi}=\cos\zeta(\chi)\,\cos\theta(\chi)
\quad\Longrightarrow\quad
\mathcal{J}(\zeta,\theta):=\frac{d\chi}{d\tau}=\frac{1}{\cos\zeta\,\cos\theta}.
\qquad(3.5.3)
$$

В результате компоненты фазовой скорости мгновенно масштабируются:
$$
\dot{H}
=\frac{dx_0}{d\tau}
=\frac{dx_0}{d\chi}\frac{d\chi}{d\tau}
=\tilde{H}\,\mathcal{J}
=\frac{\tilde{H}}{\cos\zeta\,\cos\theta},
\qquad
\dot{L}
=\tilde{L}\,\mathcal{J}
=\frac{\tilde{L}}{\cos\zeta\,\cos\theta}.
\qquad(3.5.4)
$$

В инфинитезимальной форме:
$$
d\ln\dot{H}=d\ln\mathcal{J}
=\tan\zeta\,d\zeta+\tan\theta\,d\theta
\;\Rightarrow\;
d\dot{H}=\dot{H}\,(\tan\zeta\,d\zeta+\tan\theta\,d\theta).
\qquad(3.5.5)
$$

- **Чистый буст** ($d\zeta=0$, $d\tilde{H}=0$):
  $$
  d\dot{H}=\dot{H}\,\tan\theta\,d\theta.
  $$
- **Чистое внутреннее изменение** ($d\theta=0$):
  $$
  d\dot{H}=\dot{H}\,\tan\zeta\,d\zeta.
  $$

> Если в нормировке §3.3 фиксировать $\dot{H}\equiv \mathtt{c}$, то (3.5.4) означает, что «изменение» достигается умножением на текущий Якобиан $\mathcal{J}$: физического скачка нет, есть локальная смена параметра.

Таким образом, (3.5.2) с учётом перепараметризации при чистом бусте $d\zeta=0$ (здесь и далее в статье внутренние изменения не рассматриваются) даёт псевдо-евклидову форму интервала в пространстве Минковского:
$$
\tilde{H}^2 = \dot{H}^2 - \dot{L}^2
= \sec^2\theta \ (\tilde{H}^2 - \tilde{L}^2)
= \gamma^2 \ (\tilde{H}^2 - \tilde{L}^2)
\qquad(3.5.6)
$$

**Вывод.**  
Переход нормы следует из инфинитезимального поворота фазы и пересчёта меры времени.  
- В хоре сохраняется полная скорость $\tilde{H}$ (евклидова норма).  
- В наблюдаемом времени сохраняется $\dot{S}$ (минковская норма).  
- Эти величины идентичны:
$$
\boxed{\;\tilde{H}=\dot{S}\;}
\qquad(3.5.7)
$$
так что наблюдаемое течение времени есть переинтерпретация полной скорости фазы в собственной системе объекта.

### 3.6. Быстрота и угол фазового поворота

По определению в униметрии скорость выражается через угол фазового поворота:
$$
\beta \equiv \frac{V}{\mathtt{c}} = \sin\theta.
\qquad(3.6.1)
$$

Быстрота задаётся соотношением $\tanh\eta=\beta$, откуда
$$
d\eta=\frac{d\beta}{1-\beta^2}.
\qquad(3.6.2)
$$

Дифференцируя $\beta=\sin\theta$ и подставляя в (3.6.2), получаем
$$
d\beta=\cos\theta\,d\theta,\qquad 1-\beta^2=\cos^2\theta
\;\;\Rightarrow\;\;
d\eta=\sec\theta\,d\theta.
\qquad(3.6.3)
$$

Интегрирование даёт
$$
\eta(\theta)=\int \sec\theta\,d\theta
=\ln\!\bigl|\sec\theta+\tan\theta\bigr|
=\tfrac12 \ln\!\left|\frac{1+\sin\theta}{1-\sin\theta}\right|.
\qquad(3.6.4)
$$

Эквивалентно (фиксируя $\eta(0)=0$):
$$
e^{\eta(\theta)}=\sqrt{\frac{1+\sin\theta}{1-\sin\theta}}.
\qquad(3.6.5)
$$

Подставляя $\sin\theta=\beta$, получаем продольный релятивистский доплер-фактор:
$$
e^{\eta}=\sqrt{\frac{1+\beta}{1-\beta}} \;\equiv\; r.
\qquad(3.6.6)
$$

Обратная зависимость угла от быстроты:
$$
\theta=\arcsin(\tanh\eta)
=\arcsin\!\left(\frac{e^{2\eta}-1}{e^{2\eta}+1}\right).
\qquad(3.6.7)
$$

Лоренц-фактор в трёх эквивалентных формах:
$$
\gamma=\frac{1}{\sqrt{1-\beta^2}}
=\frac{1}{\sqrt{1-\sin^2\theta}}
=\sec\theta
=\cosh\eta.
\qquad(3.6.8)
$$

Связь с перенормировкой темпоральной компоненты (чистый буст):
$$
d\ln \dot{H}=\tan\theta\,d\theta
=\sin\theta\,d\eta.
\qquad(3.6.9)
$$

### 3.7 Сложение скоростей

Быстрота аддитивна:
$$
\eta_{12}=\eta_1+\eta_2 
\qquad(3.7.1)
$$

Так как $\tanh\eta=\beta$ , имеем релятивистский закон сложения скоростей:
$$
\beta_{12}
=\tanh(\eta_1+\eta_2)
=\frac{\tanh\eta_1+\tanh\eta_2}{1+\tanh\eta_1\,\tanh\eta_2}
=\frac{\beta_1+\beta_2}{1+\beta_1\beta_2}.
\qquad (3.7.2)
$$

Через фазовый угол (напоминание: $\beta=\sin\theta$):
$$
\gamma=\frac{1}{\sqrt{1-\beta^2}}=\sec\theta,
\qquad
\gamma\beta=\tanh\eta=\sin\theta\,\sec\theta=\tan\theta.
\qquad(3.7.3)
$$

Отсюда, используя (3.7.1), удобно вычислять композицию как
$$
\gamma_{12}=\cosh(\eta_1+\eta_2)=\gamma_1\gamma_2(1+\beta_1\beta_2),
\quad
\gamma_{12}\beta_{12}=\sinh(\eta_1+\eta_2)=\gamma_1\gamma_2(\beta_1+\beta_2).
\qquad(3.7.4)
$$

### 3.8. Доплеровский сдвиг

Определим наблюдаемую частоту как скорость нарастания фазы по собственному времени наблюдателя:
$$
\nu := \frac{d\chi}{d\tau}
\qquad (3.8.1)
$$

Для двух последовательных фронтов волны приращение фазы одинаково, поэтому
$$
\frac{\nu_{\text{obs}}}{\nu_{\text{src}}}
=\frac{d\chi/d\tau_{\text{obs}}}{d\chi/d\tau_{\text{src}}}
=\frac{d\tau_{\text{src}}}{d\tau_{\text{obs}}}
\qquad (3.8.2)
$$

**Продольный случай.**  
Пусть источник движется со скоростью $V$ вдоль прямой к наблюдателю (подход) или от него (удаление). Интервал между эмиссиями в системе наблюдателя равен $\gamma\, d\tau_{\text{src}}$, а источник за это время смещается на $\pm V\,\gamma\, d\tau_{\text{src}}$ (знак «$+$» — удаление, «$-$» — подход). Тогда
$$
d\tau_{\text{obs}}
=\gamma\, d\tau_{\text{src}}\,(1\pm \beta),
\qquad
\beta=\frac{V}{\mathtt{c}}
\qquad (3.8.3)
$$
Отсюда продольный доплеровский фактор:
$$
\boxed{\;
\frac{\nu_{\text{obs}}}{\nu_{\text{src}}}
=\frac{1}{\gamma(1\pm \beta)}
\;}
\quad
\begin{cases}
\text{подход: } (-)\\[2pt]
\text{удаление: } (+)
\end{cases}
\qquad (3.8.4)
$$

Эквивалентные формы (с $\beta=\sin\theta,\ \gamma=\sec\theta$ и быстротой $\eta$):
$$
\frac{\nu_{\text{obs}}}{\nu_{\text{src}}}
=\sqrt{\frac{1\mp \beta}{1\pm \beta}}
=\sec\theta\, (1\mp \sin\theta)
=e^{\mp \eta}
\qquad (3.8.5)
$$

- **Подход (синее смещение):**
$$
\frac{\nu_{\text{obs}}}{\nu_{\text{src}}}
=\frac{1}{\gamma(1-\beta)}
=\sqrt{\frac{1+\beta}{1-\beta}}
=\sec\theta\,(1+\sin\theta)
=e^{+\eta}
\qquad (3.8.6)
$$

- **Удаление (красное смещение):**
$$
\frac{\nu_{\text{obs}}}{\nu_{\text{src}}}
=\frac{1}{\gamma(1+\beta)}
=\sqrt{\frac{1-\beta}{1+\beta}}
=\sec\theta\,(1-\sin\theta)
=e^{-\eta}
\qquad (3.8.7)
$$

Соответственно, отношение длин волн — обратное:
$$
\frac{\lambda_{\text{obs}}}{\lambda_{\text{src}}}
=\frac{1}{r},
\quad
r:=\frac{\nu_{\text{obs}}}{\nu_{\text{src}}}
=e^{\pm\eta}
\qquad (3.8.8)
$$

**Поперечный доплер.**  
При $\varphi=90^\circ$ (направление на источник перпендикулярно скорости в системе наблюдателя):
$$
\frac{\nu_{\text{obs}}}{\nu_{\text{src}}}
=\frac{1}{\gamma}
=\cos\theta
\qquad (3.8.9)
$$

**Общий угловой случай.**  
Если $\varphi$ — угол между вектором скорости источника и направлением на наблюдателя **в системе наблюдателя**, то
$$
\boxed{\;
\frac{\nu_{\text{obs}}}{\nu_{\text{src}}}
=\gamma\,(1-\beta\cos\varphi)
\;}
\qquad (3.8.10)
$$

> В терминах фазы (§3.4) доплеровский сдвиг — это пересчёт меры времени между источником и наблюдателем при неизменности счёта фазовых циклов: сдвигается не «сама волна», а масштаб параметра, по которому измеряется её фаза.

## 4. Заключение

Фазовый формализм униметрии показывает, что релятивистские эффекты естественно следуют из евклидовой геометрии **фазового пространства**, где время и пространство — это **производные проекции** единого параметра (фазы).

Главные итоги:

- **Интервал Минковского** возникает как сохранение нормы фазовой скорости при смене параметра (повороте фазы).
- **Замедление времени** и **сокращение длины** — геометрические следствия двойной структуры углов: внутреннего $\zeta$ (связанного с природой объекта) и внешнего $\theta$ (связанного с относительным движением).
- **Скорость света** выступает универсальной константой, связывающей темпоральное и пространственное масштабирование независимо от деталей фазовой динамики.
- **Лоренц-фактор** и **закон сложения скоростей** выводятся напрямую из геометрии поворотов на комплексной окружности, без явного введения гиперболической метрики.
- **Доплеровский сдвиг** неизбежно возникает как *масштабирование меры времени* между источником и наблюдателем при неизменности счёта фазовых циклов:  
  продольный фактор $r=\nu_{\text{obs}}/\nu_{\text{src}}=\frac{1}{\gamma(1\pm\beta)}=\sqrt{\frac{1\pm\beta}{1\mp\beta}}=e^{\pm\eta}$,  
  поперечный — $r=1/\gamma$.

Таким образом, униметрия сохраняет все проверенные экспериментом следствия СТО, но предлагает **единый фазовый язык**, в котором:
1) релятивистские эффекты — это проекции одной и той же фазовой динамики;
2) внутренние свойства объектов (масса, «плотность» потока) входят через угол $\zeta$;
3) время и пространство понимаются как интегральные образы фазовых скоростей, что создаёт мост к описанию областей, где «метрика» не задана априори.

**Перспективы.** Связать фазовую картину с квантованием (дискретизация фазы), обобщить к кривизне (ОТО) и исследовать коллективные фазовые конфигурации (поля, взаимодействия, когерентность), используя доплеровские соотношения как операциональные связки между наблюдаемыми частотой, длиной волны и фазовой быстротой.

## 5. Связь с известными теориями

В этом разделе показано, как фазовый формализм униметрии стыкуется с квантовой теорией, калибровочными полями, ОТО и космологией. Во всех случаях остаёмся в рамке: время и пространство — интегральные образы фазовой динамики (§§1–3), а релятивистские эффекты — геометрия поворота фазы.

### 5.1. Квантование как дискретизация фазы
Гипотеза: элементарная «ячейка» возникает как условие периодичности
$$
\Delta\chi=2\pi n,\quad n\in\mathbb{Z}.
$$
Связь с квантовой фазой достигается идентификацией
$$
\chi\;\hat{=}\;\frac{S}{\hbar}.
$$
Тогда интерференция — интерференция **фазовых траекторий** в хоре, а амплитуды — веса таких траекторий. Из §3.4 (частота $\nu=d\chi/d\tau$) следует
$$
E=\hbar\,\frac{d\chi}{d\tau},\qquad p_l=\hbar\,\frac{d\chi_l}{dl},
$$
что согласуется с де Бройлем при $E=\hbar\omega$, $p=\hbar k$. Квантование — ограничения на допустимые **замкнутые фазы** (условия типа Бора–Зоммерфельда) в конфигурациях с компактной $\chi$.

### 5.2. Калибровочные симметрии как перепараметризации $\chi$
Глобальный сдвиг $\chi\mapsto\chi+\chi_0$ ненаблюдаем. Разрешая **локальные** перепараметризации $\chi\mapsto\chi+\alpha(x)$, требуем когерентности сравнения фаз в разных точках — это порождает связность (калибровочное поле). На уровне волновой функции $\psi\sim e^{i\chi}$:
$$
\psi(x)\to e^{i\alpha(x)}\psi(x),\qquad \partial_\mu\to D_\mu=\partial_\mu-iA_\mu.
$$
Здесь $A_\mu$ — геометрическая «связка» между фазовыми слоями; минимальное взаимодействие возникает как требование согласованности фазового переноса.

### 5.3. Масса и внутренний угол $\zeta$
В §3.1 введён внутренний угол $\zeta$, задающий разложение фазы. Гипотеза: **масса** — инвариант, связанный с неустранимой вещественной проекцией $\cos\zeta\neq0$.
- безмассовые: $\zeta=\pm\frac{\pi}{2}$ (локальное время «заменяется» пространственным бегом; фотонное подпространство);
- массивные: $|\zeta|<\frac{\pi}{2}$, собственное время существует; внешнее замедление через $\gamma=\sec\theta$ наслаивается на внутренний множитель $\cos\zeta$ (§3.2).
Это открывает путь к фазовой трактовке массогенеза (сопоставление $\zeta$ механизмам типа Хиггса).

### 5.4. ОТО как неоднородная перепараметризация $\chi$
В §1.2 интервал задан через $g(\chi)$. При неоднородном распределении фазы Якобиан перепараметризации (§3.5) становится полем:
$$
\mathcal{J}(x)=\frac{d\chi}{d\tau}(x),\qquad dx^\mu=\tilde{X}^\mu(\chi)\,d\chi,
$$
и метрика наблюдаемого многообразия приобретает кривизну из-за вариаций $\mathcal{J}$ и $\tilde{X}^\mu$. Геодезические — траектории экстремального «фазового времени» (стационарной $\chi$). Гравитационное красное смещение трактуется как доплероподобное масштабирование меры времени в неоднородном поле $\mathcal{J}(x)$.

### 5.5. Космология и «абсолютный» калибр
Калибр «абсолютного времени» фиксируем космологически (см. Appendix): **комовинговая система** с нулевым диполем CMB. Тогда космологическое время FLRW задаёт глобальный калибр $t$; космологический красный сдвиг — накопленный фазовый (доплеровский) фактор на расширяющемся фоне; локально сохраняется лоренц-инвариантность (§§3.6–3.8).

### 5.6. Экспериментальные опоры
Фазовая интерпретация воспроизводит:
- продольный/поперечный **доплер** (§3.8);
- замедление времени (мионы, атомные часы, GPS);
- релятивистское сложение скоростей (§3.7).
Дальнейшие тесты: высокоточная спектроскопия при контролируемых $\theta$ и, по возможности, с управляемой эффективной «плотностью» (углом $\zeta$) в конденсированных системах/оптических решётках.

**Итог.** Униметрия не меняет законов, а переупорядочивает их в едином фазовом языке. В нём доплеровский сдвиг, лоренц-фактор, калибровочная связь и (эвристически) гравитация оказываются аспектами одной и той же геометрии перепараметризации $\chi$.
